## Why requests?

The `requests` library is the de facto standard for HTTP in Python. It's simpler than `urllib` and handles cookies, redirects, and encoding automatically.

```python
# Install: pip install requests
import requests

response = requests.get("https://httpbin.org/get")
print(f"Status: {response.status_code}")
print(f"Headers: {dict(response.headers)}")
print(f"JSON: {response.json()}")
```

## GET with Query Parameters

Pass `params` as a dict; requests encodes them automatically.

```python
response = requests.get(
    "https://httpbin.org/get",
    params={"page": 1, "limit": 10, "search": "python"}
)
print(f"Final URL: {response.url}")
data = response.json()
print(f"Args received: {data.get('args')}")
```

## Custom Headers

Pass `headers` as a dict. Common: User-Agent, Authorization, Accept, Content-Type.

```python
headers = {
    "User-Agent": "MyApp/1.0",
    "Accept": "application/json",
    "Authorization": "Bearer fake-token-123"
}
response = requests.get("https://httpbin.org/headers", headers=headers)
print(response.json())
```

## POST Requests

Use `data` for form-encoded, `json` for JSON body. requests sets Content-Type automatically for `json=`.

```python
# POST JSON
payload = {"name": "Alice", "email": "alice@example.com"}
response = requests.post("https://httpbin.org/post", json=payload)
print(f"Status: {response.status_code}")
print(response.json().get("json"))
```

```python
# POST form data (application/x-www-form-urlencoded)
form_data = {"username": "alice", "password": "secret"}
response = requests.post("https://httpbin.org/post", data=form_data)
print(response.json().get("form"))
```

## Sessions

Use `requests.Session()` to reuse TCP connections and persist cookies across requests.

```python
session = requests.Session()
session.headers.update({"User-Agent": "SessionApp/1.0"})

# All requests share the same session
r1 = session.get("https://httpbin.org/cookies/set?name=value")
r2 = session.get("https://httpbin.org/cookies")
print("Cookies:", r2.json())
```

## Error Handling

Check `response.raise_for_status()` to raise on 4xx/5xx. Use try/except for `requests.RequestException`.

```python
try:
    response = requests.get("https://httpbin.org/status/404", timeout=5)
    response.raise_for_status()  # Raises HTTPError for 4xx/5xx
except requests.exceptions.HTTPError as e:
    print(f"HTTP Error: {e}")
except requests.exceptions.Timeout:
    print("Request timed out")
except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
```

## Timeouts

Always set `timeout` to avoid hanging. Can use a tuple `(connect_timeout, read_timeout)`.

```python
response = requests.get("https://httpbin.org/delay/1", timeout=5)
print(f"Got response in time: {response.status_code}")
```

## Response Object Attributes

- `response.text` - body as string
- `response.content` - body as bytes
- `response.json()` - parsed JSON (raises if invalid)
- `response.headers` - dict-like headers
- `response.status_code` - integer status

```python
r = requests.get("https://httpbin.org/json")
print("Content-Type:", r.headers.get("Content-Type"))
print("Text (first 100):", r.text[:100])
print("JSON keys:", list(r.json().keys()))
```

## Common Mistakes

- **Not using timeout**: Can hang indefinitely; always set `timeout=5` or similar
- **Ignoring raise_for_status()**: 404 returns a Response object; check status or use raise_for_status()
- **Using data= instead of json=**: For JSON APIs, use `json=payload` so Content-Type is set correctly
- **Not closing sessions**: Use `with requests.Session() as s:` or session.close() when done
- **Storing passwords in code**: Use environment variables for secrets
