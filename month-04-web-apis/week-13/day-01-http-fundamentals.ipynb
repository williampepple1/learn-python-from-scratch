## What is HTTP?

**HyperText Transfer Protocol** is the foundation of data communication on the web. It's a request-response protocol: the client sends a request, the server sends a response.

## HTTP Methods

| Method | Purpose | Idempotent? |
|--------|---------|-------------|
| GET | Retrieve a resource | Yes |
| POST | Create a resource | No |
| PUT | Replace a resource | Yes |
| PATCH | Partially update | No |
| DELETE | Remove a resource | Yes |

```python
# Simulating HTTP request structure (conceptual)
request = {
    "method": "GET",
    "path": "/api/users/1",
    "headers": {
        "Host": "api.example.com",
        "Accept": "application/json",
        "User-Agent": "Python/3.10"
    },
    "body": None
}
print("GET request:", request["method"], request["path"])
```

## HTTP Status Codes

Status codes indicate the result of the request:
- **2xx** Success (200 OK, 201 Created)
- **3xx** Redirection (301 Moved, 304 Not Modified)
- **4xx** Client error (400 Bad Request, 401 Unauthorized, 404 Not Found)
- **5xx** Server error (500 Internal Server Error)

```python
STATUS_CODES = {
    200: "OK", 201: "Created", 204: "No Content",
    301: "Moved Permanently", 304: "Not Modified",
    400: "Bad Request", 401: "Unauthorized", 404: "Not Found", 500: "Internal Server Error"
}
for code, msg in STATUS_CODES.items():
    print(f"{code}: {msg}")
```

## URL Structure

Use urllib.parse for URL parsing and encoding.

```python
from urllib.parse import urlparse, parse_qs
url = "https://api.example.com:443/users?page=1&limit=10"
parsed = urlparse(url)
print("Scheme:", parsed.scheme)
print("Host:", parsed.hostname)
print("Path:", parsed.path)
print("Query:", parse_qs(parsed.query))
```

## Common Mistakes

- Confusing GET vs POST: GET for reading, POST for creating
- Ignoring status codes: Always check response.status_code
- Forgetting Content-Type for JSON APIs
- URL encoding: Use urllib.parse.urlencode for query params
