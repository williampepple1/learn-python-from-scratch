## Why BeautifulSoup?

BeautifulSoup parses HTML/XML and lets you navigate the document tree. Works well with `requests` for fetching pages.

```python
# pip install beautifulsoup4 requests
from bs4 import BeautifulSoup

html = """
<html>
  <head><title>Test Page</title></head>
  <body>
    <h1>Hello</h1>
    <p class="intro">First paragraph</p>
    <p>Second paragraph</p>
    <a href="/link">Click</a>
  </body>
</html>
"""
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify()[:300])
```

## Finding Elements by Tag

- `soup.find(tag)` - first match
- `soup.find_all(tag)` - all matches
- `soup.tag` - shortcut for first tag (e.g. `soup.head`)

```python
soup = BeautifulSoup(html, "html.parser")
print("Title:", soup.title.string)
print("First p:", soup.find("p").get_text())
print("All p:", [p.get_text() for p in soup.find_all("p")])
```

## Finding by Class and ID

Use `class_` (underscore) because `class` is a Python keyword.

```python
intro = soup.find("p", class_="intro")
print(intro.get_text())

# Multiple classes
html2 = '<p class="a b">Text</p>'
soup2 = BeautifulSoup(html2, "html.parser")
p = soup2.find("p", class_="a")
print(p.get_text())
```

## CSS Selectors

Use `select()` and `select_one()` with CSS selector syntax:
- `div` - tag
- `.class` - class
- `#id` - id
- `div p` - descendant
- `div > p` - direct child

```python
html = """
<div id="main">
  <ul class="items">
    <li><a href="/1">Link 1</a></li>
    <li><a href="/2">Link 2</a></li>
  </ul>
</div>
"""
soup = BeautifulSoup(html, "html.parser")
links = soup.select("#main .items a")
for a in links:
    print(a.get_text(), a["href"])
```

## Extracting Attributes

Use `tag["attr"]` or `tag.get("attr")` for optional attributes.

```python
soup = BeautifulSoup('<a href="https://example.com" data-id="1">Link</a>', "html.parser")
a = soup.find("a")
print("href:", a["href"])
print("data-id:", a.get("data-id"))
print("Text:", a.string)
```

## Fetching and Parsing a Real Page

Combine requests + BeautifulSoup. Always respect robots.txt and rate limits.

```python
import requests
from bs4 import BeautifulSoup

# Example: parse a simple page
url = "https://example.com"
response = requests.get(url, timeout=5)
response.raise_for_status()
soup = BeautifulSoup(response.text, "html.parser")
print("Title:", soup.title.string)
print("First paragraph:", soup.find("p").get_text()[:100])
```

## Handling Missing Elements

`find()` returns None if not found. Check before accessing.

```python
soup = BeautifulSoup("<div>No links</div>", "html.parser")
link = soup.find("a")
if link:
    print(link["href"])
else:
    print("No link found")

# Safe extraction with default
title = soup.find("title")
print("Title:", title.string if title else "N/A")
```

## Web Scraping Ethics

- Check robots.txt
- Respect rate limits (add delays)
- Use a proper User-Agent
- Prefer APIs when available
- Don't scrape personal/sensitive data without permission

## Common Mistakes

- **Using `class` instead of `class_`**: Python keyword conflict
- **Assuming find() always returns**: Check for None before accessing attributes
- **Not specifying parser**: Use `html.parser` (built-in) or `lxml` (faster)
- **Scraping too aggressively**: Add `time.sleep(1)` between requests
- **Ignoring encoding**: Use `response.content` and pass encoding if needed
