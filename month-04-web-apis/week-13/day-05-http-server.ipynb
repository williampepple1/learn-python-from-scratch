## The http.server Module

Python's built-in http.server provides simple HTTP servers. Good for development, not for production.

```python
from http.server import HTTPServer, BaseHTTPRequestHandler

class MyHandler(BaseHTTPRequestHandler):
    def do_GET(self):
        self.send_response(200)
        self.send_header("Content-Type", "text/plain")
        self.end_headers()
        self.wfile.write(b"Hello from Python HTTP server!")

print("Use HTTPServer(('', 8000), MyHandler).serve_forever() to run")
```

## Routing by Path

Use self.path to route different URLs.

```python
from http.server import BaseHTTPRequestHandler

class RouterHandler(BaseHTTPRequestHandler):
    def do_GET(self):
        if self.path == "/":
            self.send_html("<h1>Home</h1>")
        elif self.path == "/about":
            self.send_html("<h1>About</h1>")
        else:
            self.send_error(404)

    def send_html(self, html):
        self.send_response(200)
        self.send_header("Content-Type", "text/html")
        self.end_headers()
        self.wfile.write(html.encode())

print("RouterHandler with / and /about")
```

## WSGI

WSGI is a standard interface. An app is a callable: app(environ, start_response).

```python
def simple_wsgi_app(environ, start_response):
    status = "200 OK"
    headers = [("Content-Type", "text/plain")]
    start_response(status, headers)
    return [b"Hello WSGI!"]

print("WSGI app: callable(environ, start_response) -> iterable of bytes")
```

## Common Mistakes

- Using http.server in production: Use gunicorn instead
- Not encoding strings: self.wfile.write expects bytes
- Forgetting Content-Length for POST body
