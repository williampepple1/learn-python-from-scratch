## What is REST?

**Representational State Transfer** - an architectural style for APIs. Key ideas: resources as URLs, HTTP methods for actions, stateless communication.

## REST Principles

- **Resources**: Everything is a resource (users, products) identified by URL
- **HTTP methods**: GET (read), POST (create), PUT/PATCH (update), DELETE (remove)
- **Stateless**: Each request contains all info; no server-side session
- **Representations**: JSON (or XML) for data exchange

```python
# Typical REST endpoints for a "users" resource
endpoints = {
    "GET /users": "List all users",
    "GET /users/1": "Get user with id 1",
    "POST /users": "Create user",
    "PUT /users/1": "Replace user 1",
    "PATCH /users/1": "Partially update user 1",
    "DELETE /users/1": "Delete user 1"
}
for ep, desc in endpoints.items():
    print(f"{ep}: {desc}")
```

## JSON in Python

Use `json` module: `json.loads()` (string to dict) and `json.dumps()` (dict to string).

```python
import json

# Parse JSON string
json_str = '{"name": "Alice", "age": 30, "active": true}'
data = json.loads(json_str)
print(type(data), data)

# Serialize to JSON
obj = {"id": 1, "items": ["a", "b"]}
json_out = json.dumps(obj, indent=2)
print(json_out)
```

## JSON with Files and APIs

`json.load()` / `json.dump()` for file handles. APIs typically return JSON.

```python
import json

# Write to file
with open("temp_data.json", "w") as f:
    json.dump({"key": "value"}, f)

# Read from file
with open("temp_data.json") as f:
    loaded = json.load(f)
print(loaded)

# Clean up
import os
os.remove("temp_data.json")
```

## Consuming a REST API

httpbin.org provides test endpoints. Real APIs: GitHub, OpenWeatherMap, etc.

```python
import requests

response = requests.get("https://httpbin.org/json")
data = response.json()  # Same as json.loads(response.text)
print("Keys:", data.keys())
slideshow = data.get("slideshow", {})
print("Author:", slideshow.get("author"))
```

## API Versioning

Common patterns: `/v1/users` or `Accept: application/vnd.api.v1+json`

```python
# Version in URL
base_url = "https://api.example.com/v1"
users_url = f"{base_url}/users"
print(users_url)

# Version in header
headers = {"Accept": "application/vnd.api+json;version=1"}
print(headers)
```

## Pagination

APIs often paginate: `?page=1&limit=10` or `?offset=0&limit=10`

```python
def fetch_page(base_url, page=1, per_page=10):
    url = f"{base_url}?page={page}&per_page={per_page}"
    return url

print(fetch_page("https://api.example.com/users"))
print(fetch_page("https://api.example.com/users", page=2))
```

## Error Responses

APIs return error JSON with status 4xx/5xx. Structure often: `{"error": "message", "code": 400}`

```python
response = requests.get("https://httpbin.org/status/404")
print("Status:", response.status_code)
if not response.ok:
    try:
        err = response.json()
        print("Error body:", err)
    except:
        print("Error text:", response.text[:200])
```

## Common Mistakes

- **Not handling JSON errors**: `response.json()` can fail on non-JSON responses
- **Ignoring status codes**: Check before parsing; 200 doesn't always mean success in some APIs
- **Hardcoding base URLs**: Use config or env vars
- **No rate limiting**: Respect API limits; use exponential backoff
- **Storing API keys in code**: Use environment variables
