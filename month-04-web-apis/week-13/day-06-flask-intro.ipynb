## Why Flask?

Flask is a lightweight WSGI web framework. Easy to learn, flexible, and great for APIs and small web apps.

```python
# pip install flask
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return "Hello, Flask!"

if __name__ == "__main__":
    app.run(debug=True, port=5000)
# Run: python app.py or flask run
```

## Routing

Use `@app.route("/path")` to map URLs to functions. Add `methods=["POST"]` for POST.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def index():
    return "<h1>Home</h1>"

@app.route("/about")
def about():
    return "About page"

@app.route("/api/hello")
def api_hello():
    return {"message": "Hello"}  # Flask auto-converts dict to JSON

print("Routes: /, /about, /api/hello")
```

## URL Parameters

Use `<variable>` or `<int:variable>` in the route. Pass as function argument.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/user/<name>")
def user(name):
    return f"Hello, {name}!"

@app.route("/post/<int:post_id>")
def post(post_id):
    return f"Post #{post_id}"

# /user/Alice -> Hello, Alice!
# /post/42 -> Post #42
print("Dynamic routes: /user/<name>, /post/<int:post_id>")
```

## Query Parameters

Use `request.args.get("key")` for query string parameters.

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/search")
def search():
    q = request.args.get("q", "")
    page = request.args.get("page", 1, type=int)
    return f"Search: q={q}, page={page}"

# /search?q=python&page=2
print("Use request.args.get('key', default)")
```

## Request Object

- `request.method` - GET, POST, etc.
- `request.form` - form data
- `request.json` - JSON body
- `request.headers` - headers

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/api/echo", methods=["POST"])
def echo():
    data = request.get_json(silent=True) or {}
    return {"received": data}

print("POST /api/echo returns received JSON")
```

## Returning JSON

Return a dict (Flask converts) or use `jsonify()` for full control.

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route("/api/users/1")
def get_user():
    return jsonify({"id": 1, "name": "Alice"})

@app.route("/api/error")
def error():
    return jsonify({"error": "Not found"}), 404

print("jsonify() for JSON; (body, status_code) for custom status")
```

## Templates with render_template

Create a `templates/` folder. Use `render_template("name.html", var=value)`.

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/greet/<name>")
def greet(name):
    return render_template("greet.html", name=name)

# templates/greet.html: <h1>Hello, {{ name }}!</h1>
print("Templates go in templates/ folder. Use {{ var }} in Jinja2.")
```

```python
# Create a minimal template inline (for demo - normally in templates/)
import os
os.makedirs("templates", exist_ok=True)
with open("templates/greet.html", "w") as f:
    f.write("<h1>Hello, {{ name }}!</h1>")
print("Created templates/greet.html")
```

## Common Mistakes

- **Forgetting methods=["POST"]**: Default is GET only
- **Not using request.get_json(silent=True)**: Can raise if body isn't JSON
- **Templates in wrong folder**: Must be in `templates/` relative to app
- **Running with debug=True in production**: Security risk
- **Circular imports**: Keep app in one module or use application factory
