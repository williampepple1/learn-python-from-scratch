## Production Tips

- Use connection pooling
- Set statement_timeout
- Use SSL for connections
- Regular backups
- Monitor slow queries

## Common Mistakes

- **Wrong placeholder**: PostgreSQL uses %s with psycopg2, $1 with asyncpg
- **Leaking connections**: Always close or use context manager
- **No connection pooling**: Creating new connection per request is slow
- **Credentials in code**: Use env vars or secrets manager
- **Sync in async app**: Use asyncpg with FastAPI async routes
