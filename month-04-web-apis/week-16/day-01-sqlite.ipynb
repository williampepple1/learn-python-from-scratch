## Why SQLite?

SQLite is file-based, no server needed. Built into Python. Great for dev, prototypes, and small apps.

```python
import sqlite3

conn = sqlite3.connect("demo.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE
)
""")
conn.commit()
print("Table created")
```

## INSERT

Use parameterized queries with `?` placeholders. Never concatenate user input!

```python
cursor.execute("INSERT INTO users (name, email) VALUES (?, ?)", ("Alice", "alice@example.com"))
cursor.execute("INSERT INTO users (name, email) VALUES (?, ?)", ("Bob", "bob@example.com"))
conn.commit()
print("Inserted:", cursor.lastrowid)
```

## SELECT

`cursor.execute()` then `cursor.fetchone()`, `fetchall()`, or `fetchmany(n)`.

```python
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
for row in rows:
    print(row)

cursor.execute("SELECT name, email FROM users WHERE id = ?", (1,))
one = cursor.fetchone()
print("One:", one)
```

## UPDATE and DELETE

Same pattern: execute with params, commit.

```python
cursor.execute("UPDATE users SET email = ? WHERE id = ?", ("alice.new@example.com", 1))
conn.commit()
print("Updated:", cursor.rowcount)

# cursor.execute("DELETE FROM users WHERE id = ?", (2,))
# conn.commit()
```

## Row Factory

Use `row_factory` to get dict-like access.

```python
conn.row_factory = sqlite3.Row
cursor = conn.cursor()
cursor.execute("SELECT * FROM users LIMIT 1")
row = cursor.fetchone()
print("By name:", row["name"])
print("Keys:", row.keys())
```

## Context Manager

Use `with` for automatic commit/rollback and close.

```python
with sqlite3.connect("demo.db") as conn:
    cursor = conn.cursor()
    cursor.execute("SELECT COUNT(*) FROM users")
    print("Count:", cursor.fetchone()[0])
# conn closed automatically
```

## executemany

Insert multiple rows efficiently.

```python
data = [("Carol", "carol@example.com"), ("Dave", "dave@example.com")]
with sqlite3.connect("demo.db") as conn:
    conn.executemany("INSERT INTO users (name, email) VALUES (?, ?)", data)
print("executemany for batch insert")
```

## Common Mistakes

- **SQL injection**: Always use ? placeholders, never f-strings with user input
- **Forgetting commit()**: Changes not persisted without commit
- **Not closing connection**: Use context manager or conn.close()
- **Shared connection across threads**: SQLite has threading limitations; use one connection per thread
- **Storing binary**: Use BLOB type; encode/decode as needed
