## Why Migrations?

Migrations track schema changes over time. Enable rollback, team collaboration, and reproducible deploys.

## Create Migration

`alembic revision -m "add users table"` creates a new migration file.

```python
# Migration file has upgrade() and downgrade()
migration_example = """
def upgrade():
    op.create_table(
        'users',
        sa.Column('id', sa.Integer(), primary_key=True),
        sa.Column('name', sa.String(50), nullable=False),
        sa.Column('email', sa.String(100))
    )

def downgrade():
    op.drop_table('users')
"""
print("upgrade() and downgrade() in each revision")
```

## Auto-generate from Models

`alembic revision --autogenerate -m "message"` compares models to DB and generates migration.

```python
# In env.py, set target_metadata = Base.metadata
# from app.models import Base
# target_metadata = Base.metadata
#
# Then: alembic revision --autogenerate -m "add posts table"
print("target_metadata = Base.metadata for autogenerate")
```

## Run Migrations

`alembic upgrade head` - apply all. `alembic downgrade -1` - rollback one.

```python
# alembic upgrade head      # Apply all pending
# alembic upgrade +1        # Upgrade one
# alembic downgrade -1      # Downgrade one
# alembic downgrade base    # Back to start
# alembic current           # Show current revision
# alembic history          # Show history
print("upgrade head, downgrade -1, current, history")
```

## Best Practices

- One logical change per migration
- Test downgrade
- Don't edit applied migrations; create new one
- Backup before migration in production

## Common Mistakes

- **Editing old migrations**: Create new migration to fix
- **Not testing downgrade**: Downgrade might fail
- **Large data migrations**: Split schema change and data migration
- **Forgetting target_metadata**: Autogenerate won't see models
- **Running migrations manually**: Use alembic, not raw SQL
