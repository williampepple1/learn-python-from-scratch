## Why Docker?

Docker containers package app + dependencies. Consistent across dev/staging/prod. Isolated, reproducible.

```python
# Dockerfile example
dockerfile_content = """
FROM python:3.10-slim
WORKDIR /app
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt
COPY . .
EXPOSE 8000
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
"""
print(dockerfile_content)
```

## Dockerfile Instructions

- FROM: base image
- WORKDIR: set working directory
- COPY: copy files into image
- RUN: execute command during build
- CMD: default command when container runs
- EXPOSE: document port (doesn't publish)

```python
with open("Dockerfile.example", "w") as f:
    f.write("""FROM python:3.10-slim
WORKDIR /app
COPY requirements.txt .
RUN pip install -r requirements.txt
COPY . .
CMD ["python", "-m", "http.server", "8000"]
""")
print("Created Dockerfile.example")
```

## docker-compose

Define multi-container apps. App + database + redis.

```python
compose_content = """
version: "3.8"
services:
  web:
    build: .
    ports:
      - "8000:8000"
    environment:
      - DATABASE_URL=postgresql://user:pass@db:5432/mydb
      - REDIS_URL=redis://redis:6379
    depends_on:
      - db
      - redis
  db:
    image: postgres:15
    environment:
      POSTGRES_USER: user
      POSTGRES_PASSWORD: pass
      POSTGRES_DB: mydb
    volumes:
      - pgdata:/var/lib/postgresql/data
  redis:
    image: redis:7-alpine
volumes:
  pgdata:
"""
print(compose_content)
```

## .dockerignore

Exclude files from build context. Speeds build, reduces image size.

```python
dockerignore = """
__pycache__
*.pyc
.git
.env
venv
*.db
"""
print(".dockerignore excludes unnecessary files")
```

## Multi-stage Build

Build in one stage, copy artifacts to smaller final image.

```python
multistage = """
FROM python:3.10 as builder
WORKDIR /app
COPY requirements.txt .
RUN pip install --user -r requirements.txt

FROM python:3.10-slim
WORKDIR /app
COPY --from=builder /root/.local /root/.local
COPY . .
ENV PATH=/root/.local/bin:$PATH
CMD ["uvicorn", "main:app", "--host", "0.0.0.0"]
"""
print("Multi-stage: smaller final image")
```

## Common Mistakes

- **Copying too much**: Use .dockerignore
- **Running as root**: Add USER in Dockerfile
- **Hardcoding secrets**: Use env vars or secrets
- **Not using slim images**: python:3.10-slim vs python:3.10
- **Forgetting depends_on**: DB may not be ready when app starts
