## What is CI/CD?

**Continuous Integration**: Run tests on every push. **Continuous Deployment**: Deploy automatically when tests pass.

```python
# .github/workflows/test.yml
workflow = """
name: Test
on:
  push:
    branches: [main, develop]
  pull_request:
    branches: [main]
jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with:
          python-version: '3.10'
      - run: pip install -r requirements.txt
      - run: pytest
"""
print(workflow)
```

## Basic Test Workflow

Checkout, setup Python, install deps, run tests.

```python
import os
os.makedirs(".github/workflows", exist_ok=True)

with open(".github/workflows/test.yml", "w") as f:
    f.write("""name: Test
on: [push, pull_request]
jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with:
          python-version: '3.10'
      - run: pip install pytest
      - run: pytest tests/ -v
""")
print("Created .github/workflows/test.yml")
```

## Caching Dependencies

Cache pip packages to speed up runs.

```python
cache_step = """
      - uses: actions/cache@v4
        with:
          path: ~/.cache/pip
          key: pip-${{ hashFiles('requirements.txt') }}
"""
print("Cache pip for faster installs")
```

## Lint and Format

Run ruff, black, or flake8.

```python
lint_workflow = """
      - run: pip install ruff black
      - run: ruff check .
      - run: black --check .
"""
print("Lint: ruff, black --check")
```

## Matrix Testing

Test on multiple Python versions.

```python
matrix = """
    strategy:
      matrix:
        python-version: ['3.9', '3.10', '3.11']
    steps:
      - uses: actions/setup-python@v5
        with:
          python-version: ${{ matrix.python-version }}
"""
print("Matrix: test on 3.9, 3.10, 3.11")
```

## Deployment (Example)

Deploy to cloud on push to main. Use secrets for credentials.

```python
deploy = """
  deploy:
    needs: test
    runs-on: ubuntu-latest
    if: github.ref == 'refs/heads/main'
    steps:
      - uses: actions/checkout@v4
      - name: Deploy
        env:
          API_KEY: ${{ secrets.API_KEY }}
        run: echo "Deploy with API_KEY"
"""
print("needs: test, secrets.API_KEY")
```

## Common Mistakes

- **No requirements.txt**: CI can't install deps
- **Hardcoding secrets**: Use GitHub Secrets
- **Skipping cache**: Slow builds
- **Not failing on lint**: Add lint job that fails
- **Deploying without test**: Use needs: test
