## RESTful Resource Design

Map HTTP methods to CRUD: GET (read), POST (create), PUT/PATCH (update), DELETE.

```python
from flask import Flask, jsonify, request

app = Flask(__name__)
items = {1: {"id": 1, "name": "Item 1", "price": 10}}
next_id = 2

@app.route("/api/items", methods=["GET"])
def list_items():
    return jsonify(list(items.values()))

@app.route("/api/items", methods=["POST"])
def create_item():
    global next_id
    data = request.get_json() or {}
    item = {"id": next_id, "name": data.get("name", ""), "price": data.get("price", 0)}
    items[next_id] = item
    next_id += 1
    return jsonify(item), 201

print("REST: GET list, POST create")
```

## Single Resource and Error Handlers

GET/PUT/DELETE for /api/items/<id>. Custom error handlers.

```python
@app.route("/api/items/<int:id>", methods=["GET"])
def get_item(id):
    if id not in items:
        return jsonify({"error": "Not found"}), 404
    return jsonify(items[id])

@app.errorhandler(404)
def not_found(e):
    return jsonify({"error": "Not found"}), 404

print("Error handlers for consistent API responses")
```

## Common Mistakes

- Not using 201 for created
- Skipping validation of get_json()
- No rate limiting in production
