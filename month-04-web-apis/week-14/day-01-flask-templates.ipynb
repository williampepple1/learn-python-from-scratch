## Jinja2 Basics

Flask uses Jinja2 for templating. Variables: `{{ name }}`, logic: `{% if %}`, loops: `{% for %}`.

```python
from jinja2 import Template

t = Template("Hello, {{ name }}!")
print(t.render(name="Alice"))

t2 = Template("{% for x in items %}{{ x }} {% endfor %}")
print(t2.render(items=[1, 2, 3]))
```

## Template Inheritance

Base template defines `{% block content %}`. Child extends with `{% extends "base.html" %}` and `{% block content %}...{% endblock %}`.

```python
base_html = """
<html>
<head><title>{% block title %}My App{% endblock %}</title></head>
<body>
  {% block content %}{% endblock %}
</body>
</html>
"""

page_html = """
{% extends "base" %}
{% block title %}Home{% endblock %}
{% block content %}<h1>Welcome</h1>{% endblock %}
"""

from jinja2 import Environment, DictLoader
env = Environment(loader=DictLoader({"base": base_html, "page": page_html}))
t = env.get_template("page")
print(t.render())
```

## Creating Base Template

Create `templates/base.html` with common structure.

```python
import os
os.makedirs("templates", exist_ok=True)

base_content = """<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8">
  <title>{% block title %}Flask App{% endblock %}</title>
  <link rel="stylesheet" href="{{ url_for('static', filename='style.css') }}">
</head>
<body>
  <nav>{% block nav %}<a href="/">Home</a>{% endblock %}</nav>
  <main>{% block content %}{% endblock %}</main>
</body>
</html>"""

with open("templates/base.html", "w") as f:
    f.write(base_content)
print("Created templates/base.html")
```

## Child Template

Create `templates/home.html` that extends base.

```python
home_content = """{% extends "base.html" %}
{% block title %}Home{% endblock %}
{% block content %}
  <h1>Welcome</h1>
  <p>Hello, {{ username }}!</p>
{% endblock %}"""

with open("templates/home.html", "w") as f:
    f.write(home_content)
print("Created templates/home.html")
```

## url_for for Links

`url_for('view_name')` generates URLs from route names. Use for static files: `url_for('static', filename='style.css')`.

```python
from flask import Flask, render_template, url_for

app = Flask(__name__)

@app.route("/")
def index():
    return render_template("home.html", username="Alice")

with app.test_request_context():
    print("url_for('index'):", url_for("index"))
    print("url_for('static', filename='style.css'):", url_for("static", filename="style.css"))
```

## Static Files

Put CSS/JS in `static/` folder. Reference with `url_for('static', filename='file.css')`.

```python
os.makedirs("static", exist_ok=True)
with open("static/style.css", "w") as f:
    f.write("body { font-family: sans-serif; }\nh1 { color: #333; }")
print("Created static/style.css")
```

## Filters

`{{ name|upper }}`, `{{ items|length }}`, `{{ value|default('N/A') }}`

```python
from jinja2 import Template
t = Template("{{ name|upper }} has {{ items|length }} items. Score: {{ score|default(0) }}")
print(t.render(name="alice", items=[1,2,3]))
print(t.render(name="bob", items=[], score=None))
```

## Loops and Conditionals

`{% for item in items %}`, `{% if condition %}`, `{% else %}`, `{% endif %}`, `{% endfor %}`

```python
html = """
{% for user in users %}
  <li>{{ user.name }} - {{ user.role }}</li>
{% else %}
  <p>No users</p>
{% endfor %}
{% if admin %}<p>Admin view</p>{% endif %}
"""
t = Template(html)
print(t.render(users=[{"name": "A", "role": "user"}], admin=True))
```

## Common Mistakes

- **Wrong template path**: Templates must be in `templates/` folder
- **Block name mismatch**: Child block names must match base
- **Forgetting url_for for static**: Hardcoding `/static/` can break with blueprints
- **XSS**: Use `{{ value }}` (auto-escaped) not `{{ value|safe }}` for user input
- **Circular extends**: Don't create circular inheritance
