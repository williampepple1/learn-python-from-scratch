## What is a Blueprint?

A Blueprint is a way to organize routes, templates, and static files into reusable modules. Enables larger, maintainable applications.

````python
from flask import Blueprint

auth_bp = Blueprint("auth", __name__, url_prefix="/auth")

@auth_bp.route("/login")
def login():
    return "Login page"

@auth_bp.route("/logout")
def logout():
    return "Logout"

print("Blueprint 'auth' with /auth/login, /auth/logout")
````

## Registering Blueprints

Call `app.register_blueprint(bp)` to add routes to the app.

````python
from flask import Flask

app = Flask(__name__)
app.register_blueprint(auth_bp)

@app.route("/")
def index():
    return "Home"

with app.test_request_context():
    from flask import url_for
    print("auth.login:", url_for("auth.login"))
    print("auth.logout:", url_for("auth.logout"))
````

## Application Factory

Create the app in a function. Enables testing, multiple instances, and config loading.

````python
def create_app(config=None):
    app = Flask(__name__)
    app.config.from_object(config or "config.DefaultConfig")
    app.register_blueprint(auth_bp)
    return app

app = create_app()
print("create_app() returns configured Flask app")
````

## Blueprint with Templates

Use `template_folder` to scope templates to the blueprint.

````python
users_bp = Blueprint("users", __name__, url_prefix="/users", template_folder="templates")

from flask import render_template

@users_bp.route("/<int:id>")
def profile(id):
    return render_template("users/profile.html", user_id=id)

# Templates: users/templates/users/profile.html or app templates/users/profile.html
print("Blueprint can have its own template_folder")
````

## url_for with Blueprints

Use `url_for('blueprint_name.view_name')` for blueprint routes.

````python
with app.test_request_context():
    print(url_for("auth.login"))  # /auth/login
    print(url_for("auth.logout"))  # /auth/logout
````

## Before/After Request (Blueprint-scoped)

Use `@bp.before_request` for logic that runs before each request in that blueprint.

````python
@auth_bp.before_request
def require_auth():
    pass  # Could check login, etc.

print("Blueprint-level before_request available")
````

## Common Mistakes

- **Circular imports**: Define blueprint first, then import routes
- **Wrong url_for**: Use 'blueprint.view' not just 'view' for blueprint routes
- **Forgetting to register**: app.register_blueprint(bp) must be called
- **Duplicate url_prefix**: Can conflict if not careful
- **Template path**: Blueprint templates need correct folder structure
