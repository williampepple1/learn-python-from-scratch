## Basic Form Handling

Use `request.form` for POST form data. Keys are the `name` attributes.

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        username = request.form.get("username", "")
        password = request.form.get("password", "")
        return f"Submitted: {username}"
    return """
    <form method="POST">
      <input name="username" placeholder="Username">
      <input name="password" type="password" placeholder="Password">
      <button type="submit">Login</button>
    </form>
    """

print("Basic form: request.form.get('key')")
```

## WTForms

WTForms provides validation, CSRF, and clean form definitions. `pip install flask-wtf`

```python
from flask_wtf import FlaskForm
from wtforms import StringField, PasswordField, BooleanField, SubmitField
from wtforms.validators import DataRequired, Email, Length

class LoginForm(FlaskForm):
    username = StringField("Username", validators=[DataRequired(), Length(min=3, max=20)])
    password = PasswordField("Password", validators=[DataRequired()])
    remember = BooleanField("Remember me", default=False)
    submit = SubmitField("Login")

form = LoginForm()
print("Fields:", [f.name for f in form])
```

## Validators

Common validators: `DataRequired`, `Email`, `Length`, `NumberRange`, `Optional`, `Regexp`

```python
from wtforms import IntegerField
from wtforms.validators import NumberRange, Optional

class ProductForm(FlaskForm):
    name = StringField("Name", validators=[DataRequired()])
    price = IntegerField("Price", validators=[NumberRange(min=0, max=10000)])
    description = StringField("Description", validators=[Optional()])

pf = ProductForm()
pf.name.data = ""
pf.validate()
print("Name errors:", pf.name.errors)
```

## CSRF Protection

Flask-WTF enables CSRF by default. Set `SECRET_KEY` in app config. Include `{{ form.hidden_tag() }}` in forms.

```python
app = Flask(__name__)
app.config["SECRET_KEY"] = "dev-secret-change-in-production"

class MyForm(FlaskForm):
    name = StringField("Name")
    submit = SubmitField("Submit")

# In template: {{ form.hidden_tag() }} includes CSRF token
print("CSRF: Add SECRET_KEY, use form.hidden_tag() in template")
```

## Rendering Forms in Templates

Use `form.field_name` or `form.field_name()` for labels and inputs.

```python
form_html = """
<form method="POST">
  {{ form.hidden_tag() }}
  <p>{{ form.username.label }} {{ form.username() }}</p>
  <p>{{ form.password.label }} {{ form.password() }}</p>
  <p>{{ form.remember.label }} {{ form.remember() }}</p>
  <p>{{ form.submit() }}</p>
</form>
"""
print("Template: form.field.label, form.field(), form.hidden_tag()")
```

## Form Validation in Route

`form.validate_on_submit()` returns True for POST with valid data.

```python
from flask import render_template, redirect, url_for

class RegisterForm(FlaskForm):
    username = StringField("Username", validators=[DataRequired()])
    email = StringField("Email", validators=[DataRequired(), Email()])
    submit = SubmitField("Register")

@app.route("/register", methods=["GET", "POST"])
def register():
    form = RegisterForm()
    if form.validate_on_submit():
        # Process valid form
        return redirect(url_for("success"))
    return render_template("register.html", form=form)

print("validate_on_submit() = POST + form.validate()")
```

## Displaying Errors

`form.field.errors` contains validation error messages.

```python
error_html = """
{% for error in form.username.errors %}
  <span class="error">{{ error }}</span>
{% endfor %}
"""
print("Errors: form.username.errors")
```

## Pre-populating Forms

Set `form.field.data = value` before rendering, or pass `obj=model` to form constructor.

```python
form = LoginForm()
form.username.data = "alice"  # Pre-fill
print("Pre-fill: form.username.data = 'value'")
```

## Common Mistakes

- **Missing SECRET_KEY**: CSRF will fail
- **Forgetting methods=["POST"]**: Form won't submit
- **Not using form.hidden_tag()**: CSRF token missing
- **Using request.form with WTForms**: Use form.field.data instead
- **Not validating**: Always call validate_on_submit() before processing
