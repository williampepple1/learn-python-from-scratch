## Why FastAPI?

FastAPI is a modern, fast web framework with automatic OpenAPI docs, type hints, and async support. Great for APIs.

```python
# pip install fastapi uvicorn
from fastapi import FastAPI

app = FastAPI(title="My API", version="1.0")

@app.get("/")
def root():
    return {"message": "Hello, FastAPI!"}

# Run: uvicorn main:app --reload
print("FastAPI app created. Run: uvicorn main:app --reload")
```

## Path Parameters

Use type hints. FastAPI validates and converts automatically.

```python
@app.get("/users/{user_id}")
def get_user(user_id: int):
    return {"user_id": user_id}

@app.get("/items/{item_name}")
def get_item(item_name: str):
    return {"item": item_name}

print("Path params: {name} in route, name: type in function")
```

## Query Parameters

Add parameters with defaults. Optional with `Optional[str] = None`.

```python
from typing import Optional

@app.get("/items")
def list_items(skip: int = 0, limit: int = 10, q: Optional[str] = None):
    return {"skip": skip, "limit": limit, "q": q}

# /items -> skip=0, limit=10, q=None
# /items?skip=5&limit=3&q=test
print("Query params: function args with defaults")
```

## Automatic Documentation

FastAPI generates Swagger UI at `/docs` and ReDoc at `/redoc`.

```python
# Visit http://localhost:8000/docs for interactive API docs
# Visit http://localhost:8000/redoc for alternative docs
print("Docs: /docs (Swagger), /redoc (ReDoc)")
```

## POST with Request Body

Define a Pydantic model. FastAPI validates and parses JSON.

```python
from pydantic import BaseModel

class Item(BaseModel):
    name: str
    price: float
    in_stock: bool = True

@app.post("/items")
def create_item(item: Item):
    return {"received": item.model_dump()}

print("POST body: Pydantic model as parameter")
```

## Response Model

Use `response_model` to document and validate response.

```python
class UserOut(BaseModel):
    id: int
    name: str

@app.get("/users/1", response_model=UserOut)
def get_user_1():
    return {"id": 1, "name": "Alice"}

print("response_model documents and filters output")
```

## Common Mistakes

- **Wrong param order**: Path params before query params in route definition
- **Not using type hints**: FastAPI needs them for validation and docs
- **Running with flask**: Use uvicorn, not app.run()
- **Forgetting Optional**: Required by default; use Optional for optional params
- **Pydantic model in wrong place**: Define before route that uses it
