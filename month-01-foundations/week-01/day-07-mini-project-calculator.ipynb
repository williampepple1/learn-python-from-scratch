## Step 2: All Basic Operations

Add subtract, multiply, and divide functions.

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero"
    return a / b

# Test all operations
print(add(10, 5))       # 15
print(subtract(10, 5))  # 5
print(multiply(10, 5))  # 50
print(divide(10, 5))    # 2.0
print(divide(10, 0))    # Error message
```

## Step 3: Calculator Function with Operator

Create a single function that takes an operator symbol and two numbers.

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return None  # Signal error
    return a / b

def calculate(operator, a, b):
    """Perform calculation based on operator. Returns result or None for invalid/error."""
    if operator == "+":
        return add(a, b)
    elif operator == "-":
        return subtract(a, b)
    elif operator == "*":
        return multiply(a, b)
    elif operator == "/":
        return divide(a, b)
    else:
        return None

print(calculate("+", 10, 5))   # 15
print(calculate("*", 7, 6))   # 42
```

## Step 5: Complete Calculator with Loop

Full implementation: run repeatedly until user quits.

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return None
    return a / b

def calculate(operator, a, b):
    """Perform calculation. Returns result or None for error."""
    if operator == "+":
        return add(a, b)
    elif operator == "-":
        return subtract(a, b)
    elif operator == "*":
        return multiply(a, b)
    elif operator == "/":
        return divide(a, b)
    else:
        return None

def run_calculator():
    """Run the calculator. Type 'q' to quit."""
    print("=== Python Calculator ===")
    print("Enter 'q' to quit.\n")

    while True:
        try:
            user_input = input("Enter expression (e.g., 5 + 3): ").strip()
            if user_input.lower() == 'q':
                print("Goodbye!")
                break

            parts = user_input.split()
            if len(parts) != 3:
                print("Please enter: number operator number (e.g., 10 + 5)")
                continue

            num1 = float(parts[0])
            operator = parts[1]
            num2 = float(parts[2])

            result = calculate(operator, num1, num2)
            if result is not None:
                print(f"Result: {result}\n")
            else:
                print("Error: Invalid operator or division by zero.\n")

        except ValueError:
            print("Error: Please enter valid numbers.\n")

# Run the calculator (comment out in notebook to avoid blocking)
# run_calculator()

# Demo without blocking:
print("Demo: 10 + 5 =", calculate("+", 10, 5))
print("Demo: 20 / 4 =", calculate("/", 20, 4))
```

## Step 6: Final Complete Calculator

Here's the full calculator ready to run as a script. Copy to a .py file and run with `python calculator.py`.

```python
def add(a, b):
    """Return a + b."""
    return a + b

def subtract(a, b):
    """Return a - b."""
    return a - b

def multiply(a, b):
    """Return a * b."""
    return a * b

def divide(a, b):
    """Return a / b. Returns None if b is 0."""
    if b == 0:
        return None
    return a / b

def calculate(operator, a, b):
    """
    Perform calculation based on operator.
    Supported: +, -, *, /
    Returns result or None for invalid input/division by zero.
    """
    if operator == "+":
        return add(a, b)
    elif operator == "-":
        return subtract(a, b)
    elif operator == "*":
        return multiply(a, b)
    elif operator == "/":
        return divide(a, b)
    else:
        return None

def main():
    print("=== Python Calculator ===")
    print("Enter expressions like: 10 + 5")
    print("Type 'q' to quit.\n")

    while True:
        try:
            user_input = input("> ").strip()
            if user_input.lower() == 'q':
                print("Goodbye!")
                break

            parts = user_input.split()
            if len(parts) != 3:
                print("Usage: number operator number (e.g., 10 + 5)")
                continue

            num1 = float(parts[0])
            operator = parts[1]
            num2 = float(parts[2])

            result = calculate(operator, num1, num2)
            if result is not None:
                print(f"= {result}")
            else:
                print("Error: Invalid operator or division by zero.")

        except ValueError:
            print("Error: Please enter valid numbers.")

if __name__ == "__main__":
    main()
```
