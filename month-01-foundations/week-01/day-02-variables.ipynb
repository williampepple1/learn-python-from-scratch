## Variable Assignment

Variables store values. In Python, you assign with `=`. No need to declare the type!

```python
name = "Alice"
age = 25
height = 5.6
is_student = True

print(name, age, height, is_student)
```

## Naming Conventions (PEP 8)

Python follows PEP 8 style guide. Use `snake_case` for variables and functions.

```python
# Good names
user_name = "Bob"
total_count = 42
is_active = True

# Avoid: single letters (except i, j in loops), unclear abbreviations
# Bad: UserName (use for classes), totalCount (camelCase is for other languages)
```

```python
# Valid but reserved - don't use: and, or, if, else, for, while, def, class, etc.
# Valid names can have letters, numbers, underscores (but not start with number)
my_var_2 = 10  # OK
# 2nd_var = 10  # Error!
```

## Dynamic Typing

Python variables can change type. The same name can hold different types over time.

```python
x = 10
print(x, type(x))

x = "hello"
print(x, type(x))

x = [1, 2, 3]
print(x, type(x))
```

## Multiple Assignment

Assign multiple variables in one line.

```python
# Same value to multiple variables
a = b = c = 0
print(a, b, c)
```

```python
# Different values (tuple unpacking)
x, y, z = 1, 2, 3
print(x, y, z)
```

```python
# Swap variables easily
a, b = 5, 10
a, b = b, a
print(f"a={a}, b={b}")
```

## type() and id()

- `type()` returns the type of an object
- `id()` returns the unique memory address (identity) of an object

```python
value = 42
print(type(value))
print(id(value))
```

```python
# Two variables pointing to same object (small integers are cached)
a = 100
b = 100
print(id(a), id(b))  # May be same for small integers

# Different objects
c = [1, 2, 3]
d = [1, 2, 3]
print(id(c), id(d))  # Different - two separate lists
```

## Common Mistakes

- **Using `=` for comparison**: Use `==` for equality, `=` for assignment
- **Reserved words**: Don't use `class`, `def`, `if` as variable names
- **Starting with number**: `2fast` is invalid
- **Assuming type persists**: After `x = "hi"`, `x` is a string even if it was an int before
