## Defining Functions

Use `def` to define a function. Functions encapsulate reusable logic.

```python
def greet():
    print("Hello, World!")

greet()
```

## Parameters and Arguments

Parameters are the names in the definition; arguments are the values you pass.

```python
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
greet("Bob")
```

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

## Default Arguments

Parameters can have default values. Defaults must come after non-default parameters.

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alice")           # Hello, Alice!
greet("Bob", "Hi")       # Hi, Bob!
```

```python
# Keyword arguments - order doesn't matter
def describe_pet(animal, name):
    print(f"I have a {animal} named {name}.")

describe_pet(name="Fluffy", animal="cat")
```

## Docstrings

Docstrings document your function. Use triple quotes right after the def line.

```python
def calculate_area(length, width):
    """
    Calculate the area of a rectangle.

    Args:
        length: The length of the rectangle.
        width: The width of the rectangle.

    Returns:
        The area (length * width).
    """
    return length * width

print(calculate_area(5, 3))
print(calculate_area.__doc__)
```

## Scope Introduction

Variables inside a function are local. Variables outside are global (use with care).

```python
x = 10  # global

def show_scope():
    y = 20  # local
    print("Inside:", x, y)

show_scope()
print("Outside:", x)
# print(y)  # Error - y is not defined here
```

```python
# To modify global variable (avoid when possible)
count = 0

def increment():
    global count
    count += 1

increment()
print(count)
```

## Common Mistakes

- **Mutable default argument**: `def f(x=[])` — list is shared! Use `def f(x=None)` then `x = x or []`
- **Forgetting return**: Function returns `None` if no return statement
- **Parameter order**: Default params must come after required params
- **Modifying global without `global`**: Assignment creates local variable instead
