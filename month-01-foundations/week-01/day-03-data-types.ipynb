## Integers (int)

Whole numbers with no decimal point. Python handles arbitrarily large integers.

```python
age = 25
negative = -100
big_number = 10**100  # Python handles this!
print(age, negative, big_number)
```

## Floats (float)

Decimal numbers. Be aware of floating-point precision.

```python
price = 19.99
scientific = 1.5e6  # 1.5 * 10^6 = 1500000.0
print(price, scientific)
```

```python
# Floating-point quirk
print(0.1 + 0.2)  # Not exactly 0.3!
# Use round() for display: round(0.1 + 0.2, 1)
```

## Booleans (bool)

`True` and `False` (capitalized!). Used for logic and conditions.

```python
is_raining = True
has_umbrella = False
print(is_raining, has_umbrella)

# Booleans are actually integers: True=1, False=0
print(True + True)  # 2
```

## None

Represents the absence of a value. Only one `None` object exists.

```python
result = None
print(result)
print(type(result))

# Common use: default/placeholder value
def find_user(id):
    # Simulate not found
    return None
```

## Complex Numbers (complex)

Numbers with real and imaginary parts. Use `j` for the imaginary unit.

```python
z = 3 + 4j
print(z)
print(z.real, z.imag)
print(abs(z))  # Magnitude: sqrt(3^2 + 4^2) = 5
```

## Type Conversion

Convert between types using `int()`, `float()`, `str()`, `bool()`.

```python
s = "42"
n = int(s)
print(n + 8)  # 50
```

```python
x = 3.14
print(int(x))   # Truncates: 3
print(str(x))   # "3.14"
print(bool(0))  # False
print(bool(1))  # True
print(bool("")) # False (empty string)
print(bool("hi")) # True (non-empty)
```

## Type Checking

Use `type()` or `isinstance()` to check types.

```python
value = 42
print(type(value) == int)
print(isinstance(value, int))  # Preferred
```

```python
# isinstance handles inheritance (e.g., bool is subclass of int)
print(isinstance(True, bool))  # True
print(isinstance(True, int))   # Also True!
print(type(True) == int)       # False
```

## Common Mistakes

- **`true`/`false`**: Python uses `True`/`False` (capitalized)
- **`int("3.14")`**: Fails! Use `int(float("3.14"))`
- **Assuming exact floats**: `0.1 + 0.2 != 0.3` due to binary representation
- **`type(x) == str`**: Prefer `isinstance(x, str)` for flexibility
