## What is Python?

Python is a high-level, interpreted programming language known for its readability and versatility. It's used in web development, data science, automation, and more.

## Your First Python Program: print()

The `print()` function displays output to the console. It's the most common way to see your program's results.

```python
# Your very first Python program!
print("Hello, World!")
```

```python
# print() can output multiple items, separated by spaces
print("Python", "is", "awesome!")
```

```python
# Customize the separator
print("apple", "banana", "cherry", sep=" | ")
```

```python
# end= changes what comes after the output (default is newline)
print("Hello", end=" ")
print("World!")
```

## Comments

Comments help you and others understand your code. Python ignores them when running.

```python
# This is a single-line comment
print("Code runs, comment doesn't")  # Inline comment

# Multi-line comments use multiple # lines
# Or you can use triple-quoted strings (often for docstrings)
"""
This is a multi-line string.
Often used for documentation.
"""
```

## Basic I/O: Getting User Input

Use `input()` to read text from the user. It always returns a string.

```python
# Uncomment and run to try interactive input:
# name = input("What's your name? ")
# print("Nice to meet you,", name)

# For notebooks, we can simulate:
name = "Alex"  # Simulated input
print("Nice to meet you,", name)
```

## Introduction to f-strings

f-strings (formatted string literals) let you embed expressions inside strings using `{expression}`.

```python
name = "Python"
version = 3
print(f"Welcome to {name} {version}!")
```

```python
# You can put any expression inside {}
a, b = 5, 3
print(f"{a} + {b} = {a + b}")
```

## Escape Characters

Escape sequences start with `\` and represent special characters.

```python
print("Line 1\nLine 2")  # \n = newline
print("Tab\tseparated")  # \t = tab
print("She said \"Hello\"")  # \" = literal quote
print("Path: C\\Users\\Documents")  # \\ = literal backslash
```

## Common Mistakes

- **Forgetting quotes**: `print(Hello)` → Error! Use `print("Hello")`
- **Wrong quote types**: Mixing `"` and `'` in unmatched pairs
- **Typo in print**: `Print()` or `prnt()` → Python is case-sensitive
- **input() returns string**: `age = input("Age? ")` gives `"25"` not `25`
