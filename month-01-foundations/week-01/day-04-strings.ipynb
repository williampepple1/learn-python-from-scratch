## String Creation

Strings can use single, double, or triple quotes. They are immutable.

```python
s1 = 'single quotes'
s2 = "double quotes"
s3 = '''multi
line'''
print(s1, s2)
print(s3)
```

## String Methods: upper, lower, strip

Common methods for transforming and cleaning strings.

```python
text = "  Hello World  "
print(text.upper())   # HELLO WORLD
print(text.lower())   # hello world
print(text.strip())   # removes leading/trailing whitespace
print(text.lstrip())  # left strip
print(text.rstrip())  # right strip
```

```python
# strip() can remove specific characters
url = "https://example.com/"
print(url.strip("/"))  # https://example.com
```

## split() and join()

`split()` breaks a string into a list; `join()` combines a list into a string.

```python
sentence = "Python is awesome"
words = sentence.split()  # default: split on whitespace
print(words)
```

```python
csv = "apple,banana,cherry"
fruits = csv.split(",")
print(fruits)

# join: separator.join(iterable)
back = " | ".join(fruits)
print(back)
```

## f-strings (Formatted String Literals)

Embed expressions with `{expression}`. Add format specifiers with `:{spec}`.

```python
name = "Alice"
score = 95.5
print(f"{name} scored {score}%")
```

```python
# Format specifiers
pi = 3.14159
print(f"Pi: {pi:.2f}")      # 2 decimal places
print(f"Percent: {0.85:.1%}")  # 85.0%
print(f"Padding: {42:05d}")   # 00042
```

## Slicing

`string[start:stop:step]` — start inclusive, stop exclusive. Negative indices count from end.

```python
s = "Hello World"
print(s[0])      # H (first char)
print(s[-1])     # d (last char)
print(s[0:5])    # Hello
print(s[:5])     # Hello (same)
print(s[6:])     # World
print(s[::-1])   # dlroW olleH (reversed)
```

```python
# Step in slicing
s = "abcdefgh"
print(s[::2])   # aceg (every 2nd)
print(s[1::2])  # bdfh (every 2nd starting at 1)
```

## Other String Formatting

Legacy: `.format()` and `%` operator. f-strings are preferred in Python 3.6+.

```python
name, age = "Bob", 30
print("{} is {} years old".format(name, age))
print("%s is %d years old" % (name, age))
```

## Common Mistakes

- **Strings are immutable**: `s[0] = 'x'` fails; create new string instead
- **split() returns list**: `"a,b".split(",")` → `["a", "b"]`
- **join() syntax**: `" ".join(list)` not `list.join(" ")`
- **Slicing**: `s[5:10]` gives indices 5–9, not 5–10
