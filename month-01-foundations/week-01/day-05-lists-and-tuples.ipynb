## Lists: Creation and Access

Lists are ordered, mutable collections. Use `[]` or `list()`.

```python
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
print(fruits[0], fruits[-1])
```

## List Slicing

Same syntax as string slicing: `[start:stop:step]`.

```python
nums = [10, 20, 30, 40, 50]
print(nums[1:4])   # [20, 30, 40]
print(nums[:3])    # [10, 20, 30]
print(nums[::2])   # [10, 30, 50]
print(nums[::-1])  # reversed
```

## List CRUD: Create, Read, Update, Delete

Lists are mutable—you can change them in place.

```python
# Create
items = ["a", "b", "c"]

# Update
items[1] = "B"
print(items)

# Append (add to end)
items.append("d")
print(items)

# Insert at index
items.insert(0, "start")
print(items)
```

```python
# Delete
items = ["a", "b", "c", "b"]
items.remove("b")   # removes first "b"
print(items)

popped = items.pop()   # removes and returns last
print(popped, items)

del items[0]  # delete by index
print(items)
```

## More List Methods

`sort()`, `reverse()`, `extend()`, `count()`, `index()`.

```python
nums = [3, 1, 4, 1, 5]
nums.sort()
print(nums)

nums.reverse()
print(nums)

nums.extend([9, 2])
print(nums)

print(nums.count(1))
print(nums.index(4))
```

## Tuples: Immutable Sequences

Tuples use `()`. They cannot be modified after creation.

```python
point = (3, 4)
rgb = (255, 128, 0)
single = (1,)  # comma needed for single-element tuple!
print(point[0], point[1])
```

```python
# Tuples are immutable
# point[0] = 5  # Error!
```

## Tuple Packing and Unpacking

Packing: multiple values into one tuple. Unpacking: assign tuple elements to variables.

```python
# Packing
coords = 10, 20  # parentheses optional
print(coords)

# Unpacking
x, y = coords
print(x, y)
```

```python
# Swap with unpacking
a, b = 1, 2
a, b = b, a
print(a, b)

# Unpack with * for rest
first, *rest = [1, 2, 3, 4, 5]
print(first, rest)
```

## Common Mistakes

- **Single-element tuple**: `(1)` is int, `(1,)` is tuple
- **List vs tuple**: Use list when you need to modify; tuple for fixed data
- **sort() vs sorted()**: `list.sort()` modifies in place; `sorted(list)` returns new list
- **append vs extend**: `append([1,2])` adds one element; `extend([1,2])` adds two
