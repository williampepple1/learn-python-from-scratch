## CSV Module — Reading

Use `csv.reader()` for list of lists, or `csv.DictReader()` for list of dicts.

```python
import csv
from io import StringIO

# Sample CSV data
csv_data = """name,age,city
Alice,30,NYC
Bob,25,LA
Charlie,35,Chicago"""

reader = csv.reader(StringIO(csv_data))
for row in reader:
    print(row)
```

```python
# DictReader — access by column name
reader = csv.DictReader(StringIO(csv_data))
for row in reader:
    print(f"{row['name']} is {row['age']} in {row['city']}")
```

## CSV Module — Writing

Use `csv.writer()` or `csv.DictWriter()` with `writeheader()` for column names.

```python
import csv

with open("users.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["name", "email", "score"])
    writer.writerow(["Alice", "alice@example.com", 95])
    writer.writerow(["Bob", "bob@example.com", 87])

print("Written users.csv")
```

```python
# DictWriter
with open("users_dict.csv", "w", newline="") as f:
    fieldnames = ["name", "email", "score"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"name": "Alice", "email": "alice@example.com", "score": 95})
    writer.writerow({"name": "Bob", "email": "bob@example.com", "score": 87})

print("Written users_dict.csv")
```

## JSON Module — Basics

`json.loads()` parses a string; `json.load()` reads from a file. `json.dumps()` and `json.dump()` for serialization.

```python
import json

# Parse JSON string
data = json.loads('{"name": "Alice", "age": 30, "active": true}')
print(data)
print(data["name"])

# Serialize to string
s = json.dumps(data, indent=2)
print(s)
```

## JSON — Reading and Writing Files

Use `json.load()` and `json.dump()` with file objects.

```python
config = {
    "app_name": "MyApp",
    "version": "1.0",
    "settings": {"debug": True, "port": 8080}
}

with open("config.json", "w") as f:
    json.dump(config, f, indent=2)

with open("config.json") as f:
    loaded = json.load(f)
    print(loaded)
```

## JSON — Custom Types

Use `default` and `object_hook` for non-standard types, or `default=str` for datetime etc.

```python
from datetime import datetime

def json_serializer(obj):
    if isinstance(obj, datetime):
        return obj.isoformat()
    raise TypeError(f"Type {type(obj)} not serializable")

data = {"created": datetime.now()}
s = json.dumps(data, default=json_serializer)
print(s)
```

## Real-World: CSV to JSON

Convert CSV data to a list of dicts and save as JSON.

```python
import csv
import json

with open("users.csv") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

with open("users.json", "w") as f:
    json.dump(rows, f, indent=2)

print("Converted users.csv to users.json")
```

## Common Mistakes

1. **CSV newline**: Use `newline=''` when opening CSV files to avoid extra blank lines on Windows
2. **JSON keys**: Must be strings; Python dict keys can be other types
3. **NaN/Infinity**: JSON doesn't support them; use custom serializers
4. **Encoding**: Open files with `encoding='utf-8'` for non-ASCII data
