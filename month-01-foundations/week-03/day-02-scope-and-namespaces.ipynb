## The LEGB Rule

Python looks up names in this order: **L**ocal → **E**nclosing → **G**lobal → **B**uilt-in.

```python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(x)  # local
    inner()
    print(x)  # enclosing

outer()
print(x)  # global
```

## Global vs Local Scope

Variables assigned inside a function are local by default. Use `global` to modify a global variable.

```python
count = 0

def increment():
    global count  # Without this, we'd get UnboundLocalError
    count += 1
    return count

increment()
increment()
print(count)  # 2
```

```python
# Reading global without declaring is fine
PI = 3.14159

def circle_area(radius):
    return PI * radius ** 2  # PI is read from global

print(circle_area(5))
```

## The `nonlocal` Keyword

Use `nonlocal` to modify a variable from an enclosing (non-global) scope.

```python
def counter():
    count = 0
    def increment():
        nonlocal count  # Refers to outer count, not global
        count += 1
        return count
    return increment

c = counter()
print(c())  # 1
print(c())  # 2
print(c())  # 3
```

## globals() and locals()

These built-in functions return dictionaries of the current global and local namespaces.

```python
x, y = 10, 20

def show_scopes():
    a, b = 1, 2
    print("Local:", list(locals().keys()))
    print("Global (sample):", [k for k in globals().keys() if not k.startswith('_')][:10])

show_scopes()
```

```python
# Dynamic variable creation (use sparingly!)
def set_global(name, value):
    globals()[name] = value

set_global("dynamic_var", 42)
print(dynamic_var)  # 42
```

## Shadowing Built-ins

Be careful not to shadow built-in names like `list`, `dict`, `str`, `sum`.

```python
# BAD: shadows built-in list
# list = [1, 2, 3]  # Don't do this!

# GOOD: use different names
my_list = [1, 2, 3]
result = sum(my_list)  # sum works because we didn't shadow it
print(result)
```

## Common Mistakes

1. **Forgetting `global`**: Assigning to a global without declaring it creates a local variable
2. **Using `nonlocal` for globals**: `nonlocal` only works for enclosing scope, not global
3. **UnboundLocalError**: Referencing a variable before assignment in a function makes it local
4. **Overusing globals**: Prefer passing values as arguments and returning results
