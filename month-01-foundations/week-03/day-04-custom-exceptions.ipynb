## Creating Custom Exception Classes

Define exceptions by inheriting from `Exception` (or a more specific base). Keep them simple.

```python
class InsufficientFundsError(Exception):
    """Raised when account balance is too low."""
    pass

class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(f"Balance {self.balance} < {amount}")
        self.balance -= amount
        return self.balance

acc = BankAccount(100)
try:
    acc.withdraw(150)
except InsufficientFundsError as e:
    print(f"Bank error: {e}")
```

## Exceptions with Attributes

Add attributes to store context (e.g., the invalid value, expected range).

```python
class ValidationError(Exception):
    def __init__(self, message, field=None, value=None):
        super().__init__(message)
        self.field = field
        self.value = value

def validate_email(email):
    if "@" not in email:
        raise ValidationError("Invalid email format", field="email", value=email)
    return email

try:
    validate_email("notanemail")
except ValidationError as e:
    print(f"{e} | field={e.field}, value={e.value}")
```

## Exception Chaining

Use `raise NewError(...) from original_error` to chain exceptions and preserve the cause.

```python
class ConfigError(Exception):
    pass

def load_config(path):
    try:
        with open(path) as f:
            return f.read()
    except FileNotFoundError as e:
        raise ConfigError(f"Config file not found: {path}") from e

try:
    load_config("missing.json")
except ConfigError as e:
    print(f"Config error: {e}")
    print(f"Caused by: {e.__cause__}")
```

## Exception Hierarchy for Your Domain

Create a base exception for your module, then subclass for specific cases.

```python
class DatabaseError(Exception):
    """Base for all database-related errors."""
    pass

class ConnectionError(DatabaseError):
    pass

class QueryError(DatabaseError):
    pass

# Callers can catch DatabaseError to get all DB errors
try:
    raise QueryError("Invalid SQL")
except DatabaseError as e:
    print(f"Database problem: {e}")
```

## Best Practices

1. **Name with `Error` suffix**: `InsufficientFundsError`, not `InsufficientFunds`
2. **Inherit from `Exception`**: Not from `BaseException`
3. **Keep them simple**: Often `pass` is enough; add attributes only when useful
4. **Document**: Use docstrings to explain when the exception is raised

```python
class RateLimitError(Exception):
    """
    Raised when an API rate limit is exceeded.
    Attributes:
        retry_after: seconds until limit resets
    """
    def __init__(self, message, retry_after=60):
        super().__init__(message)
        self.retry_after = retry_after

raise RateLimitError("Too many requests", retry_after=30)
```

## Common Mistakes

1. **Overusing custom exceptions**: Use built-ins (`ValueError`, `TypeError`) when they fit
2. **Forgetting to call `super().__init__()`**: Your exception may not display correctly
3. **Catching your base too early**: Put specific handlers before general ones
4. **Losing the chain**: Use `from e` when wrapping exceptions
