## open() and File Modes

| Mode | Meaning |
|------|--------|
| `'r'` | Read (default). File must exist. |
| `'w'` | Write. Overwrites existing file. |
| `'a'` | Append. Adds to end. |
| `'x'` | Exclusive create. Fails if file exists. |
| `'b'` | Binary (e.g., `'rb'`, `'wb'`) |
| `'+'` | Read and write (e.g., `'r+'`) |

```python
# Write a file
with open("sample.txt", "w") as f:
    f.write("Hello, World!\n")
    f.write("Line 2\n")

# Read it back
with open("sample.txt", "r") as f:
    content = f.read()
    print(content)
```

## Reading Methods

- `read()` — entire file as string
- `readline()` — one line
- `readlines()` — list of lines
- Iterating — memory-efficient for large files

```python
with open("sample.txt", "r") as f:
    # Read line by line (memory efficient)
    for line in f:
        print(line.strip())  # strip removes newline

with open("sample.txt", "r") as f:
    lines = f.readlines()  # List of lines
    print(lines)
```

## The with Statement

`with` ensures the file is closed even if an exception occurs. Always use it for files!

```python
# BAD: might leak file handle if error occurs
# f = open("file.txt")
# data = f.read()
# f.close()

# GOOD: automatic cleanup
with open("sample.txt") as f:
    data = f.read()
# File is closed here, even if read() raised an error
```

## pathlib — Modern Path Handling

`pathlib.Path` provides an object-oriented, cross-platform way to work with paths.

```python
from pathlib import Path

p = Path("month-01-foundations/week-03/sample.txt")
print(f"Exists: {p.exists()}")
print(f"Parent: {p.parent}")
print(f"Name: {p.name}")
print(f"Stem: {p.stem}")
print(f"Suffix: {p.suffix}")

# Reading with pathlib
content = Path("sample.txt").read_text(encoding="utf-8")
print(content)

# Writing
Path("output.txt").write_text("Written via pathlib!\n", encoding="utf-8")
```

## Binary Files

Use `'rb'` or `'wb'` for images, executables, etc. Read/write bytes, not strings.

```python
# Copy a file in binary mode
with open("sample.txt", "rb") as src:
    data = src.read()

with open("sample_copy.txt", "wb") as dst:
    dst.write(data)

print("File copied successfully")

# pathlib for binary
Path("sample_copy.txt").write_bytes(Path("sample.txt").read_bytes())
```

## Encoding

Specify `encoding='utf-8'` for text files to avoid platform-dependent defaults.

```python
with open("utf8_file.txt", "w", encoding="utf-8") as f:
    f.write("Café, naïve, 日本語")

with open("utf8_file.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

## Common Mistakes

1. **Forgetting to close**: Use `with` always
2. **Wrong mode**: `'w'` overwrites; use `'a'` to append
3. **Path separators**: Use `pathlib` or `os.path.join` for cross-platform paths
4. **Encoding**: Specify `encoding='utf-8'` for non-ASCII text
