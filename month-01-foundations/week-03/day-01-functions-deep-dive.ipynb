## *args: Variable Positional Arguments

Use `*args` when you want to accept any number of positional arguments. The `*` unpacks arguments into a tuple.

```python
def sum_all(*args):
    """Sum any number of arguments."""
    print(f"Received args: {args}, type: {type(args)}")
    return sum(args)

print(sum_all(1, 2, 3))           # 6
print(sum_all(10, 20, 30, 40))    # 100
print(sum_all())                   # 0
```

## **kwargs: Variable Keyword Arguments

Use `**kwargs` to accept any number of keyword arguments. They're collected into a dictionary.

```python
def build_profile(**kwargs):
    """Build a user profile from keyword arguments."""
    profile = {}
    for key, value in kwargs.items():
        profile[key] = value
    return profile

user = build_profile(name="Alice", age=30, city="NYC", occupation="Engineer")
print(user)
```

## Combining *args and **kwargs

You can use both together. Order matters: regular args, *args, **kwargs.

```python
def log_message(prefix, *args, **kwargs):
    """Log with prefix, optional args, and keyword metadata."""
    message = " ".join(str(a) for a in args)
    meta = ", ".join(f"{k}={v}" for k, v in kwargs.items())
    return f"[{prefix}] {message}" + (f" ({meta})" if meta else "")

print(log_message("INFO", "User logged in", user_id=42, timestamp="2025-03-10"))
```

## Lambda Functions

Lambda creates small anonymous functions in a single expression. Syntax: `lambda args: expression`

```python
# Traditional function vs lambda
def double(x):
    return x * 2

double_lambda = lambda x: x * 2

print(double(5))           # 10
print(double_lambda(5))   # 10

# Lambda with multiple args
add = lambda a, b: a + b
print(add(3, 7))  # 10

# Common use: sorting
students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]
sorted_by_grade = sorted(students, key=lambda s: s[1], reverse=True)
print(sorted_by_grade)
```

## Closures

A closure is a nested function that "remembers" variables from its enclosing scope, even after the outer function has finished.

```python
def make_multiplier(factor):
    """Returns a function that multiplies by factor."""
    def multiplier(x):
        return x * factor  # factor is "closed over"
    return multiplier

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))   # 10
print(triple(5))   # 15
```

## Higher-Order Functions

Higher-order functions take functions as arguments or return functions. Examples: `map`, `filter`, custom wrappers.

```python
# map: apply function to each item
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # [1, 4, 9, 16, 25]

# filter: keep items where function returns True
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # [2, 4]

# Custom higher-order: retry decorator
def retry(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for i in range(times):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    if i == times - 1:
                        raise
            return None
        return wrapper
    return decorator

@retry(3)
def flaky_operation():
    return "Success!"

print(flaky_operation())
```

## Common Mistakes

1. **Mutable default arguments**: `def f(items=[])` — the list is shared! Use `def f(items=None)` and `items = items or []`
2. **Lambda overuse**: Don't write complex lambdas; use a named function for readability
3. **Wrong order**: `def f(a, *args, **kwargs)` — *args must come before **kwargs
4. **Closure gotcha**: Loop variables in closures capture by reference, not value — use default args to fix
