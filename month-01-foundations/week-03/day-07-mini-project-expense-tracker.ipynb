## Step 1: Define the Data Structure and File Path

We'll store expenses in a CSV with columns: date, description, amount, category.

```python
import csv
from pathlib import Path
from datetime import datetime

EXPENSES_FILE = Path("expenses.csv")
FIELDNAMES = ["date", "description", "amount", "category"]
```

## Step 2: Initialize the CSV File

Create the file with headers if it doesn't exist.

```python
def init_expenses_file():
    """Create expenses.csv with headers if it doesn't exist."""
    if not EXPENSES_FILE.exists():
        with open(EXPENSES_FILE, "w", newline="") as f:
            writer = csv.DictWriter(f, fieldnames=FIELDNAMES)
            writer.writeheader()
        print("Created expenses.csv")

init_expenses_file()
```

## Step 3: Add an Expense

Append a new expense row to the CSV.

```python
def add_expense(description, amount, category="Other"):
    """Add an expense. Amount can be positive (income) or negative (expense)."""
    try:
        amount = float(amount)
    except ValueError:
        raise ValueError("Amount must be a number")
    
    date = datetime.now().strftime("%Y-%m-%d")
    row = {"date": date, "description": description, "amount": amount, "category": category}
    
    with open(EXPENSES_FILE, "a", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=FIELDNAMES)
        writer.writerow(row)
    
    print(f"Added: {description} — {amount} ({category})")

add_expense("Coffee", -4.50, "Food")
add_expense("Salary", 3000, "Income")
```

## Step 4: List All Expenses

Read and display all expenses from the CSV.

```python
def list_expenses():
    """Print all expenses."""
    with open(EXPENSES_FILE) as f:
        reader = csv.DictReader(f)
        rows = list(reader)
    
    if not rows:
        print("No expenses yet.")
        return
    
    print(f"{"Date":<12} {"Description":<20} {"Amount":>10} {"Category":<10}")
    print("-" * 55)
    for row in rows:
        print(f"{row['date']:<12} {row['description']:<20} {float(row['amount']):>10.2f} {row['category']:<10}")

list_expenses()
```

## Step 5: Get Summary (Total, by Category)

Calculate total balance and optionally break down by category.

```python
def get_summary():
    """Return total and per-category totals."""
    with open(EXPENSES_FILE) as f:
        reader = csv.DictReader(f)
        rows = list(reader)
    
    total = sum(float(r["amount"]) for r in rows)
    by_category = {}
    for row in rows:
        cat = row["category"]
        by_category[cat] = by_category.get(cat, 0) + float(row["amount"])
    
    return total, by_category

def show_summary():
    total, by_category = get_summary()
    print(f"\nTotal balance: {total:.2f}")
    print("By category:")
    for cat, amt in sorted(by_category.items()):
        print(f"  {cat}: {amt:.2f}")

show_summary()
```

## Step 6: Main Menu and Run Loop

Interactive menu to add, list, summarize, or quit.

```python
def main():
    init_expenses_file()
    while True:
        print("\n--- Expense Tracker ---")
        print("1. Add expense")
        print("2. List expenses")
        print("3. Summary")
        print("4. Quit")
        choice = input("Choice (1-4): ").strip()
        
        if choice == "1":
            desc = input("Description: ").strip()
            amt = input("Amount (negative for expense): ").strip()
            cat = input("Category [Other]: ").strip() or "Other"
            try:
                add_expense(desc, amt, cat)
            except ValueError as e:
                print(f"Error: {e}")
        elif choice == "2":
            list_expenses()
        elif choice == "3":
            show_summary()
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice.")

# Uncomment to run interactively:
# main()
```

## Step 7: Complete Standalone Script

Here's the full program in one cell for easy copy-paste or running as a script.

```python
import csv
from pathlib import Path
from datetime import datetime

EXPENSES_FILE = Path("expenses.csv")
FIELDNAMES = ["date", "description", "amount", "category"]

def init_expenses_file():
    if not EXPENSES_FILE.exists():
        with open(EXPENSES_FILE, "w", newline="") as f:
            csv.DictWriter(f, fieldnames=FIELDNAMES).writeheader()

def add_expense(description, amount, category="Other"):
    amount = float(amount)
    row = {"date": datetime.now().strftime("%Y-%m-%d"), "description": description, "amount": amount, "category": category}
    with open(EXPENSES_FILE, "a", newline="") as f:
        csv.DictWriter(f, fieldnames=FIELDNAMES).writerow(row)

def list_expenses():
    with open(EXPENSES_FILE) as f:
        rows = list(csv.DictReader(f))
    for r in rows:
        print(f"{r['date']} | {r['description']} | {r['amount']} | {r['category']}")

def show_summary():
    with open(EXPENSES_FILE) as f:
        rows = list(csv.DictReader(f))
    total = sum(float(r["amount"]) for r in rows)
    print(f"Total: {total:.2f}")

def main():
    init_expenses_file()
    while True:
        print("\n1.Add 2.List 3.Summary 4.Quit")
        c = input("Choice: ").strip()
        if c == "1":
            add_expense(input("Desc: "), input("Amount: "), input("Category: ") or "Other")
        elif c == "2": list_expenses()
        elif c == "3": show_summary()
        elif c == "4": break

if __name__ == "__main__":
    main()
```

## Common Mistakes

1. **Forgetting newline=''** in CSV — causes extra blank lines on Windows
2. **Not converting amount to float** — CSV stores strings
3. **File not found** — call `init_expenses_file()` before first use
4. **No error handling** — wrap user input in try/except for invalid amounts
