## Basic try/except

Catch and handle errors to prevent program crashes.

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
    result = None

print(f"Result: {result}")
```

## try/except/else/finally

- **else**: runs only if no exception occurred
- **finally**: always runs (cleanup, close files, etc.)

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Cannot divide by zero!")
        return None
    else:
        print("Division succeeded!")
        return result
    finally:
        print("Cleanup: always runs")

print(safe_divide(10, 2))
print("---")
print(safe_divide(10, 0))
```

## Common Exceptions

| Exception | When it occurs |
|-----------|----------------|
| `ValueError` | Invalid value (e.g., `int('abc')`) |
| `TypeError` | Wrong type (e.g., `'a' + 1`) |
| `KeyError` | Missing dict key |
| `IndexError` | List index out of range |
| `FileNotFoundError` | File doesn't exist |
| `ZeroDivisionError` | Division by zero |

```python
# Catching multiple exceptions
def parse_number(s):
    try:
        return int(s)
    except (ValueError, TypeError) as e:
        print(f"Parse error: {e}")
        return None

print(parse_number("42"))   # 42
print(parse_number("abc"))  # None
print(parse_number(None))   # None
```

## Raising Exceptions

Use `raise` to signal that something went wrong. You can raise with or without an exception instance.

```python
def set_age(age):
    if not 0 <= age <= 150:
        raise ValueError(f"Age {age} must be between 0 and 150")
    return age

try:
    set_age(25)   # OK
    set_age(200)  # Raises ValueError
except ValueError as e:
    print(f"Caught: {e}")
```

## Exception Hierarchy

All exceptions inherit from `BaseException`. Catching a parent catches all children. Be specific!

```python
# Exception hierarchy (simplified)
# BaseException
#   ├── SystemExit, KeyboardInterrupt
#   └── Exception
#         ├── ValueError, TypeError, KeyError, ...
#         └── (your custom exceptions)

try:
    x = int("not a number")
except Exception as e:  # Catches almost everything (broad)
    print(f"Something went wrong: {e}")

# Prefer specific:
try:
    x = int("not a number")
except ValueError as e:
    print(f"Invalid number: {e}")
```

## Re-raising Exceptions

Use `raise` without arguments to re-raise the current exception (e.g., after logging).

```python
def process(data):
    try:
        return data["key"]  # May raise KeyError
    except KeyError as e:
        print(f"Log: Missing key - {e}")
        raise  # Re-raise for caller to handle

try:
    process({})
except KeyError:
    print("Caller caught KeyError")
```

## Common Mistakes

1. **Bare `except:`** — catches everything including `KeyboardInterrupt`; use `except Exception:`
2. **Catching too broadly** — prefer specific exceptions over `Exception`
3. **Swallowing exceptions** — don't use empty `except: pass` without logging
4. **Wrong order** — put specific exceptions before general ones
