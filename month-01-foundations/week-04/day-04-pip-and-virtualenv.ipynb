## pip — Package Installer

pip installs packages from PyPI. Use it from the command line.

```python
# Command-line examples (run in terminal, not notebook):
# pip install requests
# pip install requests==2.28.0
# pip install 'requests>=2.0,<3.0'
# pip uninstall requests
# pip list
# pip show requests

# From Python/notebook, you can run:
import subprocess
from pathlib import Path
result = subprocess.run(["pip", "list"], capture_output=True, text=True)
print(result.stdout[:500])
```

## requirements.txt

List dependencies for reproducible installs. Create with `pip freeze` or write manually.

```python
requirements_content = """# Project dependencies
requests>=2.28.0
python-dotenv>=1.0.0
"""

from pathlib import Path
Path("requirements.txt").write_text(requirements_content)
print("Created requirements.txt")
print(requirements_content)

# Install with: pip install -r requirements.txt
```

## Creating a Virtual Environment (venv)

venv creates an isolated Python environment. Each project gets its own packages.

```python
import venv
from pathlib import Path

venv_dir = Path(".venv_demo")
if not venv_dir.exists():
    venv.create(venv_dir)
    print("Created .venv_demo")
else:
    print(".venv_demo already exists")

# Activate (run in terminal):
# Windows: .venv_demo\Scripts\activate
# Linux/Mac: source .venv_demo/bin/activate
```

## Activating venv

| OS | Command |
|----|--------|
| Windows (cmd) | `.venv\Scripts\activate.bat` |
| Windows (PowerShell) | `.venv\Scripts\Activate.ps1` |
| Linux/Mac | `source .venv/bin/activate` |

```python
# After activation, 'which python' (or 'where python' on Windows)
# points to the venv's Python. pip install goes to venv only.

import sys
print("Current Python:", sys.executable)
```

## pip freeze for requirements.txt

Capture exact versions of installed packages for reproducibility.

```python
result = subprocess.run(["pip", "freeze"], capture_output=True, text=True)
lines = [l for l in result.stdout.strip().split("\n") if l and not l.startswith("#")][:5]
print("Sample pip freeze output:")
for line in lines:
    print(line)
```

## Best Practices

1. **Always use venv** for projects — avoids polluting system Python
2. **Pin versions** in requirements.txt for production (`==` or `>=` with upper bound)
3. **Separate dev deps** — use `requirements-dev.txt` or `pip install -e .[dev]`
4. **.gitignore** — add `venv/`, `.venv/` to avoid committing the env

## Common Mistakes

1. **Installing globally**: Always activate venv first
2. **Forgetting to commit requirements.txt**: Others can't reproduce your env
3. **Pip vs pip3**: Use `python -m pip` to ensure you're using the right Python
4. **Confusing venv with conda**: venv is stdlib; conda is a separate tool
