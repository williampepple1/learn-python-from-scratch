## Basic Logging

Use `logging` instead of `print()` for production-ready output. Default level is WARNING.

```python
import logging

logging.debug("Debug message")
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
logging.critical("Critical message")
```

## Setting Log Level

Configure the root logger to see all messages at or above the specified level.

```python
logging.basicConfig(level=logging.DEBUG)

logging.debug("Now visible")
logging.info("Info visible")
logging.warning("Warning visible")
```

## Formatters

Customize the format of log messages: timestamp, level, logger name, message.

```python
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("Formatted message")
logging.warning("Another formatted message")
```

## Logger Objects

Use named loggers for different modules. Use `getLogger(__name__)` for module-specific names.

```python
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)

handler = logging.StreamHandler()
handler.setLevel(logging.DEBUG)
formatter = logging.Formatter("%(name)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)
logger.addHandler(handler)

logger.debug("Debug from logger")
logger.info("Info from logger")
```

## File Handler

Write logs to a file instead of (or in addition to) stderr.

```python
file_logger = logging.getLogger("file_logger")
file_logger.setLevel(logging.INFO)

fh = logging.FileHandler("app.log", encoding="utf-8")
fh.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
file_logger.addHandler(fh)

file_logger.info("Written to app.log")
print("Check app.log for the message")
```

## Logging with Variables

Use `%s` placeholders or f-strings. Prefer `%` style for lazy evaluation (avoids formatting if level is filtered).

```python
user = "Alice"
count = 42
logging.warning("User %s failed login %d times", user, count)

# Also valid: logging.warning(f"User {user} failed {count} times")
# But % style is slightly more efficient when level is filtered
```

## Best Practices

1. **Use `getLogger(__name__)`** in modules
2. **Configure once** at application entry (e.g., main.py)
3. **DEBUG for dev, INFO for prod** — adjust level per environment
4. **Don't log sensitive data** — passwords, tokens, PII

## Common Mistakes

1. **Using print() everywhere**: Use logging for production code
2. **Logging too much**: DEBUG in production can be noisy and slow
3. **Wrong level**: Use ERROR for exceptions, INFO for normal flow
4. **Multiple basicConfig**: Only first call has effect; configure explicitly
