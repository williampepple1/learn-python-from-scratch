## sys.argv

`sys.argv` is a list of strings: script name is argv[0], rest are arguments.

```python
import sys

print("argv:", sys.argv)
print("Script:", sys.argv[0] if sys.argv else "N/A")
print("Args:", sys.argv[1:] if len(sys.argv) > 1 else "[]")

# Example: python script.py foo bar
# argv = ['script.py', 'foo', 'bar']
```

## argparse — Basic Usage

`argparse` provides automatic help, type conversion, and validation.

```python
import argparse

parser = argparse.ArgumentParser(description="A simple CLI tool")
parser.add_argument("name", help="Your name")
parser.add_argument("-v", "--verbose", action="store_true", help="Verbose output")
parser.add_argument("-n", "--count", type=int, default=1, help="Repeat count")

# Simulate: python script.py Alice -v -n 3
args = parser.parse_args(["Alice", "-v", "-n", "3"])
print(f"Name: {args.name}")
print(f"Verbose: {args.verbose}")
print(f"Count: {args.count}")
```

## Positional vs Optional

Positional args: required, no dash. Optional: prefixed with `-` or `--`.

```python
parser = argparse.ArgumentParser()
parser.add_argument("input_file", help="Input file path")
parser.add_argument("output_file", nargs="?", default="out.txt", help="Output file (optional)")
parser.add_argument("--upper", action="store_true", help="Convert to uppercase")

args = parser.parse_args(["input.txt"])
print(f"Input: {args.input_file}")
print(f"Output: {args.output_file}")
print(f"Upper: {args.upper}")
```

## Choices and nargs

Restrict values with `choices`; collect multiple with `nargs`.

```python
parser = argparse.ArgumentParser()
parser.add_argument("--mode", choices=["read", "write", "append"], default="read")
parser.add_argument("files", nargs="+", help="One or more files")

args = parser.parse_args(["--mode", "write", "a.txt", "b.txt"])
print(f"Mode: {args.mode}")
print(f"Files: {args.files}")
```

## Subcommands

Like `git commit`, `git push` — use `add_subparsers()`.

```python
parser = argparse.ArgumentParser()
subparsers = parser.add_subparsers(dest="command", help="Available commands")

add_parser = subparsers.add_parser("add", help="Add numbers")
add_parser.add_argument("numbers", nargs="+", type=int)

mul_parser = subparsers.add_parser("mul", help="Multiply numbers")
mul_parser.add_argument("numbers", nargs="+", type=int)

args = parser.parse_args(["add", "1", "2", "3"])
if args.command == "add":
    print("Sum:", sum(args.numbers))
elif args.command == "mul":
    from functools import reduce
    import operator
    print("Product:", reduce(operator.mul, args.numbers))

args2 = parser.parse_args(["mul", "2", "3", "4"])
from functools import reduce
import operator
print("Product:", reduce(operator.mul, args2.numbers))
```

## Complete CLI Example

A minimal script that greets with optional verbosity.

```python
def main():
    parser = argparse.ArgumentParser(description="Greeter")
    parser.add_argument("name", help="Name to greet")
    parser.add_argument("-v", "--verbose", action="store_true")
    args = parser.parse_args()
    msg = f"Hello, {args.name}!"
    if args.verbose:
        msg = f"[VERBOSE] {msg}"
    print(msg)

if __name__ == "__main__":
    main()

# Run with: python script.py World -v
```

## Common Mistakes

1. **parse_args() in notebook**: Pass a list to avoid using sys.argv
2. **Required optional args**: Use `required=True` for --flags that must be set
3. **Conflicting options**: Use `add_mutually_exclusive_group()`
4. **Forgetting dest**: Use `dest="my_var"` for custom attribute names
