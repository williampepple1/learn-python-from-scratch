## Package Structure

A package is a directory containing `__init__.py` (and optionally other modules). It allows hierarchical organization.

```python
# Example structure:
# myproject/
#   __init__.py
#   utils.py
#   models.py
#   subpackage/
#     __init__.py
#     helper.py
#
# Usage: from myproject import utils
#        from myproject.subpackage import helper
```

## __init__.py

Marks a directory as a package. Can be empty or used to expose a public API.

```python
from pathlib import Path

# Create a minimal package
pkg_dir = Path("mypkg")
pkg_dir.mkdir(exist_ok=True)
(pkg_dir / "__init__.py").write_text("# mypkg package\n")
(pkg_dir / "greet.py").write_text('''
def say_hi():
    return "Hi from mypkg!"
''')

import mypkg.greet
print(mypkg.greet.say_hi())
```

## Exposing API via __init__.py

Re-export from `__init__.py` for cleaner imports: `from mypkg import say_hi`

```python
# In mypkg/__init__.py:
# from .greet import say_hi
# __all__ = ['say_hi']

(pkg_dir / "__init__.py").write_text("from .greet import say_hi\n__all__ = ['say_hi']\n")

import importlib
import mypkg
importlib.reload(mypkg)
print(mypkg.say_hi())
```

## Relative Imports

Use `.` for same package, `..` for parent. Only work inside packages.

```python
# Inside mypkg/submod.py:
# from . import greet      # import greet from same package
# from .greet import say_hi
# from ..otherpkg import x  # from sibling package

sub_dir = pkg_dir / "sub"
sub_dir.mkdir(exist_ok=True)
(sub_dir / "__init__.py").write_text("")
(sub_dir / "helper.py").write_text('''
from ..greet import say_hi
def greet_twice():
    return say_hi() + " " + say_hi()
''')

from mypkg.sub.helper import greet_twice
print(greet_twice())
```

## Namespace Packages (PEP 420)

Python 3.3+ allows packages without `__init__.py` — "namespace packages." Multiple directories can contribute to the same package name.

```python
# Namespace package: dirs without __init__.py
# If both /path/a/ns and /path/b/ns exist (no __init__.py),
# 'import ns' can load from both.
# Used for plugin systems and split distributions.

print("Namespace packages allow split package layout across dirs")
```

## Common Mistakes

1. **Relative import at top level**: Relative imports only work inside a package
2. **Missing __init__.py**: Traditional packages need it; namespace packages don't
3. **Circular package imports**: Subpackage imports parent — restructure
4. **Wrong PYTHONPATH**: Package dir must be on sys.path (e.g., project root)
