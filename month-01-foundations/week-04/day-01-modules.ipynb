## import Statement

Import a module to use its attributes. Access with `module.attribute`.

```python
import math

print(math.pi)
print(math.sqrt(16))
```

## from ... import

Import specific names. Avoid `from module import *` — it pollutes the namespace.

```python
from math import sqrt, pi

print(sqrt(9))
print(pi)

# Alias to avoid name clashes
from datetime import datetime as dt
print(dt.now())
```

## __name__ == '__main__'

Code under `if __name__ == '__main__':` runs only when the file is executed directly, not when imported.

```python
# Create a file my_module.py with:
# def greet():
#     return "Hello"
# if __name__ == '__main__':
#     print(greet())
#
# Running: python my_module.py  -> prints Hello
# Importing: import my_module   -> does NOT print

# Simulate in notebook:
def greet():
    return "Hello"

if __name__ == "__main__":
    print("Running as script:", greet())
else:
    print("Imported — __name__ is:", __name__)
```

## Module Search Path (sys.path)

Python looks for modules in directories listed in `sys.path`. The current directory and site-packages are included.

```python
import sys

print("Module search path:")
for i, p in enumerate(sys.path[:5]):
    print(f"  {i}: {p}")

# Add a custom path (rarely needed)
# sys.path.insert(0, '/my/custom/module/dir')
```

## Creating Your Own Module

Any `.py` file is a module. Save as `utils.py` and import it.

```python
# Create utils.py in the same directory:
utils_content = '''
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

if __name__ == "__main__":
    print("Utils module run directly")
'''

from pathlib import Path
Path("utils.py").write_text(utils_content)

import utils
print(utils.add(2, 3))
print(utils.multiply(4, 5))
```

## dir() and __all__

`dir(module)` lists attributes. `__all__` defines what `from module import *` exports.

```python
import math
print([x for x in dir(math) if not x.startswith('_')][:10])

# __all__ in a module limits 'from module import *'
# e.g. in module: __all__ = ['add', 'multiply']
```

## Common Mistakes

1. **Circular imports**: A imports B, B imports A — restructure or use lazy imports
2. **from x import ***: Pollutes namespace; avoid in production
3. **Naming modules like stdlib**: Don't name your file `math.py` or `json.py`
4. **Forgetting __main__**: Put script code under `if __name__ == '__main__':` so it doesn't run on import
