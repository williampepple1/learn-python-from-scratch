## os Module

Operating system interface: paths, environment, directory operations.

```python
import os

print("Current dir:", os.getcwd())
print("Path exists:", os.path.exists("sample.txt"))
print("Path join:", os.path.join("dir", "sub", "file.txt"))
print("Env HOME:", os.environ.get("HOME", "N/A"))
print("List dir:", os.listdir(".")[:5])
```

## sys Module

System-specific parameters: argv, path, exit, stdin/stdout.

```python
import sys

print("Python version:", sys.version)
print("Platform:", sys.platform)
print("argv (script args):", sys.argv)
# sys.exit(0)  # Exit with code 0
# sys.exit(1)  # Exit with error
```

## datetime Module

`date`, `time`, `datetime`, `timedelta` for working with dates and times.

```python
from datetime import datetime, date, timedelta

now = datetime.now()
print("Now:", now)
print("Date:", now.date())
print("Time:", now.time())
print("Formatted:", now.strftime("%Y-%m-%d %H:%M"))

tomorrow = now + timedelta(days=1)
print("Tomorrow:", tomorrow.date())

parsed = datetime.strptime("2025-03-10", "%Y-%m-%d")
print("Parsed:", parsed)
```

## math Module

Mathematical functions: sqrt, ceil, floor, pow, pi, e, etc.

```python
import math

print("sqrt(16):", math.sqrt(16))
print("ceil(3.2):", math.ceil(3.2))
print("floor(3.8):", math.floor(3.8))
print("pi:", math.pi)
print("log(100, 10):", math.log(100, 10))
print("gcd(48, 18):", math.gcd(48, 18))
```

## random Module

Pseudorandom number generation: choice, randint, shuffle, random.

```python
import random

print("random():", random.random())  # 0.0 to 1.0
print("randint(1, 10):", random.randint(1, 10))
print("choice([a,b,c]):", random.choice(["a", "b", "c"]))

items = [1, 2, 3, 4, 5]
random.shuffle(items)
print("Shuffled:", items)

print("Sample 3:", random.sample(items, 3))
```

## collections — Counter

Count hashable objects. Perfect for word frequency, voting, etc.

```python
from collections import Counter

words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
counts = Counter(words)
print(counts)
print("Most common 2:", counts.most_common(2))

text = "hello world hello"
print(Counter(text.split()))
```

## collections — defaultdict

Dict with default factory for missing keys. No KeyError.

```python
from collections import defaultdict

dd = defaultdict(list)
dd["a"].append(1)
dd["a"].append(2)
dd["b"].append(3)
print(dict(dd))

dd_int = defaultdict(int)
for word in ["a", "b", "a", "a"]:
    dd_int[word] += 1
print(dict(dd_int))
```

## collections — deque

Double-ended queue. O(1) append/pop from both ends. Good for queues and stacks.

```python
from collections import deque

d = deque([1, 2, 3])
d.append(4)
d.appendleft(0)
print(d)
d.pop()
d.popleft()
print(d)
```

## Common Mistakes

1. **os.path vs pathlib**: Prefer pathlib for new code; os.path is older
2. **random for security**: Use `secrets` for passwords/tokens, not `random`
3. **datetime timezone**: `datetime.now()` is naive; use `datetime.now(timezone.utc)` for UTC
4. **Counter with non-hashable**: Counter requires hashable items
