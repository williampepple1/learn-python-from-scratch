## More String Methods

`replace()`, `startswith()`, `endswith()`, `find()`, `count()`, `isalpha()`, `isdigit()`

```python
s = "Hello World"
print(s.replace("World", "Python"))
print(s.startswith("Hello"))
print(s.endswith("World"))
print(s.find("o"))  # First occurrence, -1 if not found
print(s.count("l"))
```

```python
print("abc".isalpha())   # True
print("123".isdigit())  # True
print("a1b2".isalnum()) # True (letters or digits)
print("   ".isspace())  # True
```

## Regex Basics with `re`

Import `re` for pattern matching. Common: `re.search()`, `re.findall()`, `re.sub()`

```python
import re

text = "Contact: alice@email.com or bob@test.org"
match = re.search(r"\w+@\w+\.\w+", text)
if match:
    print("Found:", match.group())
```

```python
# Find all matches
emails = re.findall(r"\w+@\w+\.\w+", text)
print(emails)
```

```python
# Replace with regex
text = "Price: $99.99 and $12.50"
masked = re.sub(r"\d+\.\d+", "XX.XX", text)
print(masked)
```

```python
# Common patterns
# \d = digit, \w = word char, \s = whitespace
# + = one or more, * = zero or more
phone = "Call 555-123-4567"
nums = re.findall(r"\d{3}-\d{3}-\d{4}", phone)
print(nums)
```

## Advanced String Formatting

f-string format specifiers: alignment, width, precision.

```python
name = "Alice"
score = 95.5
print(f"{name:>10}")   # Right-align, width 10
print(f"{name:<10}")   # Left-align
print(f"{name:^10}")   # Center
print(f"{score:.1f}")  # 1 decimal
print(f"{42:05d}")     # Zero-padded, 5 chars
```

```python
# Format with thousands separator
n = 1_000_000
print(f"{n:,}")

# Percentage
p = 0.856
print(f"{p:.1%}")
```

## Common Mistakes

- **Raw string for regex**: Use `r"pattern"` to avoid escaping backslashes
- **re.search vs re.match**: search finds anywhere; match only at start
- **replace()**: `s.replace(a, b)` returns new string; doesn't modify in place
- **find() returns -1**: Check for -1 when using find()
