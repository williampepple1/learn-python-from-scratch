## for Loops

Iterate over a sequence (list, string, range, etc.).

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
for char in "Python":
    print(char, end=" ")
print()
```

## range()

`range(stop)` or `range(start, stop)` or `range(start, stop, step)`. Produces integers, stop is exclusive.

```python
print(list(range(5)))        # 0,1,2,3,4
print(list(range(2, 6)))    # 2,3,4,5
print(list(range(0, 10, 2))) # 0,2,4,6,8
```

```python
for i in range(5):
    print(i, end=" ")
print()

for i in range(1, 6):
    print(i**2, end=" ")
print()
```

```python
# enumerate: get index and value
for i, fruit in enumerate(["apple", "banana", "cherry"]):
    print(f"{i}: {fruit}")
```

## while Loops

Repeat while condition is True. Ensure condition eventually becomes False!

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Common pattern: user input until valid
n = 0
while n <= 0:
    n = 3  # Simulate: n = int(input("Enter positive number: "))
    if n <= 0:
        print("Invalid. Try again.")
print(f"You entered {n}")
```

## break and continue

- `break`: exit the loop immediately
- `continue`: skip to next iteration

```python
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")
print(" (stopped at 5)")
```

```python
for i in range(6):
    if i % 2 == 0:
        continue
    print(i, end=" ")  # Only odd: 1 3 5
print()
```

## else Clause on Loops

The `else` block runs if the loop completes without hitting `break`.

```python
for i in range(3):
    print(i)
else:
    print("Loop finished normally")

for i in range(5):
    if i == 2:
        break
    print(i)
else:
    print("This won't print")
```

## Nested Loops

Loop inside a loop. Useful for 2D structures.

```python
for i in range(3):
    for j in range(3):
        print(f"({i},{j})", end=" ")
    print()
```

```python
# Multiplication table
for row in range(1, 4):
    for col in range(1, 4):
        print(row * col, end="\t")
    print()
```

## Common Mistakes

- **Infinite while**: Forgetting to update the condition variable
- **range(5)**: 0..4, not 1..5; use range(1,6) for 1..5
- **Modifying list while iterating**: Can cause unexpected behavior—iterate over a copy
- **break/continue**: Only affect the innermost loop
