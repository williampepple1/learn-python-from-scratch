## Basic List Comprehensions

`[expression for item in iterable]` — build a list in one line.

```python
# Instead of:
squares = []
for x in range(5):
    squares.append(x ** 2)
print(squares)

# Use comprehension:
squares = [x ** 2 for x in range(5)]
print(squares)
```

```python
words = ["hello", "world", "python"]
upper_words = [w.upper() for w in words]
print(upper_words)
```

## Comprehensions with Conditions

`[expr for item in iterable if condition]` — filter while building.

```python
nums = [1, 2, 3, 4, 5, 6]
evens = [n for n in nums if n % 2 == 0]
print(evens)
```

```python
# if-else in expression (ternary)
nums = [1, 2, 3, 4, 5]
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print(labels)
```

## Nested Comprehensions

Comprehension inside comprehension for 2D structures.

```python
# 3x3 matrix
matrix = [[i * 3 + j for j in range(3)] for i in range(3)]
print(matrix)
```

```python
# Flatten a 2D list
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print(flat)
```

## Dict Comprehensions

`{key_expr: value_expr for item in iterable}`

```python
nums = [1, 2, 3, 4, 5]
squares_dict = {n: n**2 for n in nums}
print(squares_dict)
```

```python
words = ["apple", "banana", "cherry"]
lengths = {w: len(w) for w in words}
print(lengths)
```

## Set Comprehensions

`{expr for item in iterable}` — creates a set (no duplicates).

```python
nums = [1, 2, 2, 3, 3, 3, 4]
unique_squares = {n**2 for n in nums}
print(unique_squares)
```

```python
words = ["hello", "world", "hello"]
lengths = {len(w) for w in words}
print(lengths)  # {5} - both "hello" and "world" have unique lengths
```

## Generator Expressions (Bonus)

Use `()` instead of `[]` for memory efficiency—produces values one at a time.

```python
gen = (x**2 for x in range(5))
print(list(gen))  # Consume generator
```

## Common Mistakes

- **Order in nested**: `[x for row in matrix for x in row]` — for loops left to right
- **Dict vs set**: `{k: v}` is dict; `{x}` is set—syntax differs
- **Overly complex**: If comprehension is hard to read, use a regular loop
- **Variable scope**: Variable in comprehension can leak in Python 2, not in 3
