## if Statement

Execute code only when a condition is True. Use indentation (4 spaces) for the block.

```python
age = 18
if age >= 18:
    print("You can vote!")
```

## if / elif / else

Chain conditions. Only the first True block runs.

```python
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"Grade: {grade}")
```

## Comparison Operators

`==` (equal), `!=` (not equal), `<`, `>`, `<=`, `>=`

```python
a, b = 5, 10
print(a == b)   # False
print(a != b)   # True
print(a < b)    # True
print(a >= 5)   # True
```

## Logical Operators: and, or, not

Combine conditions.

```python
age = 25
has_ticket = True
if age >= 18 and has_ticket:
    print("Welcome!")

is_weekend = False
if is_weekend or age < 12:
    print("Discount applies")

if not has_ticket:
    print("Please buy a ticket")
```

## Ternary Operator

`value_if_true if condition else value_if_false`

```python
age = 20
status = "adult" if age >= 18 else "minor"
print(status)

# Equivalent to:
# if age >= 18:
#     status = "adult"
# else:
#     status = "minor"
```

## Nested Conditions

if inside if. Use sparingly—consider flattening with and/or.

```python
age = 25
is_member = True

if age >= 18:
    if is_member:
        print("Premium access")
    else:
        print("Standard access")
else:
    print("Access denied")
```

```python
# Flattened version
if age >= 18 and is_member:
    print("Premium access")
elif age >= 18:
    print("Standard access")
else:
    print("Access denied")
```

## Truthiness and Falsiness

In conditions, values are evaluated as True or False. Falsy: `False`, `0`, `0.0`, `""`, `[]`, `{}`, `None`.

```python
print(bool(0))      # False
print(bool(""))     # False
print(bool([]))    # False
print(bool(None))  # False
print(bool(1))     # True
print(bool("hi"))  # True
print(bool([1]))   # True
```

```python
# Common pattern: check if list has items
items = []
if items:
    print("Has items")
else:
    print("Empty")
```

## Common Mistakes

- **`=` vs `==`**: `if x = 5` is wrong; use `if x == 5`
- **Indentation**: Python uses indentation for blocks—must be consistent
- **Falsy gotcha**: `if not []` is True (empty list is falsy)
- **Chained comparisons**: `0 < x < 10` works in Python!
