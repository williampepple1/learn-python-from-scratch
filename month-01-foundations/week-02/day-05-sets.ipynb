## Creating Sets

Sets are unordered collections of unique elements. Use `{}` or `set()`. Empty `{}` is a dict—use `set()` for empty set.

```python
fruits = {"apple", "banana", "cherry"}
print(fruits)

# From list - removes duplicates
nums = set([1, 2, 2, 3, 3, 3])
print(nums)

empty = set()  # Not {} - that's empty dict!
print(type(empty))
```

## Set Operations: Union, Intersection, Difference

Mathematical set operations. Use methods or operators.

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a | b)   # Union: {1, 2, 3, 4}
print(a & b)   # Intersection: {2, 3}
print(a - b)   # Difference (in a, not in b): {1}
print(b - a)   # {4}
print(a ^ b)   # Symmetric difference: {1, 4}
```

```python
# Method versions
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

## Set Methods

`add()`, `remove()`, `discard()`, `clear()`, `in`

```python
s = {1, 2, 3}
s.add(4)
s.add(2)  # No duplicate
print(s)

s.remove(2)   # KeyError if not found
s.discard(99)  # No error if not found
print(s)

print(3 in s)  # True
```

## Practical Uses

Remove duplicates, membership testing, finding common elements.

```python
# Remove duplicates from list
items = [1, 2, 2, 3, 3, 3, 4]
unique = list(set(items))
print(unique)

# Fast membership (O(1) average)
valid_ids = {101, 102, 103}
print(102 in valid_ids)

# Common elements in two lists
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common = set(list1) & set(list2)
print(common)
```

## frozenset

Immutable set. Can be used as dict key or set element.

```python
fs = frozenset([1, 2, 3])
print(fs)
# fs.add(4)  # Error - immutable!

# Use as dict key
d = {frozenset([1, 2]): "pair"}
print(d[frozenset([1, 2])])
```

## Common Mistakes

- **`{}` is dict**: Use `set()` for empty set
- **Sets are unordered**: Don't rely on order
- **Unhashable in sets**: Lists can't be in sets; use tuples
- **`remove` vs `discard`**: remove raises KeyError; discard doesn't
