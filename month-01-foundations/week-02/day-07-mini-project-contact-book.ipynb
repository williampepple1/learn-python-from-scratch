# 🐍 Day 7: Mini-Project - Contact Book

Build a full CRUD contact book using:
- Dictionaries for contact storage
- Functions for each operation
- Loops and conditionals
- User input/output

## Step 1: Data Structure

Each contact is a dict with name, phone, email. We'll use name as the key for easy lookup.

```python
# contacts is a dict: name -> {phone, email}
contacts = {}

# Example contact structure
contacts["Alice"] = {"phone": "555-1234", "email": "alice@email.com"}
contacts["Bob"] = {"phone": "555-5678", "email": "bob@email.com"}
print(contacts)
```

## Step 2: Add Contact (Create)

Function to add a new contact.

```python
def add_contact(contacts, name, phone, email):
    """Add a new contact. Returns True if added, False if name already exists."""
    if name in contacts:
        return False
    contacts[name] = {"phone": phone, "email": email}
    return True

contacts = {}
add_contact(contacts, "Alice", "555-1234", "alice@email.com")
add_contact(contacts, "Bob", "555-5678", "bob@email.com")
print(contacts)
```

## Step 3: View Contact (Read)

Function to look up and display a contact.

```python
def get_contact(contacts, name):
    """Get contact by name. Returns dict or None if not found."""
    return contacts.get(name)

def display_contact(contacts, name):
    """Display a contact or message if not found."""
    contact = get_contact(contacts, name)
    if contact:
        print(f"Name:  {name}")
        print(f"Phone: {contact['phone']}")
        print(f"Email: {contact['email']}")
    else:
        print(f"Contact '{name}' not found.")

display_contact(contacts, "Alice")
display_contact(contacts, "Charlie")
```

## Step 4: List All Contacts

Display all contacts in the book.

```python
def list_contacts(contacts):
    """List all contacts."""
    if not contacts:
        print("No contacts yet.")
        return
    print("\n--- Contacts ---")
    for name, info in contacts.items():
        print(f"{name}: {info['phone']} | {info['email']}")

list_contacts(contacts)
```

## Step 5: Update Contact

Modify an existing contact's phone or email.

```python
def update_contact(contacts, name, phone=None, email=None):
    """Update contact. Only updates provided fields. Returns True if updated."""
    if name not in contacts:
        return False
    if phone is not None:
        contacts[name]["phone"] = phone
    if email is not None:
        contacts[name]["email"] = email
    return True

update_contact(contacts, "Alice", phone="555-9999")
display_contact(contacts, "Alice")
```

## Step 6: Delete Contact

Remove a contact from the book.

```python
def delete_contact(contacts, name):
    """Delete a contact. Returns True if deleted."""
    if name in contacts:
        del contacts[name]
        return True
    return False

delete_contact(contacts, "Bob")
list_contacts(contacts)
```

## Step 7: Complete Contact Book with Menu

Full implementation with a menu loop.

```python
def add_contact(contacts, name, phone, email):
    if name in contacts:
        return False
    contacts[name] = {"phone": phone, "email": email}
    return True

def get_contact(contacts, name):
    return contacts.get(name)

def display_contact(contacts, name):
    contact = get_contact(contacts, name)
    if contact:
        print(f"Name:  {name}")
        print(f"Phone: {contact['phone']}")
        print(f"Email: {contact['email']}")
    else:
        print(f"Contact '{name}' not found.")

def list_contacts(contacts):
    if not contacts:
        print("No contacts yet.")
        return
    print("\n--- Contacts ---")
    for name, info in contacts.items():
        print(f"{name}: {info['phone']} | {info['email']}")

def update_contact(contacts, name, phone=None, email=None):
    if name not in contacts:
        return False
    if phone is not None:
        contacts[name]["phone"] = phone
    if email is not None:
        contacts[name]["email"] = email
    return True

def delete_contact(contacts, name):
    if name in contacts:
        del contacts[name]
        return True
    return False

def main():
    contacts = {}
    print("=== Contact Book ===")
    print("1. Add  2. View  3. List  4. Update  5. Delete  6. Quit\n")

    while True:
        choice = input("Choice: ").strip()
        if choice == "6":
            print("Goodbye!")
            break
        elif choice == "1":
            name = input("Name: ").strip()
            phone = input("Phone: ").strip()
            email = input("Email: ").strip()
            if add_contact(contacts, name, phone, email):
                print("Added.")
            else:
                print("Name already exists.")
        elif choice == "2":
            name = input("Name: ").strip()
            display_contact(contacts, name)
        elif choice == "3":
            list_contacts(contacts)
        elif choice == "4":
            name = input("Name: ").strip()
            phone = input("New phone (enter to skip): ").strip()
            email = input("New email (enter to skip): ").strip()
            p = phone if phone else None
            e = email if email else None
            if update_contact(contacts, name, p, e):
                print("Updated.")
            else:
                print("Not found.")
        elif choice == "5":
            name = input("Name: ").strip()
            if delete_contact(contacts, name):
                print("Deleted.")
            else:
                print("Not found.")
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```

```python
# Demo without blocking - run the functions directly
contacts = {}
add_contact(contacts, "Alice", "555-1234", "alice@email.com")
add_contact(contacts, "Bob", "555-5678", "bob@email.com")
print("After add:")
list_contacts(contacts)
print("\nView Alice:")
display_contact(contacts, "Alice")
update_contact(contacts, "Alice", phone="555-9999")
print("\nAfter update:")
display_contact(contacts, "Alice")
delete_contact(contacts, "Bob")
print("\nAfter delete Bob:")
list_contacts(contacts)
```

## Key Takeaways

- **Data structure**: Dict of dicts—name as key, info as value
- **CRUD**: Create (add), Read (get/display/list), Update, Delete
- **Functions**: One function per operation for clarity
- **Menu loop**: while True with input() for choice
- **Edge cases**: Check if contact exists before update/delete
