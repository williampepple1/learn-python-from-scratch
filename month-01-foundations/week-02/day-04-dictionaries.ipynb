## Creating Dictionaries

Dictionaries store key-value pairs. Keys must be hashable (immutable); values can be anything.

```python
person = {"name": "Alice", "age": 30, "city": "NYC"}
print(person)

# Using dict() constructor
scores = dict(math=95, english=88, science=92)
print(scores)
```

## Accessing and Modifying

Use `[]` for access. `d[key]` raises KeyError if missing; use `.get()` for safe access.

```python
person = {"name": "Bob", "age": 25}
print(person["name"])
person["age"] = 26
person["email"] = "bob@example.com"  # Add new key
print(person)
```

```python
# Safe access with .get()
print(person.get("phone"))       # None (key missing)
print(person.get("phone", "N/A")) # "N/A" as default
```

## Dictionary Methods

`keys()`, `values()`, `items()`, `pop()`, `popitem()`, `update()`, `clear()`

```python
d = {"a": 1, "b": 2, "c": 3}
print(d.keys())    # dict_keys(['a', 'b', 'c'])
print(d.values())  # dict_values([1, 2, 3])
print(d.items())   # dict_items([('a', 1), ('b', 2), ('c', 3)])
```

```python
d = {"a": 1, "b": 2, "c": 3}
val = d.pop("b")  # Remove and return
print(val, d)

d.update({"d": 4, "e": 5})  # Merge
print(d)
```

## Iteration

By default, iterate over keys. Use `.items()` for key-value pairs.

```python
d = {"a": 1, "b": 2, "c": 3}
for key in d:
    print(key, d[key])

for key, value in d.items():
    print(f"{key}: {value}")
```

## Nested Dictionaries

Values can be dictionaries. Access with chained `[]`.

```python
users = {
    "alice": {"age": 30, "city": "NYC"},
    "bob": {"age": 25, "city": "LA"}
}
print(users["alice"]["city"])
users["bob"]["email"] = "bob@example.com"
print(users["bob"])
```

## Dict Comprehensions

`{key_expr: value_expr for item in iterable}`

```python
words = ["apple", "banana", "cherry"]
lengths = {w: len(w) for w in words}
print(lengths)

# With condition
nums = [1, 2, 3, 4, 5]
even_squares = {n: n**2 for n in nums if n % 2 == 0}
print(even_squares)
```

## Common Mistakes

- **Mutable keys**: Lists can't be keys; use tuples instead
- **KeyError**: Use `.get()` if key might not exist
- **Iterating and modifying**: Don't change dict size while iterating; iterate over copy
- **`in` checks keys**: `"a" in d` checks keys, not values

## Key Takeaways

- **Dict**: `{k: v}`; keys must be hashable
- **Access**: `d[key]` or `d.get(key, default)`
- **Methods**: keys(), values(), items(), pop(), update()
- **Iterate**: `for k, v in d.items()`
- **Nested**: `d["a"]["b"]`
