## Series

A 1D labeled array. Index can be custom (not just 0,1,2...).


```python
import pandas as pd

s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)
print(s['b'])
print(s.values)
```

## DataFrame

2D labeled structure with columns and index. Like a spreadsheet.


```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
print(df)
print(df['A'])
print(df.loc[0])
```

## Loading Data

Read CSV, Excel, JSON. Use `read_csv`, `read_excel`, `read_json`.


```python
df = pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})
df.to_csv('temp_data.csv', index=False)
df2 = pd.read_csv('temp_data.csv')
print(df2)
```

## Common Mistakes

- **SettingWithCopyWarning**: Use `.loc` or `.copy()` when modifying
- **Mixed loc/iloc**: Use loc for labels, iloc for integer positions
- **Forgetting index=False**: When saving CSV, index=False avoids extra column
