## Broadcasting

NumPy broadcasts arrays of different shapes during arithmetic. Smaller arrays are "stretched" to match larger ones.


```python
import numpy as np

# Row vector + column vector -> 2D result
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)
```

## Fancy Indexing

Use arrays of indices or boolean masks to select elements.


```python
arr = np.array([10, 20, 30, 40, 50])
indices = [0, 2, 4]
print(arr[indices])

# Boolean indexing
mask = arr > 25
print(arr[mask])
```

## Linear Algebra

`np.linalg` provides matrix operations: inverse, determinant, eigenvalues, etc.


```python
m = np.array([[1, 2], [3, 4]])
print('Inverse:', np.linalg.inv(m))
print('Determinant:', np.linalg.det(m))
print('Dot product:', np.dot(m, np.linalg.inv(m)))
```

## Common Mistakes

- **Broadcasting shape errors**: Dimensions must be compatible (1 or same size)
- **Fancy indexing returns copy**: Modifying result doesn't change original
- **np.linalg.inv on singular matrix**: Raises LinAlgError; check det first
