## Why NumPy?

NumPy provides fast, memory-efficient arrays for numerical computing. It's the foundation of the Python data science stack.

## Creating Arrays

```python
import numpy as np

# From list
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(type(arr))

# 2D array
matrix = np.array([[1, 2], [3, 4]])
print(matrix)
```

## Array dtypes

NumPy infers dtype from input. You can specify it explicitly.

```python
arr_int = np.array([1, 2, 3], dtype=np.int32)
arr_float = np.array([1.0, 2.0], dtype=np.float64)
arr_str = np.array(['a', 'b'], dtype='U1')
print(arr_int.dtype, arr_float.dtype, arr_str.dtype)
```

## Array Creation Helpers

```python
zeros = np.zeros((2, 3))
ones = np.ones((2, 2))
arange = np.arange(0, 10, 2)
linspace = np.linspace(0, 1, 5)
random_arr = np.random.rand(3, 3)
print('zeros:', zeros)
print('arange:', arange)
print('linspace:', linspace)
```

## Vectorized Operations

Operations apply element-wise without Python loops.

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])
print('a + b:', a + b)
print('a * 2:', a * 2)
print('a ** 2:', a ** 2)
print('np.sqrt(a):', np.sqrt(a))
```

## Indexing and Slicing

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0], arr[-1])
print(arr[1:4])
print(arr[::2])

m = np.array([[1, 2, 3], [4, 5, 6]])
print(m[1, 2])
print(m[:, 1])
```

## Common Mistakes

- **Modifying views**: Slicing returns a view; changes affect original. Use `.copy()` if needed.
- **Mixed types**: Lists with mixed types become object arrays; prefer homogeneous data.
- **Integer division**: `arr / 2` gives float; use `arr // 2` for integer.
- **Shape mismatch**: Vectorized ops need compatible shapes (broadcasting).
