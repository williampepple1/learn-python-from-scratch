## Project Overview

We'll analyze a real dataset using NumPy, Pandas, and Matplotlib. We'll load data, explore it, visualize key patterns, and draw insights.

## Step 1: Setup and Load Data

Import libraries and create or load a sample dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create sample sales dataset
np.random.seed(42)
n = 100
data = {
    'date': pd.date_range('2024-01-01', periods=n, freq='D'),
    'revenue': np.random.uniform(1000, 5000, n).round(2),
    'units': np.random.randint(10, 100, n),
    'region': np.random.choice(['North', 'South', 'East', 'West'], n)
}
df = pd.DataFrame(data)
df.head(10)
```

## Step 2: Exploratory Data Analysis

Examine shape, types, and basic statistics.

```python
print('Shape:', df.shape)
print('\nInfo:')
df.info()
print('\nDescribe:')
df.describe()
```

## Step 3: Aggregate by Region

Use groupby to summarize by region.

```python
region_stats = df.groupby('region').agg({
    'revenue': ['sum', 'mean', 'count'],
    'units': ['sum', 'mean']
}).round(2)
print(region_stats)
```

## Step 4: Visualize Revenue Over Time

Line plot of daily revenue.

```python
plt.figure(figsize=(10, 4))
plt.plot(df['date'], df['revenue'], alpha=0.7)
plt.title('Daily Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Step 5: Bar Chart by Region

Compare total revenue by region.

```python
revenue_by_region = df.groupby('region')['revenue'].sum()
plt.figure(figsize=(6, 4))
revenue_by_region.plot(kind='bar', color=['#2ecc71', '#3498db', '#e74c3c', '#9b59b6'])
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Step 6: Scatter Plot - Revenue vs Units

Explore relationship between revenue and units sold.

```python
plt.figure(figsize=(6, 5))
for region in df['region'].unique():
    subset = df[df['region'] == region]
    plt.scatter(subset['units'], subset['revenue'], label=region, alpha=0.6)
plt.legend()
plt.title('Revenue vs Units Sold by Region')
plt.xlabel('Units')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

## Step 7: Key Insights Summary

Write a function to summarize findings.

```python
def summarize_insights(df):
    total_revenue = df['revenue'].sum()
    avg_revenue = df['revenue'].mean()
    best_region = df.groupby('region')['revenue'].sum().idxmax()
    print(f'Total Revenue: ${total_revenue:,.2f}')
    print(f'Average Daily Revenue: ${avg_revenue:,.2f}')
    print(f'Top Region: {best_region}')

summarize_insights(df)
```

## Step 8: Complete Code - Full Analysis Pipeline

All steps combined into a single runnable pipeline.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
df = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=100, freq='D'),
    'revenue': np.random.uniform(1000, 5000, 100).round(2),
    'units': np.random.randint(10, 100, 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100)
})

print('=== EDA ===')
print(df.describe())
print('\n=== Region Summary ===')
print(df.groupby('region').agg({'revenue': 'sum', 'units': 'sum'}))
print('\n=== Insights ===')
print(f"Best region: {df.groupby('region')['revenue'].sum().idxmax()}")
```
