## Project Overview

Build a high-performance data processor that processes large arrays using NumPy vectorization, batch processing, and efficient memory use.

## Step 1: Setup and Data Generation

Create sample data for processing.

```python
import numpy as np

# Generate 100k rows of sample data
np.random.seed(42)
n = 100_000
data = np.random.randn(n, 5)  # 5 columns
print(f'Shape: {data.shape}')
print(f'Memory: {data.nbytes / 1024:.1f} KB')
```

## Step 2: Vectorized Transformations

Apply operations without Python loops.

```python
def process_vectorized(arr):
    """Normalize each column and compute row sums."""
    normalized = (arr - arr.mean(axis=0)) / (arr.std(axis=0) + 1e-8)
    row_sums = normalized.sum(axis=1)
    return normalized, row_sums

norm, sums = process_vectorized(data)
print('Normalized shape:', norm.shape)
print('Row sums sample:', sums[:5])
```

## Step 3: Batch Processing

Process in chunks for memory efficiency.

```python
def process_in_batches(arr, batch_size=10_000):
    results = []
    for i in range(0, len(arr), batch_size):
        batch = arr[i:i+batch_size]
        batch_norm = (batch - batch.mean(axis=0)) / (batch.std(axis=0) + 1e-8)
        results.append(batch_norm.sum(axis=1))
    return np.concatenate(results)

batch_sums = process_in_batches(data)
print('Batch result length:', len(batch_sums))
```

## Step 4: Filtering and Aggregation

Use boolean indexing for fast filtering.

```python
# Filter rows where first column > 0
mask = data[:, 0] > 0
filtered = data[mask]
print(f'Filtered rows: {len(filtered)} / {len(data)}')

# Aggregate: mean per column for filtered data
agg = filtered.mean(axis=0)
print('Column means:', agg.round(4))
```

## Step 5: Complete Processor Class

Full runnable data processor.

```python
class DataProcessor:
    def __init__(self, batch_size=10_000):
        self.batch_size = batch_size

    def normalize(self, arr):
        return (arr - arr.mean(axis=0)) / (arr.std(axis=0) + 1e-8)

    def process(self, arr):
        norm = self.normalize(arr)
        return norm, norm.sum(axis=1)

processor = DataProcessor()
norm, sums = processor.process(data)
print('Processor ready. Processed', len(sums), 'rows.')
```

## Key Takeaways

- Vectorize: avoid Python loops, use NumPy ops
- Batch processing for large data
- Boolean indexing for fast filtering
- Monitor memory with .nbytes
