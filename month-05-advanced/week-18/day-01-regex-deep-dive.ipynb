## The re Module

Python's `re` module provides regex support. Key functions: `search`, `match`, `findall`, `sub`, `compile`.

```python
import re

text = 'Contact: alice@example.com or bob@test.org'
match = re.search(r'\w+@\w+\.\w+', text)
if match:
    print('Found:', match.group())

all_emails = re.findall(r'\w+@\w+\.\w+', text)
print('All:', all_emails)
```

## Common Patterns

`\d` digits, `\w` word chars, `\s` whitespace, `+` one or more, `*` zero or more, `[]` character class.

```python
# Extract phone numbers (simplified)
text = 'Call 555-1234 or 555-5678'
phones = re.findall(r'\d{3}-\d{4}', text)
print(phones)

# Extract words
words = re.findall(r'\b\w+\b', 'Hello, world!')
print(words)
```

## Capturing Groups

Use parentheses `()` to capture parts of a match. `match.group(1)`, `match.group(2)` for subgroups.

```python
text = 'Date: 2024-03-10'
m = re.search(r'(\d{4})-(\d{2})-(\d{2})', text)
if m:
    print('Year:', m.group(1), 'Month:', m.group(2), 'Day:', m.group(3))
    print('Full match:', m.group(0))
```

## Lookahead and Lookbehind

`(?=...)` positive lookahead, `(?!...)` negative lookahead. Match without consuming.

```python
# Match 'foo' only if followed by 'bar'
text = 'foobar foobaz'
m = re.findall(r'foo(?=bar)', text)
print(m)

# Negative lookahead: digits not followed by %
text2 = '50% off 100 items'
m2 = re.findall(r'\d+(?!%)', text2)
print(m2)
```

## Substitution with re.sub

Replace matches. Use `\1`, `\2` for captured groups in replacement.

```python
text = 'Hello 123 World 456'
result = re.sub(r'\d+', 'X', text)
print(result)

# Swap with groups
text2 = 'Smith, John'
result2 = re.sub(r'(\w+), (\w+)', r'\2 \1', text2)
print(result2)
```

## re.compile for Repeated Use

Compile regex once for better performance when using the same pattern repeatedly.

```python
pattern = re.compile(r'\b\w{4,}\b')
text = 'The quick brown fox jumps over the lazy dog'
print(pattern.findall(text))
```

## Common Mistakes

- **Raw strings**: Use `r'\d'` not `'\d'` to avoid escape issues
- **Greedy vs lazy**: `.*` is greedy; use `.*?` for lazy
- **match vs search**: `match` only at start; `search` finds first anywhere
- **Unanchored patterns**: `^` and `$` for start/end of string
