## Why Type Hints?

Type hints document expected types, enable static checking (mypy), and improve IDE support. They don't affect runtime (by default).

## Basic Annotations

Annotate parameters and return values with `: Type` and `-> ReturnType`.

```python
def greet(name: str) -> str:
    return f"Hello, {name}!"

def add(a: int, b: int) -> int:
    return a + b

def process(items: list[int]) -> int:
    return len(items)

print(greet("Alice"))
print(add(1, 2))
print(process([1, 2, 3]))
```

## Optional and Union

`Optional[X]` = `Union[X, None]`. For values that may be None.

```python
from typing import Optional, Union

def find_index(lst: list[str], target: str) -> Optional[int]:
    for i, x in enumerate(lst):
        if x == target:
            return i
    return None

def parse_id(value: Union[str, int]) -> int:
    if isinstance(value, str):
        return int(value)
    return value

# Python 3.10+: use X | Y instead of Union[X, Y]
def parse_id_310(value: str | int) -> int:
    return int(value) if isinstance(value, str) else value
```

## List, Dict, Tuple, Set

Use `list[T]`, `dict[K, V]` (Python 3.9+). For older Python: `List[T]` from typing.

```python
def merge_dicts(a: dict[str, int], b: dict[str, int]) -> dict[str, int]:
    return {**a, **b}

def first_and_rest(items: list[str]) -> tuple[str, list[str]]:
    return items[0], items[1:]

def unique(items: list[int]) -> set[int]:
    return set(items)
```

## Callable and TypeAlias

`Callable[[ArgTypes], ReturnType]` for functions. `TypeAlias` for complex types.

```python
from typing import Callable, TypeAlias

Processor: TypeAlias = Callable[[str], str]

def apply(func: Callable[[int, int], int], a: int, b: int) -> int:
    return func(a, b)

def process_string(s: str, processor: Processor) -> str:
    return processor(s)

print(apply(lambda x, y: x + y, 2, 3))
```

## Generic Classes

Use `Generic[T]` for parameterized types.

```python
from typing import Generic, TypeVar

T = TypeVar('T')

class Box(Generic[T]):
    def __init__(self, value: T):
        self.value = value

    def get(self) -> T:
        return self.value

b: Box[int] = Box(42)
print(b.get())
```

## mypy

Static type checker. `pip install mypy` then `mypy myfile.py`.

```python
def add(a: int, b: int) -> int:
    return a + b

# add("1", "2")  # mypy error: incompatible type
# add(1, 2)  # OK

result = add(1, 2)
# result + "hello"  # mypy error: Unsupported operand types
```

## Any and NoReturn

`Any` opts out of type checking. `NoReturn` for functions that never return (e.g., sys.exit).

```python
from typing import Any, NoReturn

def flexible(x: Any) -> Any:
    return x

def exit_app() -> NoReturn:
    raise SystemExit(1)
```

## Common Mistakes

- **Using List instead of list**: In Python 3.9+ use built-in list[int]
- **Optional for required params**: Use Optional only when None is valid
- **Overusing Any**: Defeats type checking
- **Forgetting to run mypy**: Add to CI/pre-commit
