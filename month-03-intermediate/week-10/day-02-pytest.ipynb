## Why pytest?

pytest is simpler: no class required, better output, fixtures, parametrization. Install with `pip install pytest`.

```python
# Run: pip install pytest
def add(a, b):
    return a + b

def test_add():
    assert add(2, 3) == 5

def test_add_negative():
    assert add(-1, -1) == -2
```

## Basic Fixtures

Fixtures provide test data. Use `@pytest.fixture` and inject by name.

```python
import pytest

@pytest.fixture
def sample_list():
    return [1, 2, 3]

def test_sum(sample_list):
    assert sum(sample_list) == 6

def test_len(sample_list):
    assert len(sample_list) == 3
```

## Fixture Scope

`scope='function'` (default), `'class'`, `'module'`, `'session'`.

```python
import pytest

@pytest.fixture(scope='module')
def shared_resource():
    print("Setup (once per module)")
    resource = "data"
    yield resource
    print("Teardown")

def test_one(shared_resource):
    assert shared_resource == "data"

def test_two(shared_resource):
    assert shared_resource == "data"
```

## Parametrize

Run the same test with different inputs. Reduces duplication.

```python
import pytest

def add(a, b):
    return a + b

@pytest.mark.parametrize("a,b,expected", [
    (2, 3, 5),
    (0, 0, 0),
    (-1, 1, 0),
    (10, -5, 5),
])
def test_add_parametrized(a, b, expected):
    assert add(a, b) == expected
```

## Parametrize with ids

Add readable test IDs for clearer output.

```python
import pytest

@pytest.mark.parametrize("x,expected", [
    (1, 1),
    (2, 4),
    (3, 9),
], ids=["one", "two", "three"])
def test_square(x, expected):
    assert x ** 2 == expected
```

## Markers

Tag tests: `@pytest.mark.slow`, `@pytest.mark.skip`, `@pytest.mark.xfail`.

```python
import pytest

@pytest.mark.slow
def test_slow_operation():
    import time
    time.sleep(0.1)
    assert True

@pytest.mark.skip(reason="Not implemented yet")
def test_future_feature():
    assert False

@pytest.mark.xfail(reason="Known bug")
def test_known_bug():
    assert 1 == 2  # Fails but xfail so test passes
```

## Running Marked Tests

`pytest -m slow` runs only tests marked `slow`. Register custom markers in `pytest.ini` or `pyproject.toml`.

```python
# pytest.ini:
# [pytest]
# markers =
#     slow: marks tests as slow
#     integration: marks integration tests

# Command: pytest -m "not slow"  # exclude slow tests
```

## Fixture Composition

Fixtures can use other fixtures.

```python
import pytest

@pytest.fixture
def base_data():
    return {"count": 0}

@pytest.fixture
def extended_data(base_data):
    base_data["extra"] = "value"
    return base_data

def test_extended(extended_data):
    assert extended_data["count"] == 0
    assert extended_data["extra"] == "value"
```

## Common Mistakes

- **Fixture name typo**: Must match parameter name exactly
- **Mutable fixture**: Each test gets same object by default; use factory for fresh copies
- **Expensive fixture in function scope**: Use scope='module' or 'session'
- **Unregistered markers**: Add to pytest.ini to avoid warnings
