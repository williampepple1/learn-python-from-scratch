## Why Code Quality Tools?

Linters find bugs and style issues. Formatters enforce consistent style. Together they keep codebases maintainable.

## pylint

Checks code for errors, style, and refactoring suggestions. `pip install pylint`

```python
# pylint myfile.py
# pylint --disable=C0114,C0115 myfile.py  # disable specific checks

def bad_function( x,y ):  # Bad spacing
    unused_var = 1
    return x+y

# pylint would report: unused variable, spacing, missing docstring
```

## flake8

Lightweight: combines PyFlakes, pycodestyle, McCabe. Fast and widely used.

```python
# pip install flake8
# flake8 myfile.py
# flake8 --max-line-length=100 .

x=1+2  # E225: missing whitespace around operator
def f():
    pass
    pass  # F401: unreachable

# .flake8 config:
# [flake8]
# max-line-length = 88
# exclude = .git,__pycache__
```

## black

Opinionated formatter. No configuration needed. `black .` reformats all Python files.

```python
# pip install black
# black .
# black --check .  # Check only, don't modify

# Before black:
def foo( a,b,c ):
    return {"x":a,"y":b,"z":c}

# After black:
def foo(a, b, c):
    return {"x": a, "y": b, "z": c}
```

## isort

Sorts and groups imports. `pip install isort`

```python
# isort myfile.py

# Before:
# import os
# from collections import defaultdict
# import sys

# After isort:
# import os
# import sys
#
# from collections import defaultdict
```

## pyproject.toml Configuration

Centralize tool config in `pyproject.toml`.

```python
# pyproject.toml
# [tool.black]
# line-length = 88
# target-version = ['py310']
#
# [tool.isort]
# profile = "black"
# line_length = 88
#
# [tool.pytest.ini_options]
# testpaths = ["tests"]
```

## pre-commit Hooks

Run checks before each commit. `pip install pre-commit`

```python
# .pre-commit-config.yaml
# repos:
#   - repo: https://github.com/psf/black
#     rev: 24.1.1
#     hooks:
#       - id: black
#   - repo: https://github.com/pycqa/isort
#     rev: 5.13.2
#     hooks:
#       - id: isort
#   - repo: https://github.com/pycqa/flake8
#     rev: 7.0.0
#     hooks:
#       - id: flake8
#
# Run: pre-commit install
# Then: pre-commit run --all-files
```

## Running Tools

Typical workflow: format first (black, isort), then lint (flake8/pylint).

```python
# Common commands:
# black .
# isort .
# flake8 .
# pylint src/

# Or in one script:
# black . && isort . && flake8 .
```

## Common Mistakes

- **Fighting the formatter**: Configure sparingly; black's defaults are good
- **Skipping pre-commit**: Use `--no-verify` only when necessary
- **Conflicting configs**: Use pyproject.toml for consistency
- **Too many linter disables**: Fix the issue or disable with a comment explaining why
