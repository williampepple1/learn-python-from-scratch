## Step 1: Stats - Mean

Implement mean with tests first (TDD).

```python
def mean(values: list[float]) -> float:
    if not values:
        raise ValueError("mean requires at least one value")
    return sum(values) / len(values)

def test_mean():
    assert mean([1, 2, 3, 4, 5]) == 3.0
    assert mean([10]) == 10.0
    assert mean([0, 0, 0]) == 0.0

test_mean()
print("mean tests passed")
```

## Step 2: Stats - Median

Median: middle value (or average of two middles for even length).

```python
def median(values: list[float]) -> float:
    if not values:
        raise ValueError("median requires at least one value")
    sorted_vals = sorted(values)
    n = len(sorted_vals)
    mid = n // 2
    if n % 2 == 1:
        return sorted_vals[mid]
    return (sorted_vals[mid - 1] + sorted_vals[mid]) / 2

def test_median():
    assert median([1, 2, 3, 4, 5]) == 3.0
    assert median([1, 2, 3, 4]) == 2.5
    assert median([5]) == 5.0

test_median()
print("median tests passed")
```

## Step 3: Stats - Variance and Std Dev

Variance: average of squared deviations from mean. Std dev: sqrt(variance).

```python
def variance(values: list[float], ddof: int = 0) -> float:
    if len(values) < 2 and ddof > 0:
        raise ValueError("need at least 2 values for sample variance")
    if not values:
        raise ValueError("variance requires at least one value")
    m = mean(values)
    n = len(values)
    return sum((x - m) ** 2 for x in values) / (n - ddof)

def std_dev(values: list[float], ddof: int = 0) -> float:
    return variance(values, ddof) ** 0.5

def test_variance_std():
    vals = [2, 4, 4, 4, 5, 5, 7, 9]
    assert abs(variance(vals) - 4.0) < 0.001
    assert abs(std_dev(vals) - 2.0) < 0.001

test_variance_std()
print("variance/std tests passed")
```

## Step 4: Linear Algebra - Dot Product

Dot product of two vectors: sum(a[i] * b[i]).

```python
def dot(a: list[float], b: list[float]) -> float:
    if len(a) != len(b):
        raise ValueError("vectors must have same length")
    if not a:
        raise ValueError("vectors must not be empty")
    return sum(x * y for x, y in zip(a, b))

def test_dot():
    assert dot([1, 2, 3], [4, 5, 6]) == 32
    assert dot([1], [1]) == 1
    assert dot([0, 0], [1, 1]) == 0

test_dot()
print("dot tests passed")
```

## Step 5: Matrix Multiply (2x2)

Multiply two 2x2 matrices. Result[i][j] = sum over k of A[i][k] * B[k][j].

```python
def matmul2x2(a: list[list[float]], b: list[list[float]]) -> list[list[float]]:
    if len(a) != 2 or len(a[0]) != 2 or len(b) != 2 or len(b[0]) != 2:
        raise ValueError("matrices must be 2x2")
    return [
        [a[0][0] * b[0][0] + a[0][1] * b[1][0], a[0][0] * b[0][1] + a[0][1] * b[1][1]],
        [a[1][0] * b[0][0] + a[1][1] * b[1][0], a[1][0] * b[0][1] + a[1][1] * b[1][1]],
    ]

def test_matmul():
    a = [[1, 2], [3, 4]]
    b = [[5, 6], [7, 8]]
    result = matmul2x2(a, b)
    assert result[0][0] == 19
    assert result[0][1] == 22
    assert result[1][0] == 43
    assert result[1][1] == 50

test_matmul()
print("matmul tests passed")
```

## Step 6: Complete Library with unittest

Wrap all tests in unittest for proper test discovery.

```python
import unittest

class TestMathLib(unittest.TestCase):
    def test_mean_empty_raises(self):
        with self.assertRaises(ValueError):
            mean([])

    def test_median_empty_raises(self):
        with self.assertRaises(ValueError):
            median([])

    def test_dot_mismatched_length_raises(self):
        with self.assertRaises(ValueError):
            dot([1, 2], [1, 2, 3])

    def test_matmul_non_2x2_raises(self):
        with self.assertRaises(ValueError):
            matmul2x2([[1]], [[1]])

unittest.main(argv=[''], exit=False, verbosity=1)
```

## Step 7: Full Library Code

All functions in one module.

```python
"""math_lib - Statistics and linear algebra utilities."""

def mean(values: list[float]) -> float:
    if not values:
        raise ValueError("mean requires at least one value")
    return sum(values) / len(values)

def median(values: list[float]) -> float:
    if not values:
        raise ValueError("median requires at least one value")
    s = sorted(values)
    n = len(s)
    mid = n // 2
    return s[mid] if n % 2 else (s[mid - 1] + s[mid]) / 2

def variance(values: list[float], ddof: int = 0) -> float:
    if not values:
        raise ValueError("variance requires at least one value")
    m = mean(values)
    n = len(values)
    return sum((x - m) ** 2 for x in values) / (n - ddof)

def std_dev(values: list[float], ddof: int = 0) -> float:
    return variance(values, ddof) ** 0.5

def dot(a: list[float], b: list[float]) -> float:
    if len(a) != len(b):
        raise ValueError("vectors must have same length")
    if not a:
        raise ValueError("vectors must not be empty")
    return sum(x * y for x, y in zip(a, b))

def matmul2x2(a: list[list[float]], b: list[list[float]]) -> list[list[float]]:
    if len(a) != 2 or len(a[0]) != 2 or len(b) != 2 or len(b[0]) != 2:
        raise ValueError("matrices must be 2x2")
    return [
        [a[0][0]*b[0][0] + a[0][1]*b[1][0], a[0][0]*b[0][1] + a[0][1]*b[1][1]],
        [a[1][0]*b[0][0] + a[1][1]*b[1][0], a[1][0]*b[0][1] + a[1][1]*b[1][1]],
    ]
```
