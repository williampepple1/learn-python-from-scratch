## Why Test?

Tests verify your code works, catch regressions, and document behavior. Python's `unittest` is the standard library testing framework.

## Basic TestCase

Subclass `unittest.TestCase` and define methods starting with `test_`.

```python
import unittest

def add(a, b):
    return a + b

class TestAdd(unittest.TestCase):
    def test_add_positive(self):
        self.assertEqual(add(2, 3), 5)

    def test_add_negative(self):
        self.assertEqual(add(-1, -1), -2)

    def test_add_zero(self):
        self.assertEqual(add(0, 0), 0)

unittest.main(argv=[''], exit=False, verbosity=2)
```

## Common Assertions

`assertEqual`, `assertTrue`, `assertFalse`, `assertRaises`, `assertIn`, `assertIsNone`, etc.

```python
import unittest

class TestAssertions(unittest.TestCase):
    def test_basics(self):
        self.assertEqual(1 + 1, 2)
        self.assertTrue(1 < 2)
        self.assertFalse(1 > 2)
        self.assertIsNone(None)
        self.assertIn(2, [1, 2, 3])

    def test_raises(self):
        with self.assertRaises(ValueError):
            int("not a number")

    def test_raises_with_message(self):
        with self.assertRaises(ValueError) as ctx:
            int("x")
        self.assertIn("invalid literal", str(ctx.exception))

unittest.main(argv=[''], exit=False, verbosity=1)
```

## setUp and tearDown

`setUp` runs before each test; `tearDown` runs after. Use for fixtures (DB connections, temp files).

```python
import unittest

class TestWithFixture(unittest.TestCase):
    def setUp(self):
        self.data = [1, 2, 3]

    def tearDown(self):
        self.data = None

    def test_sum(self):
        self.assertEqual(sum(self.data), 6)

    def test_len(self):
        self.assertEqual(len(self.data), 3)

unittest.main(argv=[''], exit=False, verbosity=1)
```

## setUpClass and tearDownClass

Run once per test class (expensive setup).

```python
import unittest

class TestClassFixture(unittest.TestCase):
    @classmethod
    def setUpClass(cls):
        cls.shared = "expensive_resource"

    @classmethod
    def tearDownClass(cls):
        cls.shared = None

    def test_one(self):
        self.assertEqual(self.shared, "expensive_resource")

    def test_two(self):
        self.assertIn("resource", self.shared)

unittest.main(argv=[''], exit=False, verbosity=1)
```

## Testing Exceptions

Use `assertRaises` as context manager for flexibility.

```python
import unittest

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b

class TestDivide(unittest.TestCase):
    def test_normal(self):
        self.assertAlmostEqual(divide(10, 2), 5.0)

    def test_zero_raises(self):
        with self.assertRaises(ValueError):
            divide(1, 0)

    def test_zero_message(self):
        with self.assertRaises(ValueError) as ctx:
            divide(1, 0)
        self.assertEqual(str(ctx.exception), "Cannot divide by zero")

unittest.main(argv=[''], exit=False, verbosity=1)
```

## assertAlmostEqual for Floats

Floats have precision issues. Use `assertAlmostEqual` with optional `places` or `delta`.

```python
import unittest

class TestFloats(unittest.TestCase):
    def test_float(self):
        self.assertAlmostEqual(0.1 + 0.2, 0.3, places=7)

    def test_float_delta(self):
        self.assertAlmostEqual(1.0, 1.0000001, delta=0.001)

unittest.main(argv=[''], exit=False, verbosity=1)
```

## Common Mistakes

- **Test methods must start with test_**: Otherwise they're not run
- **Using assert instead of self.assertEqual**: unittest gives better failure messages
- **Shared mutable state**: Each test gets fresh setUp; don't rely on order
- **Testing implementation details**: Test behavior, not internals
