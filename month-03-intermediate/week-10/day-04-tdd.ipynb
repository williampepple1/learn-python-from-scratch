## What is TDD?

Test-Driven Development: write tests first, then write the minimum code to pass, then refactor. Cycle: Red (fail) → Green (pass) → Refactor.

## The Red-Green-Refactor Cycle

1. **Red**: Write a failing test for the next behavior
2. **Green**: Write the simplest code to make it pass
3. **Refactor**: Improve code while keeping tests green

```python
import unittest

# Step 1: Write test first (RED)
class TestFizzBuzz(unittest.TestCase):
    def test_returns_1_for_1(self):
        self.assertEqual(fizzbuzz(1), "1")

# Step 2: Minimal implementation (GREEN)
def fizzbuzz(n):
    return "1"  # Simplest to pass

unittest.main(argv=[''], exit=False, verbosity=1)
```

## Add More Tests

Each test drives the next piece of behavior.

```python
def fizzbuzz(n):
    if n % 3 == 0 and n % 5 == 0:
        return "FizzBuzz"
    if n % 3 == 0:
        return "Fizz"
    if n % 5 == 0:
        return "Buzz"
    return str(n)

class TestFizzBuzz(unittest.TestCase):
    def test_1(self):
        self.assertEqual(fizzbuzz(1), "1")

    def test_3(self):
        self.assertEqual(fizzbuzz(3), "Fizz")

    def test_5(self):
        self.assertEqual(fizzbuzz(5), "Buzz")

    def test_15(self):
        self.assertEqual(fizzbuzz(15), "FizzBuzz")

unittest.main(argv=[''], exit=False, verbosity=1)
```

## TDD Example: Stack

Build a Stack class step by step with tests first.

```python
import unittest

class TestStack(unittest.TestCase):
    def test_new_stack_is_empty(self):
        s = Stack()
        self.assertTrue(s.is_empty())

    def test_push_then_peek(self):
        s = Stack()
        s.push(1)
        self.assertEqual(s.peek(), 1)
        self.assertFalse(s.is_empty())

    def test_push_pop(self):
        s = Stack()
        s.push(1)
        self.assertEqual(s.pop(), 1)
        self.assertTrue(s.is_empty())

    def test_pop_empty_raises(self):
        s = Stack()
        with self.assertRaises(IndexError):
            s.pop()

class Stack:
    def __init__(self):
        self._items = []

    def is_empty(self):
        return len(self._items) == 0

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self._items.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("peek on empty stack")
        return self._items[-1]

unittest.main(argv=[''], exit=False, verbosity=1)
```

## One Test at a Time

Write one test, make it pass, refactor. Don't rush to implement everything.

```python
# Good order for Stack:
# 1. test_new_stack_is_empty
# 2. test_push_makes_not_empty
# 3. test_push_pop_returns_pushed
# 4. test_pop_empty_raises
# 5. test_peek
# 6. test_LIFO (multiple push/pop)
```

## Refactoring Safely

With tests, refactor confidently. Run tests after each change.

```python
# Before refactor:
def process(items):
    result = []
    for i in items:
        if i > 0:
            result.append(i * 2)
    return result

# After refactor (tests ensure same behavior):
def process(items):
    return [i * 2 for i in items if i > 0]

assert process([1, -1, 2, 0, 3]) == [2, 4, 6]
```

## TDD Benefits

- Forces clear API design (you write usage first)
- Prevents over-engineering (only implement what tests need)
- Living documentation
- Safe refactoring

## Common Mistakes

- **Writing too many tests at once**: One at a time
- **Testing implementation**: Test behavior, not internals
- **Skipping refactor step**: Don't leave duplicate code
- **Big steps**: Small steps, quick feedback
