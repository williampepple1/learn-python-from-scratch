## What Are Decorators?

Decorators are functions that wrap other functions to add or modify behavior. They use the `@decorator` syntax and are a powerful way to apply cross-cutting concerns (logging, timing, caching) without cluttering the original function.

## Functions as First-Class Objects

In Python, functions are first-class: they can be passed as arguments, returned from functions, and assigned to variables. This is the foundation of decorators.

```python
def greet(name):
    return f"Hello, {name}!"

# Function as object
my_func = greet
print(my_func("Alice"))

# Function as argument
def apply_twice(f, x):
    return f(f(x))

def add_one(x):
    return x + 1
print(apply_twice(add_one, 5))  # 7
```

## Writing a Simple Decorator

A decorator is a function that takes a function and returns a new function (usually a wrapper).

```python
def bold_decorator(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return f"**{result}**"
    return wrapper

@bold_decorator
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # **Hello, Alice!**
```

## Timing Decorator

Measure how long a function takes to execute. Useful for profiling and optimization.

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    total = 0
    for i in range(n):
        total += i
    return total

slow_sum(1_000_000)
```

## Logging Decorator

Log function calls with arguments and return values for debugging.

```python
def log_calls(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}(args={args}, kwargs={kwargs})")
        result = func(*args, **kwargs)
        print(f"  -> returned {result}")
        return result
    return wrapper

@log_calls
def add(a, b):
    return a + b

add(3, 5)
add(10, b=20)
```

## Retry Decorator

Retry a function on failure (e.g., network requests, file I/O).

```python
import random

def retry(max_attempts=3):
    def decorator(func):
        def wrapper(*args, **kwargs):
            last_exception = None
            for attempt in range(max_attempts):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    last_exception = e
                    print(f"Attempt {attempt + 1} failed: {e}")
            raise last_exception
        return wrapper
    return decorator

@retry(max_attempts=5)
def flaky_api_call():
    if random.random() < 0.7:
        raise ConnectionError("Network error")
    return "Success!"

# May need several runs to see success
print(flaky_api_call())
```

## Decorator Without Arguments

When the decorator takes no arguments, it receives the function directly. When it takes arguments (like `retry(max_attempts=5)`), it returns a decorator that receives the function.

```python
# No args: @decorator  -> decorator(func)
# With args: @decorator(args)  -> decorator(args)(func)

def simple_decorator(func):
    return func  # Identity decorator

def decorator_with_args(tag):
    def actual_decorator(func):
        def wrapper(*args, **kwargs):
            return f"<{tag}>{func(*args, **kwargs)}</{tag}>"
        return wrapper
    return actual_decorator

@decorator_with_args("div")
def get_text():
    return "Hello"

print(get_text())  # <div>Hello</div>
```

## Common Mistakes

- **Forgetting to return the wrapper**: Decorator must return a callable
- **Not passing through args/kwargs**: Use `*args, **kwargs` in wrapper
- **Not returning the result**: Wrapper should `return func(*args, **kwargs)`
- **Losing function metadata**: Use `functools.wraps` (next lesson)
- **Decorator with args**: Must return a decorator that takes `func`

## Key Takeaways

- Decorators wrap functions to add behavior
- `@decorator` is syntactic sugar for `func = decorator(func)`
- Use `*args, **kwargs` to pass through all arguments
- Timing, logging, retry are common real-world decorators
- Decorators with args: `decorator(args)(func)`
