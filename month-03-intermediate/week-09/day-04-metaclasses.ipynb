## type() Creates Classes

`type(name, bases, dict)` dynamically creates a class.

```python
class Foo:
    x = 1
    def bar(self): return self.x

Foo2 = type('Foo2', (), {'x': 1, 'bar': lambda self: self.x})
f1, f2 = Foo(), Foo2()
print(f1.bar(), f2.bar())
```

## Metaclass

A metaclass is the class of a class. `type` is the default metaclass.

```python
class Meta(type):
    pass
class MyClass(metaclass=Meta):
    pass
print(type(MyClass))
```

## __new__ vs __init__

__new__ creates the object; __init__ initializes it.

```python
class Meta(type):
    def __new__(mcs, name, bases, attrs):
        print('Meta.__new__:', name)
        return super().__new__(mcs, name, bases, attrs)
    def __init__(cls, name, bases, attrs):
        print('Meta.__init__:', name)
class A(metaclass=Meta):
    pass
```

## Common Mistakes

- Overusing metaclasses: prefer __init_subclass__
- Forgetting super() in metaclass
