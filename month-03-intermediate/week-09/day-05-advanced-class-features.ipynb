## __slots__

Restricts instance attributes. Saves memory.

```python
class Point:
    __slots__ = ('x', 'y')
    def __init__(self, x, y):
        self.x, self.y = x, y
p = Point(3, 4)
print(p.x, p.y)
```

## __init_subclass__

Called when a subclass is created.

```python
class PluginBase:
    registry = []
    def __init_subclass__(cls, **kwargs):
        super().__init_subclass__(**kwargs)
        PluginBase.registry.append(cls)
class A(PluginBase): pass
class B(PluginBase): pass
print([c.__name__ for c in PluginBase.registry])
```

## __class_getitem__

Enables cls[key] syntax for generics.

```python
print(list[int])
class Generic:
    @classmethod
    def __class_getitem__(cls, key):
        return f'Generic[{key}]'
print(Generic[int])
```
