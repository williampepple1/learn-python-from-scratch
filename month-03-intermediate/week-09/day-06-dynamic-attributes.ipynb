## __getattr__

Called when normal attribute lookup fails.

```python
class Lazy:
    def __getattr__(self, name):
        value = f'computed_{name}'
        setattr(self, name, value)
        return value
l = Lazy()
print(l.foo)
print(l.foo)
```

## __setattr__

Called on every attribute assignment. Use object.__setattr__ to avoid recursion.

```python
class Validated:
    def __setattr__(self, name, value):
        if name == 'age' and (not isinstance(value, int) or value < 0 or value > 150):
            raise ValueError('Invalid age')
        object.__setattr__(self, name, value)
p = Validated()
p.name = 'Alice'
p.age = 30
print(p.name, p.age)
```

## Proxy

Forward attribute access to another object.

```python
class Proxy:
    def __init__(self, target):
        object.__setattr__(self, '_target', target)
    def __getattr__(self, name):
        return getattr(object.__getattribute__(self, '_target'), name)
class Data:
    def __init__(self, x): self.x = x
d = Data(10)
p = Proxy(d)
print(p.x)
```
