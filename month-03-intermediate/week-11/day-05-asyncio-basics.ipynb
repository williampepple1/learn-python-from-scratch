## async def and await

async def defines a coroutine. await yields control.

```python
import asyncio

async def hello():
    print('Hello')
    await asyncio.sleep(1)
    print('World')

asyncio.run(hello())
```

## gather

Run multiple coroutines concurrently.

```python
import asyncio
import time

async def task(n):
    await asyncio.sleep(1)
    return n * 2

async def main():
    start = time.perf_counter()
    results = await asyncio.gather(task(1), task(2), task(3))
    print('Results:', results)
    print('Time:', time.perf_counter() - start)

asyncio.run(main())
```

## create_task

Schedule a coroutine to run.

```python
import asyncio

async def background():
    await asyncio.sleep(2)
    print('Background done')

async def main():
    t = asyncio.create_task(background())
    print('Task started')
    await t
    print('Main done')

asyncio.run(main())
```

## asyncio.sleep vs time.sleep

asyncio.sleep yields; time.sleep blocks.
