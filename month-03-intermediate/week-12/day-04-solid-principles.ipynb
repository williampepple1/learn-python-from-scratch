## SRP - Single Responsibility Principle

A class should have only one reason to change. One responsibility per class.

```python
# Bad: User handles both data and persistence
# class User:
#     def save_to_db(self): ...
#     def send_email(self): ...

# Good: Separate concerns
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email

class UserRepository:
    def save(self, user):
        pass  # DB logic

class EmailService:
    def send(self, user, message):
        pass  # Email logic
```

## OCP - Open/Closed Principle

Open for extension, closed for modification. Add behavior via new classes, not by changing existing ones.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14159 * self.r ** 2

# Add Triangle without modifying Shape, Rectangle, Circle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base, self.height = base, height
    def area(self):
        return 0.5 * self.base * self.height

def total_area(shapes):
    return sum(s.area() for s in shapes)

shapes = [Rectangle(2, 3), Circle(1), Triangle(4, 5)]
print(total_area(shapes))
```

## LSP - Liskov Substitution Principle

Subtypes must be substitutable for their base types. Don't break contracts.

```python
# Bad: Square extends Rectangle but breaks set_width/set_height contract
# class Rectangle:
#     def set_width(self, w): ...
#     def set_height(self, h): ...
# class Square(Rectangle):  # set_width also changes height - violates LSP

# Good: Both implement same interface
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side ** 2

# Any Shape can be used interchangeably
def print_area(shape: Shape):
    print(shape.area())

print_area(Rectangle(2, 3))
print_area(Square(4))
```

## ISP - Interface Segregation Principle

Clients should not depend on interfaces they don't use. Prefer many specific interfaces over one general.

```python
# Bad: Fat interface
# class Worker(ABC):
#     def work(self): ...
#     def eat(self): ...  # Robot doesn't eat!

# Good: Segregated
from abc import ABC, abstractmethod

class Workable(ABC):
    @abstractmethod
    def work(self):
        pass

class Eatable(ABC):
    @abstractmethod
    def eat(self):
        pass

class Human(Workable, Eatable):
    def work(self):
        return "Working"
    def eat(self):
        return "Eating"

class Robot(Workable):
    def work(self):
        return "Working"

# Robot doesn't need Eatable
```

## DIP - Dependency Inversion Principle

Depend on abstractions, not concretions. High-level modules shouldn't depend on low-level modules.

```python
from abc import ABC, abstractmethod

class NotificationService(ABC):
    @abstractmethod
    def send(self, message):
        pass

class EmailNotification(NotificationService):
    def send(self, message):
        print(f"Email: {message}")

class SMSNotification(NotificationService):
    def send(self, message):
        print(f"SMS: {message}")

class UserService:
    def __init__(self, notifier: NotificationService):
        self.notifier = notifier  # Depends on abstraction

    def register(self, name):
        self.notifier.send(f"Welcome {name}")

# Inject dependency
service = UserService(EmailNotification())
service.register("Alice")

# Easy to swap
service2 = UserService(SMSNotification())
service2.register("Bob")
```

## Common Mistakes

- **SRP violation**: God classes doing everything
- **OCP violation**: if/elif for new types instead of polymorphism
- **LSP violation**: Subclass that throws in base method
- **ISP violation**: Implementing unused methods with pass
- **DIP violation**: new ConcreteClass() inside high-level code
