## EAFP vs LBYL

**EAFP** (Easier to Ask Forgiveness than Permission): try first, handle exceptions.
**LBYL** (Look Before You Leap): check first, then act.

Python favors EAFP.

```python
# LBYL
d = {"a": 1}
if "a" in d:
    value = d["a"]

# EAFP (Pythonic)
try:
    value = d["a"]
except KeyError:
    value = None

# Or: value = d.get("a")
```

## List Comprehensions

Concise way to build lists. Prefer over imperative loops when building a list.

```python
squares = [x ** 2 for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]
matrix = [[i * j for j in range(3)] for i in range(3)]
print(squares[:5])
print(evens[:5])
print(matrix)
```

## Dict and Set Comprehensions

Same syntax for dicts and sets.

```python
d = {k: v * 2 for k, v in [("a", 1), ("b", 2)]}
s = {x % 5 for x in range(10)}
print(d)
print(s)
```

## Unpacking

`*` for iterables, `**` for dicts.

```python
a, b, *rest = [1, 2, 3, 4, 5]
print(a, b, rest)

first, *middle, last = [1, 2, 3, 4]
print(first, middle, last)

def f(a, b, c):
    return a + b + c
args = [1, 2, 3]
print(f(*args))

kwargs = {"a": 1, "b": 2, "c": 3}
print(f(**kwargs))
```

## Walrus Operator :=

Assign and use in one expression. Python 3.8+.

```python
# Without walrus
n = len([1, 2, 3])
if n > 2:
    print(n)

# With walrus
if (n := len([1, 2, 3])) > 2:
    print(n)

# In while loop (when reading lines):
# while (line := input().strip()):
#     process(line)
```

```python
# Simpler walrus example
items = [1, 2, 3, 4, 5]
if (n := len(items)) > 3:
    print(f"List has {n} items")
```

## enumerate and zip

Iterate with index or over multiple iterables.

```python
for i, x in enumerate(["a", "b", "c"]):
    print(f"{i}: {x}")

for a, b in zip([1, 2, 3], ["x", "y", "z"]):
    print(a, b)

# zip with strict=True (3.10+) for unequal lengths
list(zip([1, 2], ["a", "b"]))
```

## Context Managers

Use `with` for resource management.

```python
with open("temp.txt", "w") as f:
    f.write("hello")
# File closed automatically

from contextlib import contextmanager

@contextmanager
def managed():
    print("enter")
    yield
    print("exit")

with managed():
    print("inside")
```

## any and all

Short-circuit logic for iterables.

```python
nums = [2, 4, 6, 8]
print(all(n % 2 == 0 for n in nums))
print(any(n > 5 for n in nums))
```

## Common Mistakes

- **Comprehension overuse**: Complex logic can be unreadable
- **Mutable default**: `def f(x=[])` - use `x=None`
- **LBYL for rare cases**: EAFP is often cleaner
- **Not using with**: Resources may leak
