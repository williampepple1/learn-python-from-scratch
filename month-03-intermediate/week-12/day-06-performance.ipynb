## Measure First

Don't optimize blindly. Profile to find bottlenecks.

## timeit

Measure execution time of small code snippets. Use for micro-benchmarks.

```python
import timeit

# timeit.repeat runs multiple times
result = timeit.timeit("sum(range(1000))", number=10000)
print(f"10000 runs: {result:.4f}s")

# Compare two approaches
t1 = timeit.timeit("[x**2 for x in range(100)]", number=10000)
t2 = timeit.timeit("list(map(lambda x: x**2, range(100)))", number=10000)
print(f"List comp: {t1:.4f}, map: {t2:.4f}")
```

## cProfile

Profile which functions take the most time. `python -m cProfile script.py`

```python
import cProfile

def slow():
    total = 0
    for i in range(10000):
        total += i
    return total

def fast():
    return sum(range(10000))

def main():
    for _ in range(100):
        slow()
        fast()

cProfile.run("main()", sort="cumulative")
```

## Optimization: Use Built-ins

Built-ins are implemented in C. Prefer `sum()`, `map()`, `filter()` over manual loops for simple operations.

```python
import timeit

# Slower
def manual_sum(n):
    total = 0
    for i in range(n):
        total += i
    return total

# Faster
def builtin_sum(n):
    return sum(range(n))

n = 100000
t1 = timeit.timeit(lambda: manual_sum(n), number=100)
t2 = timeit.timeit(lambda: builtin_sum(n), number=100)
print(f"Manual: {t1:.4f}, Built-in: {t2:.4f}")
```

## Optimization: Avoid Repeated Lookups

Cache method/attribute lookups in loops.

```python
data = list(range(1000))

# Slower: repeated len() and append lookup
# for i in range(len(data)): ...

# Faster: cache append
append = [].append  # Just for demo
result = []
for x in data:
    result.append(x * 2)

# Or use list comp (often fastest)
result2 = [x * 2 for x in data]
```

## Optimization: Local Variables

Local variable access is faster than global. Use `local_func = global_func` in hot loops.

```python
import math

def with_global(n):
    total = 0
    for i in range(n):
        total += math.sqrt(i)
    return total

def with_local(n):
    sqrt = math.sqrt
    total = 0
    for i in range(n):
        total += sqrt(i)
    return total

n = 10000
import timeit
t1 = timeit.timeit(lambda: with_global(n), number=100)
t2 = timeit.timeit(lambda: with_local(n), number=100)
print(f"Global: {t1:.4f}, Local: {t2:.4f}")
```

## Generators for Memory

Use generators when you don't need the full list in memory.

```python
# List: all in memory
def get_squares_list(n):
    return [x**2 for x in range(n)]

# Generator: one at a time
def get_squares_gen(n):
    for x in range(n):
        yield x**2

# sum(get_squares_gen(10**6))  # Low memory
# sum(get_squares_list(10**6))  # High memory
```

## Common Mistakes

- **Optimizing too early**: Measure first
- **Premature complexity**: Simple code first
- **Ignoring algorithms**: O(n²) → O(n log n) beats micro-optimizations
- **timeit with setup**: Use `setup=` for imports
