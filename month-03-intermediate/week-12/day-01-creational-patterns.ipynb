## Singleton

Ensure only one instance exists. Useful for config, DB connection, logging.

````python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

a = Singleton()
b = Singleton()
print(a is b)
````

## Factory (Simple Factory)

Create objects without specifying exact class. Encapsulate object creation.

````python
class Dog:
    def speak(self):
        return "Woof"

class Cat:
    def speak(self):
        return "Meow"

def animal_factory(animal_type):
    if animal_type == "dog":
        return Dog()
    elif animal_type == "cat":
        return Cat()
    raise ValueError(f"Unknown animal: {animal_type}")

a = animal_factory("dog")
print(a.speak())
````

## Factory Method

Subclasses override the factory method to create the appropriate object.

````python
from abc import ABC, abstractmethod

class Document(ABC):
    @abstractmethod
    def render(self):
        pass

class PDFDocument(Document):
    def render(self):
        return "Rendering PDF"

class HTMLDocument(Document):
    def render(self):
        return "Rendering HTML"

class Application(ABC):
    @abstractmethod
    def create_document(self) -> Document:
        pass

    def open_document(self):
        doc = self.create_document()
        return doc.render()

class PDFApp(Application):
    def create_document(self) -> Document:
        return PDFDocument()

class HTMLApp(Application):
    def create_document(self) -> Document:
        return HTMLDocument()

app = PDFApp()
print(app.open_document())
````

## Builder

Construct complex objects step by step. Fluent interface.

````python
class QueryBuilder:
    def __init__(self):
        self._select = "*"
        self._table = ""
        self._where = []

    def select(self, *columns):
        self._select = ", ".join(columns) if columns else "*"
        return self

    def from_table(self, table):
        self._table = table
        return self

    def where(self, condition):
        self._where.append(condition)
        return self

    def build(self):
        sql = f"SELECT {self._select} FROM {self._table}"
        if self._where:
            sql += " WHERE " + " AND ".join(self._where)
        return sql + ";"

q = QueryBuilder().select("id", "name").from_table("users").where("id > 0").build()
print(q)
````

## Prototype

Clone objects instead of creating from scratch. Use `copy.copy` or `copy.deepcopy`.

````python
import copy

class Config:
    def __init__(self, settings):
        self.settings = settings

    def clone(self):
        return copy.deepcopy(self)

original = Config({"theme": "dark", "font": 12})
clone = original.clone()
clone.settings["theme"] = "light"
print(original.settings["theme"])
print(clone.settings["theme"])
````

## Common Mistakes

- **Singleton and testing**: Hard to mock; consider dependency injection
- **Overusing Singleton**: Often a module-level instance is enough
- **Builder without return self**: Breaks fluent interface
- **Shallow copy when deep needed**: Nested objects shared
