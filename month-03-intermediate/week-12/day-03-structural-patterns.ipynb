## Adapter

Convert interface to another clients expect.

```python
class NewAPI:
    def fetch(self): return 42

class Adapter:
    def __init__(self, adaptee): self.adaptee = adaptee
    def get_data(self): return {'value': self.adaptee.fetch()}

adapter = Adapter(NewAPI())
print(adapter.get_data())
```

## Decorator

Add behavior by wrapping. Same interface.

```python
class Component:
    def operation(self): return 'Component'

class Decorator:
    def __init__(self, component): self._component = component
    def operation(self): return 'Decorator(' + self._component.operation() + ')'

c = Component()
d = Decorator(c)
print(d.operation())
```

## Proxy

Control access, lazy loading, caching.

```python
class RealSubject:
    def request(self): return 'Real data'

class Proxy:
    def __init__(self): self._real = None
    def request(self):
        if self._real is None:
            self._real = RealSubject()
        return self._real.request()

p = Proxy()
print(p.request())
print(p.request())
```
