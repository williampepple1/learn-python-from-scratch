### Exercise 455: Descriptor

Create a descriptor class that stores and returns values.


```python
class Desc:
    def __get__(self,obj,owner): return getattr(obj,'_v',None)
    def __set__(self,obj,v): setattr(obj,'_v',v)
```

```python
# Test
class C: x=Desc()
c=C(); c.x=5; assert c.x==5
print("✅ Exercise 455 passed!")
```

### Exercise 459: Algorithm

Implement binary search on a sorted list.


```python
def bsearch(lst, x):
    lo,hi=0,len(lst)-1
    while lo<=hi:
        m=(lo+hi)//2
        if lst[m]==x: return m
        if lst[m]<x: lo=m+1
        else: hi=m-1
    return -1
```

```python
# Test
assert bsearch([1,2,3,4,5], 3)==2
print("✅ Exercise 459 passed!")
```

### Exercise 460: System Design

Design a simple Cache class with get/set and max_size.


```python
class Cache:
    def __init__(self, max_size=10): self.d={}; self.max=max_size; self.order=[]
    def get(self,k): return self.d.get(k)
    def set(self,k,v):
        if k in self.d: self.order.remove(k)
        elif len(self.d)>=self.max: del self.d[self.order.pop(0)]
        self.d[k]=v; self.order.append(k)
```

```python
# Test
c=Cache(2); c.set('a',1); c.set('b',2); c.set('c',3); assert c.get('a') is None
print("✅ Exercise 460 passed!")
```
