### Exercise 351: Basic Decorator

Write a decorator @uppercase that converts the return value to uppercase.


```python
def uppercase(fn):
    def wrapper(*a,**k): return fn(*a,**k).upper()
    return wrapper
```

```python
# Test
@uppercase
def greet(): return 'hi'
assert greet()=='HI'
print("✅ Exercise 351 passed!")
```

### Exercise 352: Timing Decorator

Write @timing that prints elapsed time and returns the result.


```python
import time
def timing(fn):
    def w(*a,**k):
        t=time.perf_counter(); r=fn(*a,**k); print(time.perf_counter()-t); return r
    return w
```

```python
# Test
assert timing(lambda: 1)()==1
print("✅ Exercise 352 passed!")
```

### Exercise 353: Caching Decorator

Write @cache that memoizes results for same args.


```python
def cache(fn):
    d={}
    def w(*a):
        if a not in d: d[a]=fn(*a)
        return d[a]
    return w
```

```python
# Test
@cache
def fib(n): return n if n<2 else fib(n-1)+fib(n-2)
assert fib(10)==55
print("✅ Exercise 353 passed!")
```

### Exercise 354: Validation Decorator

Write @positive_args that raises ValueError if any arg<=0.


```python
def positive_args(fn):
    def w(*a):
        if any(x<=0 for x in a): raise ValueError('args must be positive')
        return fn(*a)
    return w
```

```python
# Test
@positive_args
def mul(a,b): return a*b
assert mul(2,3)==6
print("✅ Exercise 354 passed!")
```

### Exercise 355: Decorator with Args

Write @repeat(n) that calls the function n times and returns last result.


```python
def repeat(n):
    def dec(fn):
        def w(*a,**k):
            r=None
            for _ in range(n): r=fn(*a,**k)
            return r
        return w
    return dec
```

```python
# Test
@repeat(3)
def r(): return 1
assert r()==1
print("✅ Exercise 355 passed!")
```

### Exercise 357: functools.reduce

Use reduce to compute product of a list.


```python
from functools import reduce
def product(nums): return reduce(lambda a,b:a*b, nums, 1)
```

```python
# Test
assert product([1,2,3,4])==24
print("✅ Exercise 357 passed!")
```

### Exercise 358: functools.lru_cache

Use lru_cache to memoize a recursive fibonacci.


```python
from functools import lru_cache
@lru_cache(maxsize=None)
def fib(n): return n if n<2 else fib(n-1)+fib(n-2)
```

```python
# Test
assert fib(20)==6765
print("✅ Exercise 358 passed!")
```

### Exercise 359: Class Decorator

Write a class decorator @CountCalls that adds .call_count attribute.


```python
def CountCalls(cls):
    cls.call_count=0
    orig=cls.__init__
    def init(self,*a,**k): orig(self,*a,**k); cls.call_count+=1
    cls.__init__=init
    return cls
```

```python
# Test
@CountCalls
class C: pass
c=C(); c2=C(); assert C.call_count==2
print("✅ Exercise 359 passed!")
```

### Exercise 360: Wraps Preservation

Write @log_calls that preserves __name__ using functools.wraps.


```python
from functools import wraps
def log_calls(fn):
    @wraps(fn)
    def w(*a,**k): return fn(*a,**k)
    return w
```

```python
# Test
@log_calls
def f(): pass
assert f.__name__=='f'
print("✅ Exercise 360 passed!")
```
