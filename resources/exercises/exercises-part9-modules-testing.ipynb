### Exercise 401: Create Module

Create mymath.py with def add(a,b): return a+b. Write the file in your code.


```python
with open('mymath.py','w') as f:
    f.write('def add(a,b): return a+b')
```

```python
# Test
import importlib.util; spec=importlib.util.spec_from_file_location('mymath','mymath.py'); m=importlib.util.module_from_spec(spec); spec.loader.exec_module(m); assert m.add(1,2)==3
print("✅ Exercise 401 passed!")
```

### Exercise 402: Import Function

Import add from mymath and call it. (Run E401 first to create mymath.py)


```python
from mymath import add
```

```python
# Test
import sys
if 'mymath' in sys.modules: del sys.modules['mymath']
open('mymath.py','w').write('def add(a,b): return a+b')
from mymath import add
assert add(1,2)==3
print("✅ Exercise 402 passed!")
```

### Exercise 403: Package Init

Create a package dir with __init__.py defining VERSION=1.


```python
import os
os.makedirs('_ex403_pkg', exist_ok=True)
with open('_ex403_pkg/__init__.py','w') as f: f.write('VERSION=1')
```

```python
# Test
import importlib.util; spec=importlib.util.spec_from_file_location('_ex403_pkg','_ex403_pkg/__init__.py'); m=importlib.util.module_from_spec(spec); spec.loader.exec_module(m); assert m.VERSION==1
print("✅ Exercise 403 passed!")
```

### Exercise 405: unittest TestCase

Create a TestCase with test_add method.


```python
import unittest
class TestAdd(unittest.TestCase):
    def test_add(self): self.assertEqual(1+2, 3)
```

```python
# Test
suite=unittest.TestLoader().loadTestsFromTestCase(TestAdd); unittest.TextTestRunner().run(suite)
print("✅ Exercise 405 passed!")
```

### Exercise 406: Mock Return Value

Use unittest.mock.Mock to mock a function returning 42.


```python
from unittest.mock import Mock
m=Mock(return_value=42)
```

```python
# Test
assert m()==42
print("✅ Exercise 406 passed!")
```

### Exercise 407: Patch Function

Use patch to replace open with a mock.


```python
from unittest.mock import patch, MagicMock
```

```python
# Test
with patch('builtins.open', MagicMock()) as m: open('x'); assert m.called
print("✅ Exercise 407 passed!")
```

### Exercise 408: Parametrized Test

Use subtests or loop to test multiple inputs.


```python
def sq(x): return x**2
```

```python
# Test
for i,o in [(0,0),(1,1),(2,4)]: assert sq(i)==o
print("✅ Exercise 408 passed!")
```

### Exercise 409: Fixture Setup

Use setUp to create test data before each test.


```python
import unittest
class T(unittest.TestCase):
    def setUp(self): self.data=[1,2,3]
    def test_len(self): self.assertEqual(len(self.data),3)
```

```python
# Test
t=T(); t.setUp(); t.test_len()
print("✅ Exercise 409 passed!")
```
