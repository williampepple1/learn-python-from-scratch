# File I/O

Welcome! Python provides powerful tools for reading and writing files. This guide covers `open()`, read/write modes, context managers, pathlib, CSV, JSON, and binary files.

## open() and Basic Modes

`open(path, mode)` returns a file object. Modes: `r` (read), `w` (write, truncate), `a` (append), `x` (exclusive create), `b` (binary), `+` (read+write).

```python
with open("example.txt", "w") as f:
    f.write("Hello, World!\n")
    f.write("Line 2\n")

with open("example.txt", "r") as f:
    content = f.read()
    print(content)

with open("example.txt", "r") as f:
    lines = f.readlines()
    print(lines)
```

## Reading Files

`read()`, `readline()`, `readlines()`. Files are iterable line-by-line.

```python
with open("example.txt", "r") as f:
    for line in f:
        print(line.rstrip())

with open("example.txt", "r") as f:
    first = f.readline()
    rest = f.read()
```

## Writing Files

`write()` and `writelines()`. Use `\n` for newlines. Flush or close to ensure data is written.

```python
with open("output.txt", "w") as f:
    f.write("Line 1\n")
    f.writelines(["Line 2\n", "Line 3\n"])

with open("output.txt", "a") as f:
    f.write("Appended\n")
```

## Context Managers (with)

`with` ensures the file is closed even if an exception occurs. Always use it for files.

```python
with open("example.txt") as f:
    data = f.read()
# f is closed here automatically

try:
    f = open("example.txt")
    data = f.read()
finally:
    f.close()  # Manual cleanup - with is simpler
```

## pathlib

`pathlib.Path` provides object-oriented path handling. Cross-platform and convenient.

```python
from pathlib import Path

p = Path("resources/guides")
print(p.exists())
print(p.is_dir())
print(p / "01-variables-and-types.ipynb")

file_path = Path("example.txt")
print(file_path.read_text())
file_path.write_text("New content\n")

for f in Path(".").glob("*.ipynb"):
    print(f.name)
```

## CSV Files

Use the `csv` module for reading and writing CSV.

```python
import csv

with open("data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["name", "age", "city"])
    writer.writerow(["Alice", 30, "NYC"])
    writer.writerow(["Bob", 25, "LA"])

with open("data.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
```

## JSON Files

Use `json` module for JSON serialization.

```python
import json

data = {"name": "Alice", "age": 30, "scores": [90, 85, 88]}

with open("config.json", "w") as f:
    json.dump(data, f, indent=2)

with open("config.json", "r") as f:
    loaded = json.load(f)
    print(loaded)

s = json.dumps(data)
parsed = json.loads(s)
```

## Binary Files

Use `rb` or `wb` mode for binary data. Use `struct` for packed binary formats.

```python
import struct

with open("binary.dat", "wb") as f:
    f.write(struct.pack("i", 42))
    f.write(struct.pack("f", 3.14))

with open("binary.dat", "rb") as f:
    n = struct.unpack("i", f.read(4))[0]
    x = struct.unpack("f", f.read(4))[0]
    print(n, x)
```

## Encoding

Specify `encoding` for text files. Default is platform-dependent.

```python
with open("utf8.txt", "w", encoding="utf-8") as f:
    f.write("café résumé")

with open("utf8.txt", "r", encoding="utf-8") as f:
    print(f.read())

with open("utf8.txt", "r", encoding="utf-8", errors="replace") as f:
    pass  # errors='replace' handles bad bytes
```
