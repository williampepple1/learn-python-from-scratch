# Modules and Packages

Welcome! Python's import system lets you organize code into reusable modules and packages. This guide covers the import system, packages, `__init__.py`, pip, virtualenv, and a tour of the standard library.

## What is a Module?

A module is a `.py` file. It can define functions, classes, and variables.

```python
# In mymodule.py you might have:
# def greet(name):
#     return f"Hello, {name}!"
# PI = 3.14159

# Then: import mymodule
# mymodule.greet("Alice")
# print(mymodule.PI)
```

## import Statement

`import module` loads the module. `from module import name` imports specific names.

```python
import math
print(math.sqrt(16))
print(math.pi)

from math import sqrt, pi
print(sqrt(16), pi)

from math import *  # Avoid: pollutes namespace
print(sin(0))
```

## Standard Library Tour: os and pathlib

`os` and `pathlib` for filesystem operations.

```python
import os
from pathlib import Path

print(os.getcwd())
print(os.listdir("."))
print(Path(".").resolve())
```

## Standard Library Tour: collections

`collections`: defaultdict, Counter, deque, namedtuple.

```python
from collections import Counter, defaultdict, deque

c = Counter("hello world")
print(c.most_common(3))

d = defaultdict(list)
d["a"].append(1)
print(dict(d))

q = deque([1, 2, 3])
q.appendleft(0)
q.pop()
print(list(q))
```

## Standard Library Tour: datetime and json

`datetime` for dates/times. `json` for JSON.

```python
from datetime import datetime, timedelta
import json

now = datetime.now()
print(now.strftime("%Y-%m-%d %H:%M"))
tomorrow = now + timedelta(days=1)

data = {"name": "Alice", "age": 30}
s = json.dumps(data)
parsed = json.loads(s)
print(s, parsed)
```

## __name__ == "__main__"

Code under `if __name__ == "__main__":` runs only when the file is executed directly, not when imported.

```python
def main():
    print("Running as script")

if __name__ == "__main__":
    main()

print(__name__)
```
