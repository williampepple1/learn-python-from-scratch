# Control Flow

Welcome! Control flow determines the order in which statements execute. This guide covers conditionals (if/elif/else), loops (for, while), break and continue, ternary expressions, match/case (Python 3.10+), and truthiness.

## if / elif / else

Conditionals branch execution based on boolean expressions. Use `elif` for multiple conditions.

```python
age = 20
if age < 13:
    print("child")
elif age < 20:
    print("teen")
else:
    print("adult")
```

## Truthiness

Python uses "truthiness": values are either truthy or falsy in boolean context. Falsy: `False`, `None`, `0`, `""`, `[]`, `{}`, `()`.

```python
if []:
    print("list has items")
else:
    print("empty list is falsy")

if "hello":
    print("non-empty string is truthy")

name = ""
if not name:
    print("name is empty or falsy")
```

## for Loops

`for` iterates over iterables: sequences, dicts, files, etc.

```python
for i in [1, 2, 3]:
    print(i)

for char in "hello":
    print(char)

for key, value in {"a": 1, "b": 2}.items():
    print(key, value)
```

## range()

`range()` produces a sequence of integers. Use it for counted loops.

```python
for i in range(5):
    print(i, end=" ")
print()

for i in range(2, 8):
    print(i, end=" ")
print()

for i in range(0, 10, 2):
    print(i, end=" ")
print()

for i in range(5, 0, -1):
    print(i, end=" ")
print()
```

## enumerate() and zip()

`enumerate()` adds an index; `zip()` pairs elements from multiple iterables.

```python
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

names = ["Alice", "Bob", "Carol"]
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f"{name} is {age}")
```

## while Loops

`while` repeats as long as the condition is true.

```python
n = 0
while n < 5:
    print(n)
    n += 1

# Common pattern: read until empty
data = [1, 2, 3]
while data:
    item = data.pop()
    print(item)
```

## break and continue

`break` exits the loop immediately. `continue` skips to the next iteration.

```python
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")
print(" (break)")

for i in range(6):
    if i % 2 == 0:
        continue
    print(i, end=" ")
print(" (continue)")
```

## else on Loops

A loop's `else` block runs when the loop completes normally (not via `break`).

```python
for i in range(3):
    print(i)
else:
    print("Loop completed normally")

for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("This won't run due to break")
```

## Ternary (Conditional) Expression

`value_if_true if condition else value_if_false`

```python
age = 18
status = "adult" if age >= 18 else "minor"
print(status)

x, y = 10, 20
max_val = x if x > y else y
print(max_val)
```

## match / case (Python 3.10+)

Structural pattern matching for cleaner multi-way branching.

```python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500 | 502 | 503:
            return "Server Error"
        case _:
            return "Unknown"

print(http_status(200))
print(http_status(503))
print(http_status(999))

command = ["echo", "hello"]
match command:
    case ["quit"]:
        print("Quitting")
    case ["echo", *args]:
        print(" ".join(args))
    case _:
        print("Unknown command")
```
