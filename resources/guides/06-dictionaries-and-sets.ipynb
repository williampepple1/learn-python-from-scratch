# Dictionaries and Sets

Welcome! Dictionaries map keys to values; sets store unique, unordered elements. Both use hash tables for fast lookups. This guide covers dict operations, methods, comprehensions, sets, frozenset, and set operations.

## Creating Dictionaries

Use `{}` or `dict()`. Keys must be hashable (immutable).

```python
d = {"a": 1, "b": 2, "c": 3}
d2 = dict(a=1, b=2, c=3)
d3 = dict([("x", 1), ("y", 2)])
print(d, d2, d3)
```

## Accessing and Modifying Dicts

Use `[]` for access. `d[key]` raises KeyError if missing; use `get()` for safe access.

```python
d = {"name": "Alice", "age": 30}
print(d["name"])
print(d.get("age"))
print(d.get("city", "Unknown"))

d["city"] = "NYC"
d["age"] = 31
del d["city"]
print(d)
```

## Dict Methods

`keys()`, `values()`, `items()`, `pop()`, `popitem()`, `update()`, `setdefault()`.

```python
d = {"a": 1, "b": 2, "c": 3}
print(list(d.keys()))
print(list(d.values()))
print(list(d.items()))

val = d.pop("b")
print(d, val)

d.update({"b": 2, "d": 4})
print(d)

d.setdefault("e", 5)
d.setdefault("a", 99)
print(d)
```

## Dict Comprehensions

Build dicts from iterables: `{k: v for item in iterable}`.

```python
squares = {x: x**2 for x in range(5)}
print(squares)

words = ["hello", "world"]
lengths = {w: len(w) for w in words}
print(lengths)

filtered = {k: v for k, v in {"a": 1, "b": 0, "c": 3}.items() if v > 0}
print(filtered)
```

## Creating Sets

Sets store unique elements. Use `{}` or `set()`. Empty set must be `set()`, not `{}`.

```python
s = {1, 2, 3, 2, 1}
print(s)

s2 = set([1, 2, 3, 2, 1])
print(s2)

empty = set()
print(type(empty))
print(type({}))
```

## Set Methods

`add`, `remove`, `discard`, `pop`, `clear`, `update`.

```python
s = {1, 2, 3}
s.add(4)
s.add(2)
print(s)

s.remove(2)
s.discard(99)
print(s)

s.update([5, 6])
print(s)
```

## Set Operations

Union (`|`), intersection (`&`), difference (`-`), symmetric difference (`^`).

```python
a = {1, 2, 3}
b = {2, 3, 4}
print(a | b)
print(a & b)
print(a - b)
print(a ^ b)

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

## Set Comprehensions

Build sets from iterables: `{expr for item in iterable}`.

```python
squares = {x**2 for x in range(5)}
print(squares)

chars = {c.lower() for c in "Hello"}
print(chars)
```

## frozenset

`frozenset` is an immutable set. Can be used as dict keys or set elements.

```python
fs = frozenset([1, 2, 3])
print(fs)
d = {fs: "value"}
print(d[fs])
s = {fs, frozenset([4, 5])}
print(s)
```

## Membership and Iteration

Both dicts and sets support `in` and iteration. Dicts iterate over keys by default.

```python
d = {"a": 1, "b": 2}
print("a" in d)
for k in d:
    print(k, d[k])

s = {1, 2, 3}
print(2 in s)
for x in s:
    print(x)
```
