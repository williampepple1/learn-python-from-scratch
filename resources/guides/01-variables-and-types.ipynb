# Variables and Types

Welcome! This guide covers Python's variable system and built-in types. Python uses dynamic typing—variables can hold any type and can change type during execution. Understanding types, conversion, and mutability is essential for writing robust Python code.

## Variables and Assignment

Variables are created by assignment. No declaration is needed. Names follow the same rules as identifiers.

```python
x = 42
name = "Alice"
pi = 3.14159
is_active = True
print(x, name, pi, is_active)
```

## Dynamic Typing

Python is dynamically typed. A variable can be reassigned to a different type.

```python
value = 10
print(type(value), value)

value = "ten"
print(type(value), value)

value = [1, 2, 3]
print(type(value), value)
```

## Integer (int)

Integers have arbitrary precision in Python—no overflow for whole numbers.

```python
a = 42
b = -17
c = 1_000_000  # Underscores for readability
big = 10 ** 100
print(a, b, c, big)
```

## Float (float)

Floating-point numbers use IEEE 754 double precision. Be aware of precision limits.

```python
x = 3.14
y = 2e-3  # Scientific notation: 0.002
z = float("inf")  # Infinity
print(x, y, z)
print(0.1 + 0.2)  # Floating-point precision quirk
```

## String (str)

Strings are immutable sequences of Unicode characters.

```python
s1 = 'single quotes'
s2 = "double quotes"
s3 = '''multi
line'''
print(s1, s2)
print(s3)
```

## Boolean (bool)

Booleans are `True` and `False` (capitalized). They subclass `int`.

```python
flag = True
print(flag, type(flag))
print(True + 1)  # 2, because True == 1
```

## None

`None` is the sole instance of `NoneType`. It represents the absence of a value.

```python
result = None
print(result, type(result))

def no_return():
    pass

print(no_return())  # None
```

## Type Conversion

Use built-in constructors to convert between types: `int()`, `float()`, `str()`, `bool()`.

```python
print(int("42"))
print(float("3.14"))
print(str(42))
print(bool(1), bool(0), bool(""), bool("hello"))
```

## Type Checking

Use `type()` for exact type or `isinstance()` for type hierarchy checks.

```python
x = 42
print(type(x) == int)
print(isinstance(x, int))
print(isinstance(True, int))  # bool subclasses int
```

## The id() Function

`id()` returns the memory address (identity) of an object. Two variables pointing to the same object have the same id.

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(id(a), id(b), id(c))
print(a is b, a is c)
```

## Mutability

Mutable objects can be changed in place; immutable objects cannot. Lists and dicts are mutable; int, float, str, tuple, frozenset are immutable.

```python
lst = [1, 2, 3]
lst.append(4)  # In-place modification
print(lst)

s = "hello"
# s[0] = 'H'  # TypeError: str is immutable
s = s.upper()  # Returns new string
print(s)
```

## Summary

You've learned about variables, dynamic typing, built-in types (int, float, str, bool, None), type conversion, type checking, `id()`, and mutability. These fundamentals underpin all Python programming.
