# Strings

Welcome! Strings are immutable sequences of Unicode characters. This guide covers string creation, common methods, f-strings and formatting, slicing, encoding, and regex basics.

## String Creation

Strings can use single, double, or triple quotes. Triple quotes allow multi-line strings.

```python
s1 = 'single'
s2 = "double"
s3 = '''multi
line'''
s4 = r"raw\\n"  # Raw: backslashes literal
print(s1, s2)
print(s3)
print(s4)
```

## Indexing and Slicing

Strings support zero-based indexing and slicing with `[start:stop:step]`.

```python
s = "Python"
print(s[0], s[-1])
print(s[1:4])
print(s[:3], s[3:])
print(s[::-1])  # Reverse
```

## Common String Methods

Strings have many useful methods. They return new strings (strings are immutable).

```python
s = "  Hello, World!  "
print(s.strip())
print(s.lower(), s.upper())
print(s.replace("World", "Python"))
print("one,two,three".split(","))
print(" ".join(["a", "b", "c"]))
print("hello".startswith("he"), "world".endswith("ld"))
print("42".isdigit(), "abc".isalpha())
```

## f-Strings (Formatted String Literals)

f-strings embed expressions inside `{}`. Available in Python 3.6+.

```python
name = "Alice"
age = 30
print(f"{name} is {age} years old")
print(f"{age=}")
print(f"{3.14159:.2f}")
print(f"{42:05d}")
print(f"{2 + 3}")
```

## str.format()

The `format()` method provides positional and named placeholders.

```python
print("{0} {1}".format("Hello", "World"))
print("{name} is {age}".format(name="Bob", age=25))
print("{:.2f}".format(3.14159))
print("{:>10}".format("right"))
print("{:<10}".format("left"))
```

## % Formatting (Legacy)

The `%` operator is the oldest formatting style. Still used in some codebases.

```python
print("%s %d" % ("Score:", 100))
print("%.2f" % 3.14159)
print("%10s" % "right")
```

## String Membership and Searching

Use `in` for membership. `find()` and `index()` locate substrings.

```python
s = "hello world"
print("world" in s)
print(s.find("world"))
print(s.find("xyz"))  # -1 if not found
print(s.index("world"))
print(s.count("l"))
```

## String Encoding

Python 3 strings are Unicode. Use `encode()` and `decode()` for bytes.

```python
s = "café"
b = s.encode("utf-8")
print(b)
print(b.decode("utf-8"))
print("hello".encode("ascii"))
```

## Regular Expressions Basics

The `re` module provides regex support. Common functions: `search`, `match`, `findall`, `sub`.

```python
import re

text = "Contact: alice@example.com or bob@test.org"
match = re.search(r"[\w.-]+@[\w.-]+\.\w+", text)
print(match.group() if match else None)

emails = re.findall(r"[\w.-]+@[\w.-]+\.\w+", text)
print(emails)

cleaned = re.sub(r"\d+", "#", "a1b2c3")
print(cleaned)

print(re.match(r"^Hello", "Hello world"))
```

## Useful String Constants

The `string` module provides useful constants.

```python
import string

print(string.ascii_lowercase)
print(string.ascii_uppercase)
print(string.digits)
print(string.punctuation)
```
