# Functions

Welcome! Functions are the primary way to organize and reuse code in Python. This guide covers defining functions, parameters, return values, advanced argument handling, lambdas, docstrings, scope, and closures.

## Defining Functions with def

Use the `def` keyword to define a function. The function body is indented.

```python
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
```

## Parameters and Arguments

Parameters are the names in the function definition. Arguments are the values passed when calling.

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)

result = add(b=5, a=3)  # Keyword arguments
print(result)
```

## Return Values

Use `return` to send a value back. Without `return`, the function returns `None`.

```python
def square(x):
    return x * x

print(square(4))

def multi_return():
    return 1, 2, 3  # Returns a tuple

a, b, c = multi_return()
print(a, b, c)
```

## Default Arguments

Parameters can have default values. Defaults are evaluated once at definition time.

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Alice"))
print(greet("Bob", "Hi"))

# Mutable default gotcha
def append_to(element, target=[]):  # Avoid! Use target=None
    target.append(element)
    return target

def append_to_safe(element, target=None):
    if target is None:
        target = []
    target.append(element)
    return target
```

## *args: Variable Positional Arguments

`*args` collects extra positional arguments into a tuple.

```python
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4, 5))
print(sum_all())

def log(prefix, *messages):
    for msg in messages:
        print(f"[{prefix}] {msg}")

log("INFO", "Started", "Running", "Done")
```

## **kwargs: Variable Keyword Arguments

`**kwargs` collects extra keyword arguments into a dictionary.

```python
def print_config(**kwargs):
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print_config(host="localhost", port=8080, debug=True)

def wrapper(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

wrapper(1, 2, x=3, y=4)
```

## Lambda Functions

Lambdas are anonymous, single-expression functions. Use them for short callbacks.

```python
square = lambda x: x * x
print(square(5))

pairs = [(1, 'one'), (3, 'three'), (2, 'two')]
pairs.sort(key=lambda p: p[0])
print(pairs)

add = lambda a, b: a + b
print(add(10, 20))
```

## Docstrings

Docstrings document functions. They're stored in `__doc__` and used by `help()`.

```python
def factorial(n):
    """Compute n! (n factorial).

    Args:
        n: A non-negative integer.

    Returns:
        The factorial of n.
    """
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(factorial.__doc__)
help(factorial)
```

## Scope: LEGB Rule

Python looks up names in order: Local, Enclosing, Global, Built-in.

```python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(x)
    inner()
    print(x)

outer()
print(x)
```

## global and nonlocal

Use `global` to modify a global variable. Use `nonlocal` to modify a variable in an enclosing scope.

```python
counter = 0

def increment():
    global counter
    counter += 1

increment()
increment()
print(counter)

def outer():
    count = 0
    def inner():
        nonlocal count
        count += 1
        return count
    return inner

inc = outer()
print(inc(), inc(), inc())
```

## Closures

A closure is a function that captures variables from its enclosing scope.

```python
def make_multiplier(factor):
    def multiply(x):
        return x * factor  # factor is captured
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)
print(double(5), triple(5))

def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

c = make_counter()
print(c(), c(), c())
```

## Summary

You've learned about `def`, parameters, `return`, default args, `*args`, `**kwargs`, lambdas, docstrings, scope (LEGB), and closures. Functions are first-class objects in Python—you can pass them around and return them.
