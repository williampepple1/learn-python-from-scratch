# Testing

Welcome! Testing ensures your code works correctly. This guide covers unittest, pytest, mocking, TDD, coverage, type hints, and mypy.

## Why Test?

Tests catch bugs, document behavior, and enable confident refactoring. Write tests for critical paths and edge cases.

```python
def add(a, b):
    return a + b

assert add(2, 3) == 5
assert add(-1, 1) == 0
print("Simple asserts passed")
```

## unittest

The built-in `unittest` module provides TestCase, assertions, and test discovery.

```python
import unittest

def add(a, b):
    return a + b

class TestAdd(unittest.TestCase):
    def test_add_positive(self):
        self.assertEqual(add(2, 3), 5)

    def test_add_negative(self):
        self.assertEqual(add(-1, -1), -2)

    def test_add_zero(self):
        self.assertEqual(add(0, 0), 0)

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
```

## unittest Assertions

`assertEqual`, `assertTrue`, `assertRaises`, `assertIn`, `assertAlmostEqual`, etc.

```python
import unittest

class TestAssertions(unittest.TestCase):
    def test_basic(self):
        self.assertEqual(1 + 1, 2)
        self.assertTrue(1 < 2)
        self.assertIn(1, [1, 2, 3])
        self.assertAlmostEqual(0.1 + 0.2, 0.3, places=5)

    def test_exception(self):
        with self.assertRaises(ValueError):
            int("not a number")

unittest.main(argv=[''], exit=False, verbosity=0)
```

## pytest

pytest is simpler: no class inheritance, plain assert, and better output. Install with `pip install pytest`.

```python
def add(a, b):
    return a + b

def test_add():
    assert add(2, 3) == 5

def test_add_negative():
    assert add(-1, -1) == -2

def test_add_type_error():
    import pytest
    with pytest.raises(TypeError):
        add("a", "b")

# Run with: pytest test_file.py -v
```

## pytest Fixtures

Fixtures provide setup/teardown. Use `@pytest.fixture` decorator.

```python
import pytest

@pytest.fixture
def sample_list():
    return [1, 2, 3]

def test_len(sample_list):
    assert len(sample_list) == 3

def test_sum(sample_list):
    assert sum(sample_list) == 6
```

## Mocking

`unittest.mock` lets you replace dependencies with fake objects for isolated testing.

```python
from unittest.mock import Mock, patch, MagicMock

def fetch_user(user_id):
    # Would call an API
    return {"id": user_id, "name": "Alice"}

def get_user_name(user_id):
    user = fetch_user(user_id)
    return user["name"]

with patch("__main__.fetch_user"):
    fetch_user = Mock(return_value={"id": 1, "name": "Bob"})
    result = get_user_name(1)
    # Need to patch where it's used

mock = Mock()
mock.method.return_value = 42
print(mock.method())
```

## TDD (Test-Driven Development)

TDD: Write tests first, then implement. Red → Green → Refactor.

```python
# 1. Write failing test
def test_factorial():
    assert factorial(0) == 1
    assert factorial(5) == 120

# 2. Implement minimal code to pass
def factorial(n):
    if n < 2:
        return 1
    return n * factorial(n - 1)

# 3. Run test
test_factorial()
print("TDD cycle complete")
```

## Type Hints

Type hints document expected types. Use for function parameters and return values.

```python
def greet(name: str) -> str:
    return f"Hello, {name}!"

def add(a: int, b: int) -> int:
    return a + b

from typing import List, Optional, Dict

def process(items: List[str]) -> Optional[Dict[str, int]]:
    if not items:
        return None
    return {item: len(item) for item in items}

print(greet("Alice"))
print(process(["a", "bb"]))
```
