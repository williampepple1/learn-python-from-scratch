# Classes and OOP

Welcome! Object-oriented programming in Python uses classes to bundle data and behavior. This guide covers class definition, `__init__`, `self`, methods, properties, magic methods, and encapsulation.

## Defining a Class

Use the `class` keyword. Class names are typically PascalCase.

```python
class Dog:
    pass

dog = Dog()
print(dog)
print(type(dog))
```

## __init__ and self

`__init__` is the constructor. `self` refers to the instance being created.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog = Dog("Rex", 3)
print(dog.name, dog.age)
```

## Instance Methods

Methods receive `self` as the first parameter. They operate on instance data.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name} says woof!"

    def human_age(self):
        return self.age * 7

dog = Dog("Rex", 3)
print(dog.bark())
print(dog.human_age())
```

## Class and Instance Attributes

Class attributes are shared; instance attributes are per-object.

```python
class Dog:
    species = "Canis familiaris"

    def __init__(self, name):
        self.name = name

d1 = Dog("Rex")
d2 = Dog("Max")
print(d1.species, d2.species)
print(Dog.species)
```

## Properties (@property)

Properties provide controlled access to attributes. Use getters and setters without explicit method calls.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Radius must be non-negative")
        self._radius = value

    @property
    def area(self):
        return 3.14159 * self._radius ** 2

c = Circle(5)
print(c.radius, c.area)
c.radius = 10
print(c.area)
```

## Magic Methods (Dunder Methods)

Magic methods customize behavior. Common ones: `__str__`, `__repr__`, `__len__`, `__eq__`.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point({self.x}, {self.y})"

    def __repr__(self):
        return f"Point({self.x!r}, {self.y!r})"

    def __eq__(self, other):
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y

p = Point(3, 4)
print(str(p))
print(repr(p))
print(Point(3, 4) == Point(3, 4))
```

## Encapsulation: Private and Protected

Convention: single underscore `_` for protected; double `__` for name mangling ("private").

```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance  # Protected by convention

    def deposit(self, amount):
        self._balance += amount

    def get_balance(self):
        return self._balance

class Secret:
    def __init__(self):
        self.__internal = 42  # Name mangled to _Secret__internal

s = Secret()
# print(s.__internal)  # AttributeError
print(s._Secret__internal)
```

## Class Methods and Static Methods

`@classmethod` receives the class; `@staticmethod` receives neither class nor instance.

```python
class MyClass:
    count = 0

    @classmethod
    def from_string(cls, s):
        return cls(int(s))

    @classmethod
    def get_count(cls):
        return cls.count

    @staticmethod
    def utility(x, y):
        return x + y

print(MyClass.utility(1, 2))
```

## __slots__

`__slots__` restricts instance attributes and saves memory.

```python
class Point:
    __slots__ = ("x", "y")

    def __init__(self, x, y):
        self.x = x
        self.y = y

p = Point(1, 2)
# p.z = 3  # AttributeError
```
