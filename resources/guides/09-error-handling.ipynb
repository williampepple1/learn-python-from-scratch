# Error Handling

Welcome! Python uses exceptions for error handling. This guide covers try/except/finally/else, custom exceptions, the exception hierarchy, and best practices.

## try and except

Catch exceptions with `try`/`except`. Only catch what you can handle.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")

try:
    value = int("not a number")
except ValueError as e:
    print(f"Invalid value: {e}")
```

## Multiple except Clauses

Handle different exception types separately. Order matters: more specific first.

```python
try:
    # some operation
    x = int("42")
    y = 1 / 0
except ValueError:
    print("Invalid conversion")
except ZeroDivisionError:
    print("Division by zero")
except (TypeError, KeyError):
    print("Type or key error")
```

## else and finally

`else` runs if no exception occurred. `finally` always runs (cleanup).

```python
try:
    f = open("nonexistent.txt")
except FileNotFoundError:
    print("File not found")
else:
    print(f.read())
    f.close()
finally:
    print("Cleanup done")

try:
    x = 1 + 1
except Exception:
    pass
else:
    print("No exception", x)
```

## Raising Exceptions

Use `raise` to throw exceptions. Optionally chain with `from`.

```python
def validate_age(age):
    if age < 0 or age > 150:
        raise ValueError("Age must be between 0 and 150")
    return age

try:
    validate_age(-5)
except ValueError as e:
    print(e)

try:
    int("x")
except ValueError as e:
    raise RuntimeError("Parse failed") from e
```

## Custom Exceptions

Define custom exceptions by subclassing `Exception` (or a more specific base).

```python
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Balance {balance} < amount {amount}")

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount

try:
    withdraw(10, 20)
except InsufficientFundsError as e:
    print(e, e.balance, e.amount)
```

## Exception Hierarchy

All exceptions inherit from `BaseException`. Use `Exception` for most custom exceptions.

```python
import builtins

print(issubclass(ValueError, Exception))
print(issubclass(Exception, BaseException))

for name in dir(builtins):
    obj = getattr(builtins, name)
    if isinstance(obj, type) and issubclass(obj, BaseException):
        print(name)
```

## Catching BaseException

Avoid catching `BaseException` or bare `except:`—you'll catch `KeyboardInterrupt` and `SystemExit`.

```python
try:
    x = 1 / 0
except Exception:
    print("Caught")

# Prefer specific:
try:
    x = int("x")
except ValueError:
    print("Invalid int")
```

## Best Practices

1. Catch specific exceptions. 2. Use `else` for success path. 3. Use `finally` for cleanup. 4. Don't swallow exceptions silently. 5. Document what you raise.

```python
def read_config(path):
    """Read config file. Raises FileNotFoundError if missing."""
    try:
        with open(path) as f:
            return f.read()
    except FileNotFoundError:
        raise
    except PermissionError:
        raise PermissionError(f"Cannot read {path}") from None
```
