# Iterators and Generators

Welcome! Iterators and generators provide lazy, memory-efficient iteration. This guide covers the iterator protocol, generators, `yield`, generator expressions, `itertools`, and `functools`.

## The Iterator Protocol

An iterator implements `__iter__` (returns self) and `__next__` (returns next item or raises StopIteration).

```python
class CountDown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        self.current -= 1
        return self.current + 1

for n in CountDown(3):
    print(n)

it = iter([1, 2, 3])
print(next(it), next(it), next(it))
```

## Iterables vs Iterators

Iterables have `__iter__`; iterators also have `__next__`. `iter()` gets an iterator from an iterable.

```python
lst = [1, 2, 3]
it = iter(lst)
print(hasattr(lst, '__iter__'))
print(hasattr(it, '__next__'))
print(next(it), next(it), next(it))
# next(it)  # StopIteration
```

## Generators

Generators are functions that use `yield`. They produce values lazily and maintain state between calls.

```python
def count_up_to(n):
    i = 0
    while i < n:
        yield i
        i += 1

for x in count_up_to(5):
    print(x, end=" ")
print()

gen = count_up_to(3)
print(next(gen), next(gen), next(gen))
```

## yield and yield from

`yield` produces a value. `yield from` delegates to another generator.

```python
def chain(*iterables):
    for it in iterables:
        yield from it

print(list(chain([1, 2], [3, 4], [5])))

def flatten(nested):
    for item in nested:
        if isinstance(item, list):
            yield from flatten(item)
        else:
            yield item

print(list(flatten([1, [2, [3, 4], 5], 6])))
```

## Generator Expressions

Generator expressions are like list comprehensions but produce generators. Use `()` instead of `[]`.

```python
squares = (x**2 for x in range(5))
print(squares)
print(list(squares))

evens = (x for x in range(10) if x % 2 == 0)
print(sum(evens))

total = sum(x**2 for x in range(100))
print(total)
```

## itertools

`itertools` provides many iterator-building tools: count, cycle, chain, islice, combinations, permutations.

```python
import itertools

print(list(itertools.islice(itertools.count(10), 5)))
print(list(itertools.islice(itertools.cycle([1, 2]), 6)))
print(list(itertools.chain([1, 2], [3, 4])))
print(list(itertools.combinations([1, 2, 3], 2)))
print(list(itertools.permutations([1, 2], 2)))
print(list(itertools.groupby("aaaabbbcc")))
for k, g in itertools.groupby("aaaabbbcc"):
    print(k, list(g))
```

## functools

`functools` provides `reduce`, `partial`, `lru_cache`, `wraps`.

```python
from functools import reduce, partial, lru_cache

print(reduce(lambda a, b: a * b, [1, 2, 3, 4]))

def power(base, exp):
    return base ** exp

square = partial(power, exp=2)
print(square(5))

@lru_cache(maxsize=128)
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)

print(fib(30))
```

## Sending Values to Generators

Generators can receive values via `send()`. Use `yield` as an expression.

```python
def echo():
    while True:
        received = yield
        print(f"Echo: {received}")

g = echo()
next(g)
g.send("hello")
g.send("world")
```
