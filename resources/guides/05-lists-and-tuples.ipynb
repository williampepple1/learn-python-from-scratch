# Lists and Tuples

Welcome! Lists and tuples are sequence types. Lists are mutable; tuples are immutable. This guide covers CRUD operations, slicing, sorting, unpacking, and list comprehensions.

## Creating Lists and Tuples

Lists use `[]`; tuples use `()`. A single-element tuple needs a trailing comma.

```python
lst = [1, 2, 3]
tup = (1, 2, 3)
single = (1,)  # Not (1) which is just 1
empty_list = []
empty_tup = ()
print(lst, tup, single)
```

## Indexing and Slicing

Both support zero-based indexing and slicing. Negative indices count from the end.

```python
lst = [10, 20, 30, 40, 50]
print(lst[0], lst[-1])
print(lst[1:4])
print(lst[::2])
print(lst[::-1])
```

## List CRUD: Create, Read, Update, Delete

Lists support in-place modification. Tuples do not.

```python
lst = [1, 2, 3]
lst.append(4)
lst.extend([5, 6])
lst.insert(0, 0)
print(lst)

lst[1] = 99
lst[2:4] = [10, 11]
print(lst)

lst.remove(99)
popped = lst.pop()
del lst[0]
print(lst, popped)
```

## List Methods

`sort`, `reverse`, `count`, `index`, `clear`.

```python
lst = [3, 1, 4, 1, 5]
lst.sort()
print(lst)

lst.sort(reverse=True)
print(lst)

lst = [3, 1, 4, 1, 5]
sorted_lst = sorted(lst)
print(lst, sorted_lst)

print(lst.count(1))
print(lst.index(4))
lst.clear()
print(lst)
```

## Tuples: Immutable Sequences

Tuples are fixed once created. Use them for fixed data, dict keys, and multiple return values.

```python
point = (3, 4)
rgb = (255, 128, 0)
print(point[0], point[1])

# point[0] = 5  # TypeError

d = {(1, 2): "value"}
print(d[(1, 2)])
```

## Unpacking

Unpack sequences into variables. Use `*` to capture remaining elements.

```python
a, b, c = [1, 2, 3]
print(a, b, c)

first, *rest = [1, 2, 3, 4, 5]
print(first, rest)

*front, last = [1, 2, 3, 4, 5]
print(front, last)

x, y = 10, 20
x, y = y, x  # Swap
print(x, y)
```

## List Comprehensions

List comprehensions build lists concisely. Syntax: `[expr for item in iterable if condition]`.

```python
squares = [x**2 for x in range(5)]
print(squares)

evens = [x for x in range(10) if x % 2 == 0]
print(evens)

matrix = [[i*j for j in range(3)] for i in range(3)]
print(matrix)

flat = [x for row in matrix for x in row]
print(flat)
```

## Tuple Comprehensions?

Use generator expressions with `tuple()` for tuple "comprehensions".

```python
tup = tuple(x**2 for x in range(5))
print(tup)
```

## Copying Lists

Assignment creates a reference. Use `copy()` or slicing for a shallow copy.

```python
a = [1, 2, 3]
b = a
b.append(4)
print(a)

a = [1, 2, 3]
c = a.copy()
d = a[:]
c.append(4)
print(a, c)

import copy
nested = [[1, 2], [3, 4]]
deep = copy.deepcopy(nested)
```

## Common Operations

`len`, `min`, `max`, `sum`, `in`, `+`, `*`.

```python
lst = [3, 1, 4, 1, 5]
print(len(lst), min(lst), max(lst), sum(lst))
print(4 in lst)
print([1, 2] + [3, 4])
print([0] * 5)
```
