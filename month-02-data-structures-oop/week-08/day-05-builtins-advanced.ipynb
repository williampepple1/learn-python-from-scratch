## map - Apply Function to Each Element

`map(function, iterable, ...)` returns an iterator that applies function to each element. Stops at shortest iterable if multiple.

```python
nums = [1, 2, 3, 4, 5]
squared = map(lambda x: x ** 2, nums)
print(list(squared))

# Multiple iterables
a = [1, 2, 3]
b = [10, 20, 30]
sums = map(lambda x, y: x + y, a, b)
print(list(sums))
```

## filter - Keep Elements That Pass Test

`filter(function, iterable)` yields elements for which function returns truthy. Use `None` to filter out falsy values.

```python
nums = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, nums)
print(list(evens))

# Filter out None/falsy
data = [1, 0, None, 2, '', 3, False]
truthy = filter(None, data)
print(list(truthy))
```

## zip - Pair Elements

`zip(*iterables)` yields tuples (a, b, c...) from parallel iterables. Stops at shortest.

```python
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
pairs = list(zip(names, ages))
print(pairs)

# Unzip with *
zipped = [(1, 'a'), (2, 'b'), (3, 'c')]
nums, letters = zip(*zipped)
print(nums, letters)
```

## zip_longest (itertools)

For padding to longest length, use `itertools.zip_longest`.

```python
from itertools import zip_longest

a = [1, 2, 3]
b = ['a', 'b']
print(list(zip_longest(a, b, fillvalue='?')))
```

## enumerate - Index with Value

`enumerate(iterable, start=0)` yields (index, value) pairs.

```python
fruits = ['apple', 'banana', 'cherry']
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

print(list(enumerate(fruits, start=1)))
```

## any and all

`any(iterable)` → True if any element is truthy. `all(iterable)` → True if all are truthy. Empty: any→False, all→True.

```python
print(any([False, False, True]))
print(all([True, True, True]))
print(any([]))
print(all([]))

# Practical: check if any string is empty
names = ['Alice', 'Bob', 'Charlie']
print(all(len(n) > 0 for n in names))
```

## sorted with key

`sorted(iterable, key=None, reverse=False)` returns a new sorted list. `key` is a function applied to each element for comparison.

```python
words = ['banana', 'pie', 'Washington', 'book']
by_length = sorted(words, key=len)
print(by_length)

students = [('Alice', 85), ('Bob', 72), ('Charlie', 90)]
by_grade = sorted(students, key=lambda x: x[1], reverse=True)
print(by_grade)

# Case-insensitive sort
print(sorted(['Banana', 'apple', 'Cherry'], key=str.lower))
```

## reversed

`reversed(seq)` returns a reverse iterator. Works on sequences (list, tuple, string) or objects with __reversed__.

```python
print(list(reversed([1, 2, 3, 4])))
print(list(reversed('hello')))

# reversed returns iterator - use once
r = reversed([1, 2, 3])
print(list(r))
print(list(r))  # Empty - exhausted
```

## sum, min, max with key

`sum(iterable, start=0)`. `min`/`max` accept `key` like sorted.

```python
print(sum([1, 2, 3, 4, 5]))
print(sum([[1], [2], [3]], []))  # Flatten with start=[]

words = ['apple', 'pie', 'banana']
print(max(words, key=len))
print(min(words, key=len))
```

## Practical: Data Processing Pipeline

Combine map, filter, zip in a pipeline.

```python
data = [('Alice', 25), ('Bob', 17), ('Charlie', 22), ('Diana', 16)]
# Adults only, extract names, uppercase
adults = filter(lambda p: p[1] >= 18, data)
names = map(lambda p: p[0], adults)
upper_names = map(str.upper, names)
print(list(upper_names))
```

## Common Mistakes

- **map/filter return iterators**: Consume once; convert to list if needed multiple times
- **zip stops at shortest**: Use zip_longest for padding
- **sorted vs list.sort**: sorted returns new list; .sort() mutates in place
- **key function**: Must return comparable values (e.g., (priority, name) for multi-criteria sort)

```python
# WRONG: Reusing map iterator
m = map(str.upper, ['a', 'b', 'c'])
list(m)  # Consumes
# list(m)  # Empty!

# CORRECT: Create new or store list
result = list(map(str.upper, ['a', 'b', 'c']))
print(result)
print(result)
```

## Key Takeaways

1. `map(f, it)` applies f to each; `filter(pred, it)` keeps where pred is truthy
2. `zip` pairs iterables; stops at shortest
3. `enumerate` gives (index, value)
4. `any`/`all` short-circuit; empty any→False, all→True
5. `sorted(it, key=f)` uses f for comparison; `reversed` for reverse iteration
