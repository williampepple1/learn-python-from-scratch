## itertools Overview

`itertools` provides efficient iterator-building blocks. All return iterators (lazy).

```python
import itertools
print([x for x in dir(itertools) if not x.startswith('_')][:15])
```

## chain - Combine Iterables

`chain(*iterables)` yields elements from each iterable in sequence.

```python
from itertools import chain

result = chain([1, 2], [3, 4], [5, 6])
print(list(result))

# chain.from_iterable for nested iterables
nested = [[1, 2], [3, 4], [5]]
print(list(chain.from_iterable(nested)))
```

## product - Cartesian Product

`product(A, B)` yields all pairs (a, b) for a in A, b in B. Like nested loops.

```python
from itertools import product

print(list(product([1, 2], ['a', 'b'])))
print(list(product([0, 1], repeat=3)))  # 0,1 x 0,1 x 0,1
```

## permutations - All Orderings

`permutations(iterable, r=None)` yields all r-length permutations. Default r = full length.

```python
from itertools import permutations

print(list(permutations([1, 2, 3])))
print(list(permutations([1, 2, 3], r=2)))
```

## combinations - Unordered Subsets

`combinations(iterable, r)` yields all r-length combinations (order doesn't matter).

```python
from itertools import combinations

print(list(combinations([1, 2, 3, 4], 2)))
print(list(combinations('ABC', 2)))
```

## combinations_with_replacement

Allows repeated elements. `combinations_with_replacement('AB', 2)` → AA, AB, BB.

```python
from itertools import combinations_with_replacement

print(list(combinations_with_replacement([1, 2], 3)))
```

## groupby - Consecutive Groups

`groupby(iterable, key=None)` groups consecutive elements with same key. **Sort first** if not already grouped.

```python
from itertools import groupby

data = [1, 1, 2, 2, 2, 3, 1]
for key, group in groupby(data):
    print(f"{key}: {list(group)}")

# With key function
words = ['apple', 'ape', 'banana', 'berry']
for letter, group in groupby(words, key=lambda w: w[0]):
    print(f"{letter}: {list(group)}")
```

## islice - Slice an Iterator

`islice(iterable, start, stop, step)` - like slice but works on iterators. Doesn't consume the whole iterable.

```python
from itertools import islice

data = range(100)
print(list(islice(data, 5)))        # First 5
print(list(islice(data, 2, 8)))     # Indices 2-7
print(list(islice(data, 0, 10, 2)))  # Even indices 0-8
```

## count - Infinite Counter

`count(start=0, step=1)` yields an infinite sequence. Use with islice to limit.

```python
from itertools import count, islice

print(list(islice(count(10, 2), 5)))  # 10, 12, 14, 16, 18
```

## cycle - Repeat Forever

`cycle(iterable)` repeats the iterable indefinitely.

```python
from itertools import cycle, islice

cycler = cycle([1, 2, 3])
print(list(islice(cycler, 8)))
```

## repeat, tee

`repeat(obj, times)` yields obj n times. `tee(iterable, n=2)` splits one iterator into n independent iterators.

```python
from itertools import repeat, tee

print(list(repeat(7, 4)))

it = iter([1, 2, 3])
a, b = tee(it, 2)
print(list(a), list(b))
```

## Practical: Generate All Passwords

Use product to generate combinations of characters.

```python
from itertools import product

chars = 'ab'
length = 3
passwords = [''.join(p) for p in product(chars, repeat=length)]
print(passwords)
```

## Common Mistakes

- **groupby on unsorted data**: Groups only consecutive elements; sort first if needed
- **Consuming group**: Each group is an iterator; list() it before next iteration
- **cycle/count infinite**: Use islice or break to avoid infinite loops
- **tee memory**: tee stores consumed values; don't use on huge iterables

```python
# WRONG: groupby without sorting - non-consecutive same values in different groups
from itertools import groupby
data = [1, 2, 1, 2, 1]
for k, g in groupby(data):
    print(k, list(g))  # 1:[1], 2:[2], 1:[1], 2:[2], 1:[1]

# CORRECT: sort first for logical grouping
for k, g in groupby(sorted(data)):
    print(k, list(g))  # 1:[1,1,1], 2:[2,2]
```

## Key Takeaways

1. `chain` combines iterables; `product` = Cartesian product
2. `permutations` = order matters; `combinations` = order doesn't
3. `groupby` groups consecutive elements; sort first for logical groups
4. `islice` slices iterators; `count`/`cycle` are infinite
5. All itertools return iterators (lazy)
