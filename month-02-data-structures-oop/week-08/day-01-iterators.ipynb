## Iterable vs Iterator

**Iterable**: object that can be iterated (has `__iter__`). Examples: list, tuple, dict, set, string.

**Iterator**: object that produces values one at a time (has `__iter__` and `__next__`). Created by calling `iter()` on an iterable.

```python
lst = [1, 2, 3]
it = iter(lst)
print(next(it))
print(next(it))
print(next(it))
# print(next(it))  # StopIteration
```

## How for-loops Work

`for x in obj:` calls `iter(obj)` to get an iterator, then repeatedly calls `next()` until StopIteration.

```python
it = iter([10, 20, 30])
while True:
    try:
        x = next(it)
        print(x)
    except StopIteration:
        break
```

## Iterator Protocol

An iterator must implement:
- `__iter__` → returns self
- `__next__` → returns next value or raises StopIteration

```python
class CountDown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        self.current -= 1
        return self.current + 1

for n in CountDown(3):
    print(n)
```

## Iterable vs Iterator: Separate Classes

Iterable returns a **new** iterator each time `iter()` is called. Iterator returns **self**. For reusable iteration, keep them separate.

```python
class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __iter__(self):
        return MyRangeIterator(self.start, self.end)

class MyRangeIterator:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.end:
            raise StopIteration
        val = self.current
        self.current += 1
        return val

for x in MyRange(1, 4):
    print(x)
print("Again:")
for x in MyRange(1, 4):
    print(x)
```

## StopIteration

Raise `StopIteration` when there are no more items. The for-loop catches it and exits.

```python
class Squares:
    def __init__(self, n):
        self.n = n
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        val = self.i ** 2
        self.i += 1
        return val

print(list(Squares(5)))
```

## iter() with Sentinel

`iter(callable, sentinel)` creates an iterator that calls the callable until it returns the sentinel value.

```python
import random
def roll():
    return random.randint(1, 6)

# Iterate until we get 6
rolls = []
for r in iter(roll, 6):
    rolls.append(r)
print("Rolls until 6:", rolls)
```

## Practical: Fibonacci Iterator

Custom iterator that yields Fibonacci numbers.

```python
class Fibonacci:
    def __init__(self, limit=None):
        self.limit = limit
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.limit is not None and self.count >= self.limit:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

print(list(Fibonacci(8)))
```

## Common Mistakes

- **Iterator exhausted**: Same iterator can't be iterated twice; create new one with iter()
- **Forgetting __iter__ on iterator**: Iterator must have __iter__ returning self
- **Not raising StopIteration**: Loop never ends; must raise when done
- **Mutating during iteration**: Modifying a list while iterating can cause unexpected behavior

```python
# WRONG: Reusing exhausted iterator
it = iter([1, 2, 3])
list(it)  # Consumes it
# list(it)  # Empty! Iterator is exhausted

# CORRECT: Get fresh iterator
lst = [1, 2, 3]
list(iter(lst))
list(iter(lst))  # Works - new iterator each time
```

## Key Takeaways

1. Iterable has `__iter__`; Iterator has `__iter__` (returns self) and `__next__`
2. `iter(obj)` gets iterator; `next(it)` gets next value
3. Raise `StopIteration` when no more items
4. Iterators are one-shot; reuse iterable to get fresh iterator
5. for-loop = iter() + next() until StopIteration
