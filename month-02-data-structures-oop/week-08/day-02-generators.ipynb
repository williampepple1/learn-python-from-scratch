## Generator Functions

A function with `yield` is a **generator function**. Calling it returns a generator (iterator) that yields values one at a time. Execution pauses at each yield.

```python
def count_up(n):
    for i in range(1, n + 1):
        yield i

gen = count_up(3)
print(next(gen))
print(next(gen))
print(next(gen))
# next(gen)  # StopIteration
```

## yield vs return

`yield` produces a value and **pauses**; `return` ends the generator. Values after return become the StopIteration value.

```python
def simple_gen():
    yield 1
    yield 2
    yield 3

print(list(simple_gen()))
```

## Lazy Evaluation

Generators produce values on demand. They don't store the full sequence in memory.

```python
def infinite_sequence():
    n = 0
    while True:
        yield n
        n += 1

gen = infinite_sequence()
print([next(gen) for _ in range(5)])
```

## Generator Expressions

Like list comprehensions but with parentheses. Produces a generator, not a list.

```python
squares = (x**2 for x in range(5))
print(type(squares))
print(list(squares))
# squares is exhausted now
squares2 = (x**2 for x in range(5))
print(next(squares2), next(squares2))
```

## send() - Sending Values into Generator

`generator.send(value)` sends a value into the generator. The yield expression receives it. First send must be `None` (or use `next()`).

```python
def echo():
    while True:
        received = yield
        if received is None:
            break
        print(f"Echo: {received}")

g = echo()
next(g)  # Prime the generator
g.send("hello")
g.send("world")
g.send(None)  # Stop
```

## throw() - Raise Exception in Generator

`generator.throw(exc)` raises an exception at the yield point. Useful for cleanup or signaling.

```python
def fragile():
    try:
        yield 1
        yield 2
    except ValueError:
        print("Caught ValueError")
        yield "recovered"

g = fragile()
print(next(g))
print(g.throw(ValueError))
print(next(g))
```

## Generator Pipelines

Chain generators: each consumes from the previous. Memory-efficient for large data.

```python
def read_lines(lines):
    for line in lines:
        yield line.strip()

def filter_empty(lines):
    for line in lines:
        if line:
            yield line

def uppercase(lines):
    for line in lines:
        yield line.upper()

data = ['  a  ', '', '  b  ', '  c  ']
pipeline = uppercase(filter_empty(read_lines(data)))
print(list(pipeline))
```

## yield from

`yield from iterable` delegates to a sub-generator. Simplifies nested generators.

```python
def flatten(nested):
    for item in nested:
        if isinstance(item, list):
            yield from flatten(item)
        else:
            yield item

nested = [1, [2, 3, [4, 5]], 6]
print(list(flatten(nested)))
```

## Practical: Reading Large Files

Process file line by line without loading entire file.

```python
def read_large_file(filepath):
    with open(filepath) as f:
        for line in f:
            yield line.rstrip()

# Simulate with a list of lines
lines = ['line1\n', 'line2\n', 'line3\n']
def mock_read(lines_list):
    for line in lines_list:
        yield line.rstrip()

print(list(mock_read(lines)))
```

## Common Mistakes

- **Forgetting to prime**: send() requires generator to be at a yield; use next() first
- **Generator vs list**: (x for x in range(5)) is lazy; [x for x in range(5)] is eager
- **Reusing exhausted generator**: Create new one
- **yield in try/finally**: finally runs when generator is closed (e.g., garbage collected)

```python
# WRONG: send without priming
# g = (lambda: (yield))()  # Complex example
# g.send(1)  # TypeError: can't send non-None value to a just-started generator

# CORRECT: Prime with next()
def recv():
    while True:
        x = yield
        if x is None:
            return
        print(x)
g = recv()
next(g)
g.send(42)
```
