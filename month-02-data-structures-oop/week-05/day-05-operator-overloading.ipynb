## Arithmetic Operators

Override `__add__`, `__sub__`, `__mul__`, etc. to make your objects work with `+`, `-`, `*`. For `a + b`, Python calls `a.__add__(b)`.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, Vector):
            return NotImplemented
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        if not isinstance(other, Vector):
            return NotImplemented
        return Vector(self.x - other.x, self.y - other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)
print(v2 - v1)
```

## `__mul__` and `__rmul__`

`__mul__` handles `a * b`. For `3 * v` when `Vector` is on the right, Python tries `int.__mul__(v)` first (fails), then `v.__rmul__(3)`.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, scalar):
        if not isinstance(scalar, (int, float)):
            return NotImplemented
        return Vector(self.x * scalar, self.y * scalar)

    def __rmul__(self, scalar):
        return self.__mul__(scalar)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v = Vector(2, 3)
print(v * 5)
print(5 * v)
```

## `__getitem__` - Indexing

Enables `obj[key]` and slicing. Receive the key (index or slice object).

```python
class Sentence:
    def __init__(self, text):
        self.words = text.split()

    def __getitem__(self, key):
        if isinstance(key, slice):
            return ' '.join(self.words[key])
        return self.words[key]

    def __len__(self):
        return len(self.words)

s = Sentence("Hello world from Python")
print(s[0])
print(s[1:3])
print(s[-1])
```

## `__setitem__` - Assignment

Enables `obj[key] = value`. Makes your object mutable via indexing.

```python
class MutableList:
    def __init__(self, data=None):
        self._data = list(data) if data else []

    def __getitem__(self, key):
        return self._data[key]

    def __setitem__(self, key, value):
        self._data[key] = value

    def __len__(self):
        return len(self._data)

    def __repr__(self):
        return repr(self._data)

ml = MutableList([10, 20, 30])
ml[1] = 99
print(ml)
print(ml[0], ml[1])
```

## `__contains__` - The `in` Operator

Defines `item in obj`. If not implemented, Python falls back to iterating (calling `__iter__` or `__getitem__`).

```python
class Range:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __contains__(self, item):
        return self.start <= item < self.end

    def __repr__(self):
        return f"Range({self.start}, {self.end})"

r = Range(10, 20)
print(15 in r)
print(5 in r)
print(20 in r)
```

## Real-World: Money Class

Arithmetic for currency amounts.

```python
class Money:
    def __init__(self, amount, currency="USD"):
        self.amount = amount
        self.currency = currency

    def __add__(self, other):
        if not isinstance(other, Money) or other.currency != self.currency:
            return NotImplemented
        return Money(self.amount + other.amount, self.currency)

    def __sub__(self, other):
        if not isinstance(other, Money) or other.currency != self.currency:
            return NotImplemented
        return Money(self.amount - other.amount, self.currency)

    def __mul__(self, scalar):
        if not isinstance(scalar, (int, float)):
            return NotImplemented
        return Money(self.amount * scalar, self.currency)

    def __repr__(self):
        return f"Money({self.amount}, '{self.currency}')"

m1 = Money(100)
m2 = Money(50)
print(m1 + m2)
print(m1 - m2)
print(m1 * 2)
```

## Complete Indexable Example

A simple sparse list that supports get, set, and `in`.

```python
class SparseList:
    def __init__(self, default=0):
        self._data = {}
        self._default = default

    def __getitem__(self, key):
        return self._data.get(key, self._default)

    def __setitem__(self, key, value):
        self._data[key] = value

    def __contains__(self, item):
        return item in self._data.values() or item == self._default

sl = SparseList(default=0)
sl[0] = 10
sl[5] = 20
print(sl[0], sl[1], sl[5])
print(10 in sl, 0 in sl)
```

## Common Mistakes

- **Forgetting `__rmul__`**: `5 * v` fails if only `__mul__` defined (int doesn't know Vector)
- **Returning wrong type**: `__add__` should return new instance, not modify `self`
- **Not handling `NotImplemented`**: Return it for unsupported types so Python can try reverse op
- **`__setitem__` without `__getitem__`**: Usually implement both for indexing
- **`__contains__` for large ranges**: Implement it for O(1) instead of O(n) iteration

## Key Takeaways

- `__add__`, `__sub__`, `__mul__`: arithmetic; use `__rmul__` for `scalar * obj`
- `__getitem__(self, key)`: enables `obj[key]` and slicing
- `__setitem__(self, key, value)`: enables `obj[key] = value`
- `__contains__(self, item)`: enables `item in obj`
- Return `NotImplemented` for unsupported operand types
