## Project Overview

We'll build a library system with:
- **Book**: title, author, ISBN, borrowed status
- **Member**: name, member_id, list of borrowed books
- **Library**: catalog of books, members, borrow/return logic

## Step 1: Create the Book Class

Books have title, author, ISBN, and track whether they're borrowed.

```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self._borrowed = False

    @property
    def borrowed(self):
        return self._borrowed

    def set_borrowed(self, value):
        self._borrowed = value

    def __repr__(self):
        status = "borrowed" if self._borrowed else "available"
        return f"Book('{self.title}', '{self.author}', {status})"

# Test
b = Book("Python Crash Course", "Eric Matthes", "978-1593275990")
print(b)
```

## Step 2: Create the Member Class

Members have a name, ID, and can hold borrowed books.

```python
class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    def borrow(self, book):
        self.borrowed_books.append(book)

    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)

    def __repr__(self):
        return f"Member('{self.name}', '{self.member_id}', {len(self.borrowed_books)} books)"

# Test
m = Member("Alice", "M001")
print(m)
```

## Step 3: Create the Library Class - Setup

Library holds books and members. We'll add methods incrementally.

```python
class Library:
    def __init__(self, name):
        self.name = name
        self._books = []
        self._members = []

    def add_book(self, book):
        self._books.append(book)
        print(f"Added: {book.title}")

    def add_member(self, member):
        self._members.append(member)
        print(f"Registered: {member.name}")

    def find_book_by_isbn(self, isbn):
        for b in self._books:
            if b.isbn == isbn:
                return b
        return None

    def find_member_by_id(self, member_id):
        for m in self._members:
            if m.member_id == member_id:
                return m
        return None

lib = Library("Central Library")
lib.add_book(Book("Clean Code", "Robert Martin", "978-0132350884"))
lib.add_member(Member("Bob", "M002"))
print(lib.find_book_by_isbn("978-0132350884"))
```

## Step 4: Add Borrow and Return Logic

Library coordinates borrowing and returning.

```python
class Library:
    def __init__(self, name):
        self.name = name
        self._books = []
        self._members = []

    def add_book(self, book):
        self._books.append(book)

    def add_member(self, member):
        self._members.append(member)

    def find_book_by_isbn(self, isbn):
        for b in self._books:
            if b.isbn == isbn:
                return b
        return None

    def find_member_by_id(self, member_id):
        for m in self._members:
            if m.member_id == member_id:
                return m
        return None

    def borrow_book(self, isbn, member_id):
        book = self.find_book_by_isbn(isbn)
        member = self.find_member_by_id(member_id)
        if not book:
            return "Book not found"
        if not member:
            return "Member not found"
        if book.borrowed:
            return "Book already borrowed"
        book.set_borrowed(True)
        member.borrow(book)
        return f"{member.name} borrowed '{book.title}'"

    def return_book(self, isbn, member_id):
        book = self.find_book_by_isbn(isbn)
        member = self.find_member_by_id(member_id)
        if not book or not member:
            return "Book or member not found"
        if book not in member.borrowed_books:
            return f"{member.name} did not borrow this book"
        book.set_borrowed(False)
        member.return_book(book)
        return f"'{book.title}' returned"

    def list_available_books(self):
        return [b for b in self._books if not b.borrowed]

    def list_borrowed_books(self):
        return [b for b in self._books if b.borrowed]

lib = Library("Central")
lib.add_book(Book("Python", "Author", "123"))
lib.add_member(Member("Alice", "M1"))
print(lib.borrow_book("123", "M1"))
print(lib.list_available_books())
print(lib.return_book("123", "M1"))
```

## Step 5: Complete System - Full Code

All classes together with `__str__` and a demo.

```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self._borrowed = False

    @property
    def borrowed(self):
        return self._borrowed

    def set_borrowed(self, value):
        self._borrowed = value

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __repr__(self):
        status = "borrowed" if self._borrowed else "available"
        return f"Book('{self.title}', '{self.author}', {status})"


class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    def borrow(self, book):
        self.borrowed_books.append(book)

    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)

    def __repr__(self):
        return f"Member('{self.name}', '{self.member_id}')"


class Library:
    def __init__(self, name):
        self.name = name
        self._books = []
        self._members = []

    def add_book(self, book):
        self._books.append(book)

    def add_member(self, member):
        self._members.append(member)

    def find_book_by_isbn(self, isbn):
        for b in self._books:
            if b.isbn == isbn:
                return b
        return None

    def find_member_by_id(self, member_id):
        for m in self._members:
            if m.member_id == member_id:
                return m
        return None

    def borrow_book(self, isbn, member_id):
        book = self.find_book_by_isbn(isbn)
        member = self.find_member_by_id(member_id)
        if not book:
            return "Book not found"
        if not member:
            return "Member not found"
        if book.borrowed:
            return "Book already borrowed"
        book.set_borrowed(True)
        member.borrow(book)
        return f"{member.name} borrowed '{book.title}'"

    def return_book(self, isbn, member_id):
        book = self.find_book_by_isbn(isbn)
        member = self.find_member_by_id(member_id)
        if not book or not member:
            return "Book or member not found"
        if book not in member.borrowed_books:
            return f"{member.name} did not borrow this book"
        book.set_borrowed(False)
        member.return_book(book)
        return f"'{book.title}' returned"

    def list_available_books(self):
        return [b for b in self._books if not b.borrowed]

    def list_borrowed_books(self):
        return [b for b in self._books if b.borrowed]

    def __str__(self):
        return f"{self.name}: {len(self._books)} books, {len(self._members)} members"
```

## Step 6: Run the Demo

Simulate a library session.

```python
# Create library and populate
lib = Library("Downtown Public Library")

lib.add_book(Book("Python Crash Course", "Eric Matthes", "978-1593275990"))
lib.add_book(Book("Clean Code", "Robert Martin", "978-0132350884"))
lib.add_book(Book("The Pragmatic Programmer", "Hunt & Thomas", "978-0135957059"))

lib.add_member(Member("Alice", "M001"))
lib.add_member(Member("Bob", "M002"))

print(lib)
print()
```

```python
# Alice borrows a book
print(lib.borrow_book("978-1593275990", "M001"))
print("Available:", [b.title for b in lib.list_available_books()])
print("Borrowed:", [b.title for b in lib.list_borrowed_books()])
print()
```

```python
# Bob tries to borrow same book
print(lib.borrow_book("978-1593275990", "M002"))

# Bob borrows different book
print(lib.borrow_book("978-0132350884", "M002"))
print()
```

```python
# Alice returns her book
print(lib.return_book("978-1593275990", "M001"))
print("Available:", [b.title for b in lib.list_available_books()])
print("Alice's books:", [b.title for b in lib.find_member_by_id("M001").borrowed_books])
print("Bob's books:", [b.title for b in lib.find_member_by_id("M002").borrowed_books])
```
