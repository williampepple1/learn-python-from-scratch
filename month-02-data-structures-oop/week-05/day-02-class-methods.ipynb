## Instance Methods

Instance methods operate on a specific object. They receive `self` as the first parameter and can access/modify instance attributes.

```python
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

    def get_count(self):
        return self.count

c = Counter()
c.increment()
c.increment()
print(c.get_count())
```

## Class Attributes

Class attributes are shared by all instances. Define them at the class level (outside methods). Access via `ClassName.attr` or `self.attr`.

```python
class Dog:
    species = "Canis familiaris"  # Shared by all dogs

    def __init__(self, name):
        self.name = name  # Instance attribute

d1 = Dog("Rex")
d2 = Dog("Max")
print(d1.species, d2.species)  # Same value
print(Dog.species)
Dog.species = "Canis lupus"  # Affects all instances
print(d1.species)
```

## `@classmethod`

Class methods receive the class (`cls`) as first parameter, not `self`. Use for factory methods or when you need the class but not a specific instance.

```python
class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    @classmethod
    def from_string(cls, date_str):
        """Factory: create Date from 'YYYY-MM-DD' string."""
        y, m, d = map(int, date_str.split('-'))
        return cls(y, m, d)

    @classmethod
    def today(cls):
        import datetime
        now = datetime.date.today()
        return cls(now.year, now.month, now.day)

d1 = Date.from_string("2025-03-10")
print(f"{d1.year}-{d1.month:02d}-{d1.day:02d}")
d2 = Date.today()
print(f"Today: {d2.year}-{d2.month:02d}-{d2.day:02d}")
```

## `@staticmethod`

Static methods don't receive `self` or `cls`. They're just functions that belong to the class namespace. Use for utility functions related to the class.

```python
class MathUtils:
    @staticmethod
    def is_even(n):
        return n % 2 == 0

    @staticmethod
    def factorial(n):
        if n <= 1:
            return 1
        return n * MathUtils.factorial(n - 1)

print(MathUtils.is_even(4))
print(MathUtils.factorial(5))
# Can also call on instance
m = MathUtils()
print(m.is_even(7))
```

## When to Use Each

- **Instance method**: Needs `self`—operates on instance data
- **Class method**: Needs `cls`—factory, alternate constructors, class-level logic
- **Static method**: Needs neither—utility, doesn't touch instance or class state

```python
class Employee:
    count = 0  # Class attribute: total employees

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.count += 1

    def get_info(self):  # Instance method
        return f"{self.name}: ${self.salary}"

    @classmethod
    def get_count(cls):
        return cls.count

    @staticmethod
    def is_valid_salary(salary):
        return isinstance(salary, (int, float)) and salary >= 0

e1 = Employee("Alice", 50000)
e2 = Employee("Bob", 60000)
print(e1.get_info())
print(Employee.get_count())
print(Employee.is_valid_salary(-100))
```

## Real-World: Product Catalog

Class methods for creating products from different sources; static methods for validation.

```python
class Product:
    tax_rate = 0.08  # Class attribute

    def __init__(self, name, price):
        self.name = name
        self.price = price

    @classmethod
    def from_dict(cls, data):
        return cls(data['name'], data['price'])

    @staticmethod
    def validate_price(price):
        return price > 0 and isinstance(price, (int, float))

    def price_with_tax(self):
        return self.price * (1 + Product.tax_rate)

p = Product.from_dict({'name': 'Widget', 'price': 19.99})
print(f"{p.name}: ${p.price_with_tax():.2f}")
print(Product.validate_price(-5))
```

## Common Mistakes

- **Using `@classmethod` when you need `self`**: Use instance method instead
- **Using `@staticmethod` for factory methods**: Use `@classmethod` to return `cls(...)`
- **Modifying class attr via instance**: `obj.attr = x` creates instance attr, doesn't change class attr
- **Forgetting `cls` in classmethod**: First param must be `cls`
- **Using `self` in staticmethod**: Static methods get neither `self` nor `cls`
