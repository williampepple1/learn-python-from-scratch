## What is a Class?

A class is a blueprint for creating objects. It defines the structure and behavior that all instances of that class will share. Think of it like a cookie cutter—the class is the cutter, and each cookie is an object.

## Defining a Class with `class`

Use the `class` keyword followed by the class name. By convention, class names use `PascalCase`.

```python
class Dog:
    pass  # Empty class - valid but does nothing yet

# Create an instance (object)
my_dog = Dog()
print(my_dog)
print(type(my_dog))
```

## The `__init__` Constructor

`__init__` is a special method called when you create a new object. It initializes the object's state. The double underscores indicate it's a "dunder" (double underscore) method.

```python
class Dog:
    def __init__(self, name, age):
        print(f"Creating a dog named {name}")
        self.name = name
        self.age = age

rex = Dog("Rex", 3)
print(rex.name, rex.age)
```

## Understanding `self`

`self` refers to the instance being created or operated on. It's the first parameter of instance methods. Python passes it automatically—you don't pass it when calling.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

account = BankAccount("Alice", 100)
account.deposit(50)  # self is passed automatically
print(f"{account.owner}'s balance: ${account.balance}")
```

## Instance Attributes

Instance attributes are unique to each object. They're typically set in `__init__` using `self.attribute_name = value`. Each object has its own copy.

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

laptop = Product("Laptop", 999.99, 5)
mouse = Product("Mouse", 29.99, 50)

print(laptop.name, laptop.price)
print(mouse.name, mouse.price)
print(laptop.quantity != mouse.quantity)  # Each has its own attributes
```

## Creating Objects

To create an object, call the class name like a function: `ClassName(arg1, arg2)`. This invokes `__init__` and returns a new instance.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

# Create multiple objects
rect1 = Rectangle(4, 5)
rect2 = Rectangle(10, 3)
rect3 = Rectangle(7, 7)

print(f"Rect1: {rect1.width}x{rect1.height}")
print(f"Rect2: {rect2.width}x{rect2.height}")
print(f"Rect3: {rect3.width}x{rect3.height}")
```

## Real-World Example: Student Record

Here's a practical class for tracking students in a course.

```python
class Student:
    def __init__(self, name, student_id, major):
        self.name = name
        self.student_id = student_id
        self.major = major

student1 = Student("Emma", "S001", "Computer Science")
student2 = Student("Liam", "S002", "Mathematics")

print(f"{student1.name} ({student1.student_id}) - {student1.major}")
print(f"{student2.name} ({student2.student_id}) - {student2.major}")
```

## Common Mistakes

- **Forgetting `self`**: Instance methods must have `self` as first parameter
- **Passing `self` when calling**: `obj.method()` not `obj.method(obj)`
- **Typo in `__init__`**: Must be exactly `__init__` (two underscores each side)
- **Using `init` instead of `__init__`**: Won't run as constructor
- **Forgetting to call `super().__init__()` in inheritance** (we'll see this later)

## Key Takeaways

- `class` defines a blueprint; call it to create objects
- `__init__(self, ...)` is the constructor—runs when object is created
- `self` refers to the current instance; always first parameter in instance methods
- Instance attributes: `self.attr = value` in `__init__`
- Each object has its own copy of instance attributes
