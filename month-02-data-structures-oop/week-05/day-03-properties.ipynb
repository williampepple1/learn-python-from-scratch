## Why Properties?

Properties let you use attribute-style access (`obj.attr`) while running custom logic behind the scenes. No need for explicit `get_x()` and `set_x()` methods—Pythonic and clean.

## The `@property` Decorator

`@property` turns a method into a read-only attribute. Access it like `obj.property_name` without parentheses.

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        return 3.14159 * self.radius ** 2

    @property
    def diameter(self):
        return self.radius * 2

c = Circle(5)
print(c.area)      # No parentheses!
print(c.diameter)
# c.area = 10  # AttributeError - read-only
```

## Getters and Setters

Use `@property` for getter and `@attr.setter` for setter. The setter runs when you assign: `obj.attr = value`.

```python
class Temperature:
    def __init__(self, celsius=0):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero")
        self._celsius = value

    @property
    def fahrenheit(self):
        return self._celsius * 9/5 + 32

    @fahrenheit.setter
    def fahrenheit(self, value):
        self.celsius = (value - 32) * 5/9

t = Temperature(25)
print(t.celsius, t.fahrenheit)
t.fahrenheit = 212
print(t.celsius)
```

## Computed Properties

Properties that derive their value from other attributes. Recalculated each time accessed.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    @property
    def area(self):
        return self.width * self.height

    @property
    def perimeter(self):
        return 2 * (self.width + self.height)

    @property
    def is_square(self):
        return self.width == self.height

r = Rectangle(4, 5)
print(r.area, r.perimeter, r.is_square)
r.width = 5
print(r.area, r.is_square)
```

## Validation in Setters

Setters are perfect for validating input before storing. Reject invalid values with clear errors.

```python
class BankAccount:
    def __init__(self, balance=0):
        self._balance = balance

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, value):
        if not isinstance(value, (int, float)):
            raise TypeError("Balance must be a number")
        if value < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = value

acc = BankAccount(100)
acc.balance = 200
print(acc.balance)
# acc.balance = -50  # ValueError
# acc.balance = "rich"  # TypeError
```

## Deleter: `@attr.deleter`

Optional. Runs when you `del obj.attr`. Use for cleanup.

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.deleter
    def name(self):
        print("Name deleted")
        del self._name

p = Person("Alice")
print(p.name)
del p.name
# print(p.name)  # AttributeError
```

## Real-World: Product with Validation

Ensure price and quantity are always valid.

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        if value <= 0:
            raise ValueError("Price must be positive")
        self._price = value

    @property
    def quantity(self):
        return self._quantity

    @quantity.setter
    def quantity(self, value):
        if not isinstance(value, int) or value < 0:
            raise ValueError("Quantity must be non-negative integer")
        self._quantity = value

    @property
    def total_value(self):
        return self._price * self._quantity

p = Product("Widget", 9.99, 10)
print(f"Total: ${p.total_value:.2f}")
p.quantity = 20
print(p.total_value)
```

## Common Mistakes

- **Infinite recursion in setter**: Use `self._attr` not `self.attr` when storing
- **Forgetting to init backing attribute**: `__init__` must set `_attr` if you use property
- **Using property for expensive computation**: Consider caching if called often
- **Setter without getter**: You need `@property` before `@attr.setter`
- **Naming conflict**: Don't name the backing var same as property (use `_` prefix)

## Key Takeaways

- `@property`: method becomes read-only attribute
- `@attr.setter`: runs on assignment, use for validation
- `@attr.deleter`: runs on `del obj.attr`
- Use `_attr` for backing storage to avoid recursion
- Computed properties derive from other attributes
