## What Are Magic Methods?

Magic (dunder) methods have double underscores. They're called implicitly by Python for built-in operations. Override them to customize your class's behavior.

## `__str__` vs `__repr__`

- `__str__`: Human-readable, for `print()` and `str()`
- `__repr__`: Unambiguous, for developers. Ideally, `eval(repr(obj))` would recreate the object.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point({self.x}, {self.y})"

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

p = Point(3, 4)
print(p)           # Uses __str__
print(str(p))
print(repr(p))     # Uses __repr__
print([p])         # Container uses __repr__ for elements
```

## `__len__`

Enables `len(obj)`. Return an integer (usually the "size" of your object).

```python
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs)

    def __len__(self):
        return len(self.songs)

    def __str__(self):
        return f"Playlist with {len(self)} songs"

pl = Playlist(["Song A", "Song B", "Song C"])
print(len(pl))
print(pl)
```

## `__eq__` (Equality)

Defines `obj1 == obj2`. Without it, `==` compares identity (same object in memory).

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        if not isinstance(other, Point):
            return NotImplemented
        return self.x == other.x and self.y == other.y

p1 = Point(1, 2)
p2 = Point(1, 2)
p3 = Point(3, 4)
print(p1 == p2)
print(p1 == p3)
print(p1 == (1, 2))  # NotImplemented -> Python tries other.__eq__
```

## `__lt__` and Comparison

`__lt__` defines `<`. With `functools.total_ordering`, you can implement just `__lt__` and `__eq__` to get all comparisons.

```python
from functools import total_ordering

@total_ordering
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __eq__(self, other):
        if not isinstance(other, Student):
            return NotImplemented
        return self.grade == other.grade

    def __lt__(self, other):
        if not isinstance(other, Student):
            return NotImplemented
        return self.grade < other.grade

s1 = Student("Alice", 85)
s2 = Student("Bob", 92)
print(s1 < s2)
print(s1 <= s2)
print(s1 > s2)
print(sorted([s2, s1])[0].name)
```

## `__hash__`

Required for objects to be used in sets and as dict keys. If you define `__eq__`, make the object immutable and define `__hash__` consistently: `hash(a)==hash(b)` when `a==b`.

```python
class ImmutablePoint:
    def __init__(self, x, y):
        object.__setattr__(self, 'x', x)
        object.__setattr__(self, 'y', y)

    def __setattr__(self, name, value):
        raise AttributeError("Immutable")

    def __eq__(self, other):
        if not isinstance(other, ImmutablePoint):
            return NotImplemented
        return self.x == other.x and self.y == other.y

    def __hash__(self):
        return hash((self.x, self.y))

p1 = ImmutablePoint(1, 2)
p2 = ImmutablePoint(1, 2)
s = {p1, p2}
print(len(s))
d = {p1: "first"}
print(d[p2])
```

## Simpler Hash Example

For mutable objects, don't define `__hash__` (default is `id`-based). For immutable value objects, use a tuple of attributes.

```python
class Card:
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit

    def __eq__(self, other):
        if not isinstance(other, Card):
            return NotImplemented
        return self.rank == other.rank and self.suit == other.suit

    def __hash__(self):
        return hash((self.rank, self.suit))

    def __repr__(self):
        return f"Card({self.rank}, {self.suit})"

c1 = Card("A", "hearts")
c2 = Card("A", "hearts")
hand = {c1, c2}
print(hand)
```

## `__bool__`

Defines truthiness for `bool(obj)` and `if obj:`. Return `True` or `False`. If not defined, Python uses `__len__` (0 = False), else True.

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

    def __repr__(self):
        return f"BankAccount({self.balance})"

acc1 = BankAccount(100)
acc2 = BankAccount(0)
print(bool(acc1), bool(acc2))
if acc1:
    print("Account has funds")
if not acc2:
    print("Account is empty")
```

## Common Mistakes

- **Returning non-string from `__str__`/`__repr__`**: Must return str
- **`__eq__` returning wrong type**: Return `NotImplemented` for unknown types, not `False`
- **Mutable objects with `__hash__`**: Don't hash mutable objects; they can't go in sets/dicts safely
- **`__bool__` returning non-bool**: Should return `True` or `False`
- **Forgetting `__hash__` when overriding `__eq__`**: Default `__hash__` is removed; object becomes unhashable
