## What is Encapsulation?

Encapsulation means hiding internal implementation details and exposing only what's necessary. In Python, we use naming conventions since there's no true "private"—it's consenting adults.

## Single Underscore: "Internal Use"

A leading underscore (`_attr`) signals "internal use, don't rely on this from outside." It's a convention, not enforced. Not imported by `from module import *`.

```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance  # "Protected" - internal use

    def deposit(self, amount):
        self._balance += amount

    def get_balance(self):
        return self._balance

acc = BankAccount(100)
acc.deposit(50)
print(acc.get_balance())
# Still accessible - it's a convention
print(acc._balance)
```

## Double Underscore: Name Mangling

A leading double underscore (`__attr`) triggers **name mangling**. Python renames it to `_ClassName__attr` to avoid accidental overrides in subclasses. It's not truly private—you can still access `_BankAccount__balance` if you really want.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Mangled to _BankAccount__balance

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = BankAccount(100)
acc.deposit(50)
print(acc.get_balance())
# print(acc.__balance)  # AttributeError!
# But mangled name exists:
print(acc._BankAccount__balance)
```

## Why Name Mangling?

Prevents accidental override in subclasses. If a subclass defines `__balance`, it won't clash with the parent's `__balance`—they become different attributes.

```python
class Parent:
    def __init__(self):
        self.__secret = "parent secret"

    def get_secret(self):
        return self.__secret

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.__secret = "child secret"  # Different attr! _Child__secret

c = Child()
print(c.get_secret())  # Uses Parent's __secret -> "parent secret"
print(c._Child__secret)  # Child's own
```

## Trailing Underscore

Use `class_` or `type_` when the name would otherwise be a keyword. Avoids `class` or `type`.

```python
def create_object(class_, name):
    return class_(name)

class Dog:
    def __init__(self, name):
        self.name = name

d = create_object(Dog, "Rex")
print(d.name)
```

## Double Leading and Trailing

Names like `__init__` have double underscores on both sides. These are **not** mangled—they're reserved for Python's special methods.

```python
class Example:
    def __init__(self):
        self.__value = 1   # Mangled: _Example__value
        # __init__ is NOT mangled - it's a special method

e = Example()
print(e._Example__value)
# __str__, __repr__, __len__ etc. - no mangling
```

## Practical Encapsulation Pattern

Use `_` for internal helpers, `__` sparingly when you need to avoid subclass collisions. Prefer properties for controlled access.

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius  # Internal storage

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        self._validate(value)
        self._celsius = value

    def _validate(self, value):  # Internal helper
        if value < -273.15:
            raise ValueError("Below absolute zero")

t = Temperature(25)
t.celsius = 30
print(t.celsius)
```

## Real-World: API Client

Hide implementation details; expose clean public interface.

```python
class APIClient:
    def __init__(self, base_url, api_key):
        self._base_url = base_url
        self._api_key = api_key
        self._session = None  # Lazy init

    def _get_session(self):
        if self._session is None:
            self._session = "session"  # Simplified
        return self._session

    def get(self, endpoint):
        session = self._get_session()
        return f"GET {self._base_url}/{endpoint}"

client = APIClient("https://api.example.com", "key123")
print(client.get("users"))
```

## Common Mistakes

- **Overusing `__`**: Makes debugging harder; use `_` for most internal attrs
- **Assuming `_` is private**: It's accessible; it's a convention for "please don't"
- **Mangling on special methods**: `__init__` is NOT mangled (double both sides)
- **Accessing mangled names from outside**: Possible but breaks encapsulation—avoid
- **Using `__` for "really private"**: Python philosophy: we're all consenting adults

## Key Takeaways

- `_attr`: convention for "internal use"; not imported by `import *`
- `__attr`: name mangling to `_ClassName__attr`; avoids subclass collisions
- `__special__`: no mangling; reserved for Python magic methods
- Prefer `_` for most cases; use `__` when subclass collision is a real concern
