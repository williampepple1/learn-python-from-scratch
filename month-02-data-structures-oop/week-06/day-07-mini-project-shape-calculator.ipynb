## Project Overview

We'll build a shape calculator with:
- **Shape** (abstract): defines `area()` and `perimeter()`
- **Rectangle**, **Circle**, **Triangle**: concrete shapes
- **ShapeCalculator**: computes total area and perimeter for a list of shapes

## Step 1: Create the Abstract Shape Class

Define the contract: all shapes must implement `area()` and `perimeter()`.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def __str__(self):
        return f"{self.__class__.__name__}(area={self.area():.2f}, perimeter={self.perimeter():.2f})"

# s = Shape()  # Can't instantiate abstract class
```

## Step 2: Implement Rectangle

Rectangle: width × height for area, 2×(w+h) for perimeter.

```python
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

r = Rectangle(4, 5)
print(r)
print(f"Area: {r.area()}, Perimeter: {r.perimeter()}")
```

## Step 3: Implement Circle

Circle: πr² for area, 2πr for perimeter.

```python
import math

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

c = Circle(5)
print(c)
print(f"Area: {c.area():.2f}, Perimeter: {c.perimeter():.2f}")
```

## Step 4: Implement Triangle

Triangle with sides a, b, c. Use Heron's formula for area. Perimeter = a + b + c.

```python
import math

class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

t = Triangle(3, 4, 5)
print(t)
print(f"Area: {t.area()}, Perimeter: {t.perimeter()}")
```

## Step 5: Create ShapeCalculator

Polymorphic calculator: works with any Shape.

```python
class ShapeCalculator:
    @staticmethod
    def total_area(shapes):
        return sum(s.area() for s in shapes)

    @staticmethod
    def total_perimeter(shapes):
        return sum(s.perimeter() for s in shapes)

    @staticmethod
    def summarize(shapes):
        for s in shapes:
            print(f"  {s}")
        print(f"Total area: {ShapeCalculator.total_area(shapes):.2f}")
        print(f"Total perimeter: {ShapeCalculator.total_perimeter(shapes):.2f}")
```

## Step 6: Add Square (Optional)

Square is a special Rectangle. Use inheritance.

```python
class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)

sq = Square(4)
print(sq)
print(f"Square area: {sq.area()}")
```

## Step 7: Complete System - Full Code

All classes together.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def __str__(self):
        return f"{self.__class__.__name__}(area={self.area():.2f}, perimeter={self.perimeter():.2f})"


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius


class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))


class ShapeCalculator:
    @staticmethod
    def total_area(shapes):
        return sum(s.area() for s in shapes)

    @staticmethod
    def total_perimeter(shapes):
        return sum(s.perimeter() for s in shapes)

    @staticmethod
    def summarize(shapes):
        print("Shapes:")
        for s in shapes:
            print(f"  {s}")
        print(f"Total area: {ShapeCalculator.total_area(shapes):.2f}")
        print(f"Total perimeter: {ShapeCalculator.total_perimeter(shapes):.2f}")
```

## Step 8: Run the Demo

Create a mix of shapes and compute totals.

```python
shapes = [
    Rectangle(4, 5),
    Circle(3),
    Triangle(3, 4, 5),
    Square(4),
]

ShapeCalculator.summarize(shapes)
```

```python
# Polymorphism in action: any Shape works
print("Total area only:", ShapeCalculator.total_area(shapes))
print("Total perimeter only:", ShapeCalculator.total_perimeter(shapes))
```
