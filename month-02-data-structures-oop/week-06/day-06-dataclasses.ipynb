## Why Dataclasses?

Dataclasses (Python 3.7+) reduce boilerplate for classes that mainly store data. You get `__init__`, `__repr__`, `__eq__` automatically.

## Basic `@dataclass`

Add `@dataclass` and list attributes with type hints. No need to write `__init__`.

```python
from dataclasses import dataclass

@dataclass
class Point:
    x: float
    y: float

p = Point(3.0, 4.0)
print(p)
print(p.x, p.y)
p2 = Point(3.0, 4.0)
print(p == p2)
```

## Default Values

Give default values. Fields with defaults must come after fields without.

```python
from dataclasses import dataclass

@dataclass
class Product:
    name: str
    price: float
    quantity: int = 0

p1 = Product("Widget", 9.99)
p2 = Product("Gadget", 19.99, 10)
print(p1, p2)
```

## `field()` for Customization

Use `field()` for default_factory, repr=False, compare=False, etc.

```python
from dataclasses import dataclass, field

@dataclass
class Student:
    name: str
    grades: list = field(default_factory=list)  # Mutable default!

s1 = Student("Alice")
s2 = Student("Bob")
s1.grades.append(90)
print(s2.grades)  # [] - not shared

@dataclass
class Config:
    host: str = "localhost"
    debug: bool = field(default=False, repr=False)  # Exclude from __repr__

c = Config(debug=True)
print(c)
```

## Frozen Dataclasses

`frozen=True` makes instances immutable. Good for hashable, constant data.

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class ImmutablePoint:
    x: float
    y: float

p = ImmutablePoint(1.0, 2.0)
print(p)
# p.x = 5  # FrozenInstanceError

# Frozen + all fields comparable -> hashable
s = {ImmutablePoint(1, 2), ImmutablePoint(1, 2)}
print(len(s))
```

## NamedTuple

`NamedTuple` is lighter than dataclass. Immutable by default. Access by name or index.

```python
from typing import NamedTuple

class Point(NamedTuple):
    x: float
    y: float

p = Point(3.0, 4.0)
print(p.x, p[0])
print(p)
# p.x = 5  # AttributeError - immutable

class RGB(NamedTuple):
    red: int
    green: int
    blue: int

color = RGB(255, 0, 0)
print(color.red, color[1])
```

## `__slots__`

`__slots__` restricts attributes to a fixed set. Saves memory (no `__dict__`) and prevents typos. Dataclasses support it with `@dataclass(slots=True)` in Python 3.10+.

```python
class SlottedPerson:
    __slots__ = ('name', 'age')

    def __init__(self, name, age):
        self.name = name
        self.age = age

p = SlottedPerson("Alice", 30)
print(p.name)
# p.other = 1  # AttributeError
# print(p.__dict__)  # AttributeError - no __dict__
```

```python
# Python 3.10+ dataclass with slots
from dataclasses import dataclass

try:
    @dataclass(slots=True)
    class SlotPoint:
        x: float
        y: float

    sp = SlotPoint(1.0, 2.0)
    print(sp)
except TypeError:
    print("slots=True requires Python 3.10+")
```

## Dataclass with Methods

Add methods like any class. Use `field(init=False)` for computed or internal fields.

```python
from dataclasses import dataclass, field

@dataclass
class Rectangle:
    width: float
    height: float

    @property
    def area(self):
        return self.width * self.height

    def scale(self, factor):
        return Rectangle(self.width * factor, self.height * factor)

r = Rectangle(4, 5)
print(r.area)
print(r.scale(2))
```

## Common Mistakes

- **Mutable default**: Use `field(default_factory=list)` not `grades: list = []`
- **Field order**: No-default fields before default fields
- **Frozen + mutation**: Can't change; create new instance instead
- **NamedTuple vs dataclass**: NamedTuple for simple immutable; dataclass for mutable or more features
- **__slots__ + inheritance**: Child's __slots__ adds to parent's

## Key Takeaways

- `@dataclass`: auto `__init__`, `__repr__`, `__eq__`
- `field(default_factory=...)` for mutable defaults
- `frozen=True` for immutable, hashable instances
- `NamedTuple`: lightweight, immutable, indexable
- `__slots__` / `slots=True`: fixed attributes, less memory
