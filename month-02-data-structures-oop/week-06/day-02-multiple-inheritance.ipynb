## Multiple Inheritance Basics

A class can inherit from more than one parent: `class C(A, B):`. Python looks up methods in a specific order—the **Method Resolution Order (MRO)**.

```python
class Flyer:
    def fly(self):
        return "Flying"

class Swimmer:
    def swim(self):
        return "Swimming"

class Duck(Flyer, Swimmer):
    pass

d = Duck()
print(d.fly())
print(d.swim())
```

## MRO: Method Resolution Order

Use `ClassName.__mro__` or `mro()` to see the lookup order. Python uses C3 linearization to compute it.

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

print(D.__mro__)
# D -> B -> C -> A -> object
```

## The Diamond Problem

When both B and C inherit from A, and D inherits from both B and C, we get a diamond. MRO ensures each class is visited only once.

```python
class A:
    def method(self):
        return "A"

class B(A):
    def method(self):
        return "B"

class C(A):
    def method(self):
        return "C"

class D(B, C):
    pass

d = D()
print(d.method())  # B - first in MRO wins
print(D.__mro__)
```

## Using super() with Multiple Inheritance

`super()` follows MRO. Each class's `super()` calls the *next* class in MRO, not necessarily the "parent" you wrote.

```python
class A:
    def __init__(self):
        print("A")

class B(A):
    def __init__(self):
        print("B")
        super().__init__()

class C(A):
    def __init__(self):
        print("C")
        super().__init__()

class D(B, C):
    def __init__(self):
        print("D")
        super().__init__()

d = D()
# D -> B -> C -> A (each super() goes to next in MRO)
```

## C3 Linearization Rules

1. Child before parent
2. Order of bases is preserved (left to right)
3. No class appears more than once

```python
class X:
    pass

class Y:
    pass

class Z:
    pass

class A(X, Y):
    pass

class B(Y, Z):
    pass

class M(A, B):
    pass

print([c.__name__ for c in M.__mro__])
```

## Mixin Pattern

Small classes that add specific behavior. Often used with multiple inheritance. Put mixins after the main base.

```python
class JsonMixin:
    def to_json(self):
        import json
        return json.dumps(self.__dict__)

class Person:
    def __init__(self, name):
        self.name = name

class JsonPerson(Person, JsonMixin):
    pass

p = JsonPerson("Alice")
print(p.to_json())
```

## Common Mistakes

- **Confusing MRO**: Use `__mro__` to debug
- **Calling parent explicitly**: `A.method(self)` can skip classes; prefer `super()`
- **Inconsistent super()**: All classes in the chain should use super() for cooperative multiple inheritance
- **Circular inheritance**: Python will error
- **Too many bases**: Hard to reason about; prefer composition
