## Mixin Classes

Mixins are small classes that add specific behavior. They're not meant to stand alone—you combine them with other classes via multiple inheritance. Name them with a `Mixin` suffix.

```python
class JsonMixin:
    def to_json(self):
        import json
        return json.dumps(self.__dict__, indent=2)

class ReprMixin:
    def __repr__(self):
        return f"{self.__class__.__name__}({self.__dict__})"

class Person(JsonMixin, ReprMixin):
    def __init__(self, name, age):
        self.name = name
        self.age = age

p = Person("Alice", 30)
print(p)
print(p.to_json())
```

## Is-a vs Has-a

- **Is-a**: Inheritance. A Dog *is* an Animal. Use when the child is a specialized version of the parent.
- **Has-a**: Composition. A Car *has* an Engine. Use when one object contains another.

```python
# Is-a: Dog IS AN Animal
class Animal:
    def breathe(self):
        return "Breathing"

class Dog(Animal):
    def bark(self):
        return "Woof"

# Has-a: Car HAS AN Engine
class Engine:
    def start(self):
        return "Engine started"

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition

    def start(self):
        return self.engine.start()

c = Car()
print(c.start())
```

## Composition Over Inheritance

Favor composition when behavior can be swapped or combined flexibly. Inheritance creates tight coupling; composition is more modular.

```python
class Logger:
    def log(self, msg):
        print(f"LOG: {msg}")

class FileLogger:
    def __init__(self, filename):
        self.filename = filename

    def log(self, msg):
        with open(self.filename, 'a') as f:
            f.write(f"{msg}\n")

class UserService:
    def __init__(self, logger=None):
        self.logger = logger or Logger()  # Composition: inject dependency

    def create_user(self, name):
        self.logger.log(f"Creating user {name}")
        return {"name": name}

service = UserService(Logger())
service.create_user("Alice")

# Easy to swap: UserService(FileLogger('app.log'))
```

## Mixin: Comparable

Add comparison by a key. Reusable across many classes.

```python
class ComparableMixin:
    def __lt__(self, other):
        return self._compare_key() < other._compare_key()

    def _compare_key(self):
        raise NotImplementedError

class Student(ComparableMixin):
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def _compare_key(self):
        return self.grade

students = [Student("Bob", 85), Student("Alice", 92), Student("Carol", 78)]
for s in sorted(students):
    print(s.name, s.grade)
```

## Composition: Delegation

A class holds another object and delegates work to it. The contained object can be swapped.

```python
class Stack:
    def __init__(self):
        self._items = []  # Composition: Stack HAS-A list

    def push(self, item):
        self._items.append(item)

    def pop(self):
        return self._items.pop()

    def __len__(self):
        return len(self._items)

s = Stack()
s.push(1)
s.push(2)
print(s.pop(), len(s))
```

## When to Use What

- **Inheritance**: True is-a, shared interface, template method pattern
- **Mixin**: Add orthogonal behavior (JSON, repr, comparison) without deep hierarchy
- **Composition**: Has-a, dependency injection, strategy pattern

```python
class TimestampMixin:
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        from datetime import datetime
        self.created_at = datetime.now()

class Person(TimestampMixin):
    def __init__(self, name):
        super().__init__()
        self.name = name

p = Person("Bob")
print(p.name, p.created_at)
```

## Common Mistakes

- **Mixin with state and complex __init__**: Mixins should be simple; use super() carefully
- **Inheritance for has-a**: Use composition (Car has Engine, not Car extends Engine)
- **Deep inheritance**: Prefer composition + mixins for flexibility
- **Mixin depending on specific attributes**: Document what the mixin expects
- **Composition without interface**: Use duck typing or ABC for the composed part

## Key Takeaways

- Mixins: small, focused classes adding behavior; use multiple inheritance
- Is-a: inheritance (Dog is Animal)
- Has-a: composition (Car has Engine)
- Composition over inheritance for flexibility and testability
- Use super() in mixins for cooperative multiple inheritance
