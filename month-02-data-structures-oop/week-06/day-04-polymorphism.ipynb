## What is Polymorphism?

Polymorphism means "many forms." Different objects can be used interchangeably if they support the same operations. In Python, we use **duck typing**: "If it walks like a duck and quacks like a duck, it's a duck."

## Duck Typing

We don't check types. We use the object and assume it has the methods we need. If it doesn't, we get an error at runtime.

```python
def describe(thing):
    return thing.describe()  # We assume thing has describe()

class Dog:
    def describe(self):
        return "A loyal dog"

class Car:
    def describe(self):
        return "A fast car"

class Book:
    def describe(self):
        return "An interesting book"

for obj in [Dog(), Car(), Book()]:
    print(describe(obj))
```

## Polymorphic Functions

Functions that work with any object that supports the required protocol. Example: anything with `__len__` works with `len()`.

```python
def double_length(container):
    """Works with any object that supports len()."""
    return len(container) * 2

print(double_length([1, 2, 3]))
print(double_length("hello"))
print(double_length({1, 2, 3, 4}))

class MyContainer:
    def __len__(self):
        return 5

print(double_length(MyContainer()))
```

## Iterable Protocol

Anything with `__iter__` or `__getitem__` can be iterated. `for x in obj` works polymorphically.

```python
def sum_all(iterable):
    return sum(iterable)

print(sum_all([1, 2, 3]))
print(sum_all((4, 5, 6)))
print(sum_all({1, 2, 3}))
print(sum_all(range(5)))
```

## EAFP vs LBYL

- **EAFP**: Easier to Ask Forgiveness than Permission — try it, catch exceptions
- **LBYL**: Look Before You Leap — check first, then do

```python
# LBYL - check first
def get_value_lbyl(obj, key):
    if hasattr(obj, '__getitem__'):
        if key in obj:
            return obj[key]
    return None

# EAFP - try and catch
def get_value_eafp(obj, key):
    try:
        return obj[key]
    except (KeyError, TypeError):
        return None

d = {'a': 1, 'b': 2}
print(get_value_lbyl(d, 'a'))
print(get_value_eafp(d, 'a'))
print(get_value_eafp(d, 'z'))
```

## Pythonic EAFP

Python favors EAFP: exceptions are cheap, and code is cleaner. Use try/except for expected "might fail" cases.

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None
    except TypeError:
        return None

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("a", 2))
```

## Polymorphic Area Calculator

Any shape with `.area()` works—no inheritance required.

```python
def total_area(shapes):
    total = 0
    for s in shapes:
        total += s.area()  # Duck typing: we assume .area() exists
    return total

class Rectangle:
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h

class Circle:
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14159 * self.r ** 2

shapes = [Rectangle(3, 4), Circle(5), Rectangle(2, 2)]
print(total_area(shapes))
```

## hasattr and getattr

When you need to check: `hasattr(obj, 'method')` and `getattr(obj, 'method', default)`.

```python
def call_if_exists(obj, method_name, *args, default=None):
    method = getattr(obj, method_name, None)
    if callable(method):
        return method(*args)
    return default

class A:
    def greet(self):
        return "Hello"

class B:
    pass

print(call_if_exists(A(), 'greet'))
print(call_if_exists(B(), 'greet', default='No greet'))
```

## Common Mistakes

- **Over-checking types**: Prefer duck typing; avoid `if type(x) == list`
- **LBYL for rare cases**: EAFP is often cleaner
- **Catching too broad**: `except Exception` can hide bugs; catch specific exceptions
- **Assuming inheritance**: Polymorphism works without shared base class
- **hasattr for callable**: Use `callable(getattr(obj, 'm', None))` if you need to call it

## Key Takeaways

- Duck typing: use objects by behavior, not type
- Polymorphic functions work with any object supporting the protocol
- EAFP: try/except; LBYL: check first. Python favors EAFP
