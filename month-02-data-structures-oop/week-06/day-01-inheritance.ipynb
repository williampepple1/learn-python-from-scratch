## What is Inheritance?

Inheritance lets a class (child) get attributes and methods from another class (parent). The child can add new behavior or override parent behavior. Promotes code reuse.

## Single Inheritance

Define a child class by passing the parent in parentheses: `class Child(Parent):`.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "..."

class Dog(Animal):
    pass  # Inherits everything from Animal

d = Dog("Rex")
print(d.name)
print(d.speak())
```

## Using `super()`

`super()` gives access to the parent class. Call `super().__init__(args)` to run the parent's constructor. Essential when the parent needs to initialize.

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Call parent __init__
        self.breed = breed

d = Dog("Rex", "Labrador")
print(d.name, d.breed)
```

## Method Overriding

Define a method with the same name in the child to override the parent's version. The child's method is used when called on a child instance.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

animals = [Dog("Rex"), Cat("Whiskers"), Animal("Generic")]
for a in animals:
    print(f"{a.name}: {a.speak()}")
```

## Override and Extend with super()

You can call the parent's method from the child to extend behavior rather than replace it.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_info(self):
        return f"{self.name}: ${self.salary}"

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def get_info(self):
        base = super().get_info()
        return f"{base} (Manager, {self.department})"

m = Manager("Alice", 80000, "Engineering")
print(m.get_info())
```

## `isinstance` and `issubclass`

- `isinstance(obj, Class)`: True if obj is an instance of Class or a subclass
- `issubclass(Child, Parent)`: True if Child is a subclass of Parent

```python
class Animal:
    pass

class Dog(Animal):
    pass

d = Dog()
print(isinstance(d, Dog))
print(isinstance(d, Animal))
print(isinstance(d, object))
print(issubclass(Dog, Animal))
print(issubclass(Dog, object))
```

## Practical Use: Type Checking

Use `isinstance` for flexible type checks that respect inheritance.

```python
class Shape:
    def area(self):
        raise NotImplementedError

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

def total_area(shapes):
    total = 0
    for s in shapes:
        if isinstance(s, Shape):
            total += s.area()
    return total

shapes = [Rectangle(3, 4), Rectangle(2, 5)]
print(total_area(shapes))
```

## Common Mistakes

- **Forgetting `super().__init__()`**: Parent attributes never set
- **Wrong argument order in super()**: Match parent's signature
- **Using `type(x) == Class`**: Use `isinstance(x, Class)` for subclasses
- **Overriding without calling super**: Sometimes you need parent logic
- **Multiple inheritance order**: `super()` follows MRO (next lesson)

## Key Takeaways

- `class Child(Parent):` for single inheritance
- `super().__init__(args)` to call parent constructor
- Override by defining same method name in child
- `super().method()` to extend parent behavior
- `isinstance(obj, Class)` and `issubclass(Child, Parent)` for type checks
