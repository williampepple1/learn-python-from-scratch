## What Are Abstract Classes?

Abstract classes define a contract: subclasses *must* implement certain methods. You can't instantiate an abstract class directly. Use them to enforce interfaces.

## The `abc` Module

Import `ABC` (Abstract Base Class) and `abstractmethod` from `abc`. Inherit from `ABC` and decorate methods with `@abstractmethod`.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# s = Shape()  # TypeError: Can't instantiate abstract class

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

r = Rectangle(4, 5)
print(r.area(), r.perimeter())
```

## Forgetting to Implement

If a subclass doesn't implement all abstract methods, it stays abstract and can't be instantiated.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class IncompleteAnimal(Animal):
    pass  # Forgot speak()

d = Dog()
print(d.speak())
# i = IncompleteAnimal()  # TypeError!
```

## Abstract Properties

Use `@abstractmethod` with `@property` for required attributes.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @property
    @abstractmethod
    def wheels(self):
        pass

class Car(Vehicle):
    @property
    def wheels(self):
        return 4

class Bike(Vehicle):
    @property
    def wheels(self):
        return 2

c = Car()
print(c.wheels)
b = Bike()
print(b.wheels)
```

## Concrete + Abstract Methods

Abstract classes can have concrete (implemented) methods too. Subclasses inherit them.

```python
from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
    @abstractmethod
    def process(self, amount):
        pass

    def validate_amount(self, amount):
        return amount > 0 and isinstance(amount, (int, float))

class CreditCardPayment(PaymentProcessor):
    def process(self, amount):
        if not self.validate_amount(amount):
            raise ValueError("Invalid amount")
        return f"Charged ${amount} to card"

p = CreditCardPayment()
print(p.process(99.99))
```

## Interface Design Pattern

Define minimal abstract methods. Keep the interface small and focused.

```python
from abc import ABC, abstractmethod

class Drawable(ABC):
    @abstractmethod
    def draw(self):
        pass

class Circle(Drawable):
    def __init__(self, radius):
        self.radius = radius

    def draw(self):
        return f"Drawing circle r={self.radius}"

class Square(Drawable):
    def __init__(self, side):
        self.side = side

    def draw(self):
        return f"Drawing square s={self.side}"

def render(items):
    for item in items:
        if isinstance(item, Drawable):
            print(item.draw())

render([Circle(5), Square(3)])
```

## Common Mistakes

- **Forgetting to inherit ABC**: Old style used `metaclass=ABCMeta`
- **Implementing abstract method but still abstract**: If subclass has unimplemented abstract methods, it stays abstract
- **Wrong decorator order**: `@property` then `@abstractmethod` (property above)
- **Over-abstracting**: Don't make everything abstract; only what subclasses must implement
- **Abstract __init__**: Rare; usually concrete
