## Stack: LIFO

Stack = Last In, First Out. Push adds to top; pop removes from top. Use list's `append` and `pop`.

```python
stack = []
stack.append(1)
stack.append(2)
stack.append(3)
print(stack.pop())
print(stack.pop())
print(stack)
```

## Stack Class (List-Based)

Wrap list operations in a class for clarity and encapsulation.

```python
class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError('pop from empty stack')
        return self._items.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError('peek on empty stack')
        return self._items[-1]

    def is_empty(self):
        return len(self._items) == 0

    def __len__(self):
        return len(self._items)

s = Stack()
s.push(10)
s.push(20)
print(s.peek())
print(s.pop())
print(len(s))
```

## Stack with Linked List

Use head as top: push = insert at beginning, pop = remove from beginning.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class StackLL:
    def __init__(self):
        self.head = None

    def push(self, data):
        new = Node(data)
        new.next = self.head
        self.head = new

    def pop(self):
        if self.head is None:
            raise IndexError('pop from empty stack')
        data = self.head.data
        self.head = self.head.next
        return data

    def is_empty(self):
        return self.head is None

s = StackLL()
s.push(1)
s.push(2)
print(s.pop())
print(s.pop())
print(s.is_empty())
```

## Queue: FIFO

Queue = First In, First Out. Enqueue adds at rear; dequeue removes from front. Using list: append for enqueue, pop(0) for dequeue (but pop(0) is O(n)).

```python
from collections import deque

q = deque()
q.append(1)
q.append(2)
q.append(3)
print(q.popleft())
print(q.popleft())
print(list(q))
```

## Queue Class (deque-Based)

Use `deque` for O(1) enqueue and dequeue.

```python
from collections import deque

class Queue:
    def __init__(self):
        self._items = deque()

    def enqueue(self, item):
        self._items.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError('dequeue from empty queue')
        return self._items.popleft()

    def is_empty(self):
        return len(self._items) == 0

    def __len__(self):
        return len(self._items)

q = Queue()
q.enqueue('first')
q.enqueue('second')
print(q.dequeue())
print(q.dequeue())
```

## Queue with Linked List

Maintain `head` (front) and `tail` (rear). Enqueue at tail; dequeue from head.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class QueueLL:
    def __init__(self):
        self.head = None
        self.tail = None

    def enqueue(self, data):
        new = Node(data)
        if self.tail:
            self.tail.next = new
        self.tail = new
        if self.head is None:
            self.head = new

    def dequeue(self):
        if self.head is None:
            raise IndexError('dequeue from empty queue')
        data = self.head.data
        self.head = self.head.next
        if self.head is None:
            self.tail = None
        return data

    def is_empty(self):
        return self.head is None

q = QueueLL()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue(), q.dequeue(), q.dequeue())
```

## Practical: Balanced Parentheses with Stack

Use a stack to check if parentheses are balanced.

```python
def is_balanced(s):
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}
    for c in s:
        if c in '([{':
            stack.append(c)
        elif c in ')]}':
            if not stack or stack[-1] != pairs[c]:
                return False
            stack.pop()
    return len(stack) == 0

print(is_balanced('([])'))
print(is_balanced('([)]'))
```

## Practical: BFS-like processing with Queue

Process items in FIFO order (e.g., level-order traversal).

```python
from collections import deque

def process_in_order(items):
    q = deque(items)
    result = []
    while q:
        item = q.popleft()
        result.append(item)
    return result

print(process_in_order(['a', 'b', 'c']))
```

## Common Mistakes

- **Using list for queue**: `pop(0)` is O(n); use `deque` for O(1)
- **Stack pop from wrong end**: Use `pop()` (end), not `pop(0)` (beginning)
- **Queue tail update**: When dequeuing the last item, set tail = None
- **Peek on empty**: Always check is_empty before peek/pop

```python
# WRONG: queue with list - pop(0) is slow
# q = []; q.append(1); q.pop(0)  # O(n)!

# CORRECT: use deque
from collections import deque
q = deque()
q.append(1)
q.popleft()  # O(1)
```

## Key Takeaways

1. Stack: LIFO; use list append/pop or linked list with head as top
2. Queue: FIFO; use deque for O(1) enqueue/dequeue
3. List pop(0) is O(n); avoid for queues
4. Stack LL: push/pop at head; Queue LL: enqueue at tail, dequeue from head
5. Balanced parentheses: push open brackets, pop on close and match
