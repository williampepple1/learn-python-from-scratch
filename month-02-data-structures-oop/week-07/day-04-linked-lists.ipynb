## What is a Linked List?

A linked list is a linear data structure where each element (node) contains data and a reference to the next node. Unlike arrays, elements are not stored contiguously.

## Node Class

Each node holds `data` and `next` (pointer to the next node). The last node's `next` is `None`.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n1.next = n2
n2.next = n3

current = n1
while current:
    print(current.data, end=' -> ')
    current = current.next
print('None')
```

## LinkedList Class - Basic Structure

A LinkedList wraps the chain of nodes with a `head` pointer. Empty list has `head = None`.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

ll = LinkedList()
print("Empty:", ll.is_empty())
ll.head = Node(1)
print("After adding head:", ll.is_empty())
```

## Insert at Beginning

Create new node, set its `next` to current head, then update head.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

ll = LinkedList()
ll.insert_at_beginning(3)
ll.insert_at_beginning(2)
ll.insert_at_beginning(1)
print(ll.head.data, ll.head.next.data, ll.head.next.next.data)
```

## Insert at End

Traverse to the last node, then set its `next` to the new node.

```python
def insert_at_end(self, data):
    new_node = Node(data)
    if self.head is None:
        self.head = new_node
        return
    current = self.head
    while current.next:
        current = current.next
    current.next = new_node

# Add to LinkedList class and test
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

ll = LinkedList()
ll.insert_at_end(1)
ll.insert_at_end(2)
ll.insert_at_end(3)
print(ll.head.data, ll.head.next.data, ll.head.next.next.data)
```

## Delete by Value

Find the node with the value. If it's the head, update head. Otherwise, update the previous node's `next`.

```python
def delete(self, key):
    if self.head is None:
        return
    if self.head.data == key:
        self.head = self.head.next
        return
    current = self.head
    while current.next:
        if current.next.data == key:
            current.next = current.next.next
            return
        current = current.next

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def delete(self, key):
        if self.head is None:
            return
        if self.head.data == key:
            self.head = self.head.next
            return
        current = self.head
        while current.next:
            if current.next.data == key:
                current.next = current.next.next
                return
            current = current.next

    def display(self):
        cur = self.head
        while cur:
            print(cur.data, end=' -> ')
            cur = cur.next
        print('None')

ll = LinkedList()
for x in [1, 2, 3, 4]:
    ll.insert_at_end(x)
ll.delete(3)
ll.display()
```

## Complete LinkedList Example

Full implementation with all operations.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new = Node(data)
        new.next = self.head
        self.head = new

    def insert_at_end(self, data):
        new = Node(data)
        if not self.head:
            self.head = new
            return
        cur = self.head
        while cur.next:
            cur = cur.next
        cur.next = new

    def search(self, key):
        cur = self.head
        while cur:
            if cur.data == key:
                return True
            cur = cur.next
        return False

    def delete(self, key):
        if not self.head:
            return
        if self.head.data == key:
            self.head = self.head.next
            return
        cur = self.head
        while cur.next:
            if cur.next.data == key:
                cur.next = cur.next.next
                return
            cur = cur.next

    def display(self):
        cur = self.head
        while cur:
            print(cur.data, end=' -> ')
            cur = cur.next
        print('None')

ll = LinkedList()
ll.insert_at_end(10)
ll.insert_at_end(20)
ll.insert_at_beginning(5)
ll.display()
print("Search 20:", ll.search(20))
ll.delete(10)
ll.display()
```

## Key Takeaways

1. Node has `data` and `next`; last node's next is None
2. LinkedList keeps a `head`; empty list has head = None
3. Insert at beginning: new.next = head; head = new
4. Insert at end: traverse to last, set last.next = new
5. Delete: special case for head; otherwise update prev.next to skip the node
6. Search: traverse until key found or end
