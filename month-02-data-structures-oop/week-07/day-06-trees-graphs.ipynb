## Binary Tree Node

Each node has `data`, `left`, and `right` children. Leaves have `left = right = None`.

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
print(root.data, root.left.data, root.right.data)
```

## Preorder Traversal (Root, Left, Right)

Visit root first, then left subtree, then right subtree.

```python
def preorder(node):
    if node is None:
        return
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
print("Preorder:", end=' ')
preorder(root)
```

## Inorder Traversal (Left, Root, Right)

Visit left subtree, then root, then right. For BST, gives sorted order.

```python
def inorder(node):
    if node is None:
        return
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
print("Inorder:", end=' ')
inorder(root)
```

## Postorder Traversal (Left, Right, Root)

Visit left, then right, then root. Useful for deleting nodes (children before parent).

```python
def postorder(node):
    if node is None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
print("Postorder:", end=' ')
postorder(root)
```

## Level-Order (BFS) with Queue

Process nodes level by level using a queue.

```python
from collections import deque

def level_order(root):
    if root is None:
        return
    q = deque([root])
    while q:
        node = q.popleft()
        print(node.data, end=' ')
        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
print("Level-order:", end=' ')
level_order(root)
```

## Graph: Adjacency List

A graph is a set of vertices and edges. Adjacency list: each vertex maps to a list of neighbors.

```python
class Graph:
    def __init__(self):
        self.adj = {}  # vertex -> list of neighbors

    def add_edge(self, u, v):
        if u not in self.adj:
            self.adj[u] = []
        if v not in self.adj:
            self.adj[v] = []
        self.adj[u].append(v)
        self.adj[v].append(u)  # undirected

g = Graph()
g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('B', 'C')
print(g.adj)
```

## DFS on Graph

Depth-first search: go deep, use recursion or stack. Track visited to avoid cycles.

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')
    for neighbor in graph.adj.get(start, []):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    return visited

class Graph:
    def __init__(self):
        self.adj = {}
    def add_edge(self, u, v):
        if u not in self.adj: self.adj[u] = []
        if v not in self.adj: self.adj[v] = []
        self.adj[u].append(v)
        self.adj[v].append(u)

g = Graph()
g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('B', 'D')
g.add_edge('C', 'D')
print("DFS from A:", end=' ')
dfs(g, 'A')
```

## BFS on Graph

Breadth-first search: level by level using a queue.

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    q = deque([start])
    visited.add(start)
    while q:
        node = q.popleft()
        print(node, end=' ')
        for neighbor in graph.adj.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                q.append(neighbor)

class Graph:
    def __init__(self):
        self.adj = {}
    def add_edge(self, u, v):
        if u not in self.adj: self.adj[u] = []
        if v not in self.adj: self.adj[v] = []
        self.adj[u].append(v)
        self.adj[v].append(u)

g = Graph()
g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('B', 'D')
g.add_edge('C', 'D')
print("BFS from A:", end=' ')
bfs(g, 'A')
```

## Tree Height

Height = 1 + max(height(left), height(right)). Base case: None has height -1 or 0.

```python
def height(node):
    if node is None:
        return -1
    return 1 + max(height(node.left), height(node.right))

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
print("Height:", height(root))
```

## Common Mistakes

- **Forgetting None check**: Always check `if node is None` before accessing children
- **Graph cycles**: Use `visited` set in DFS/BFS to avoid infinite loops
- **Directed vs undirected**: add_edge(u,v) for undirected adds both directions; for directed, add only u->v
- **Empty graph**: Check if start vertex exists in adj before iterating

```python
# WRONG: No visited set in graph DFS -> infinite loop on cycles
# def dfs_bad(g, start):
#     print(start)
#     for n in g.adj[start]: dfs_bad(g, n)  # Loops forever!

# CORRECT: Track visited
def dfs_ok(g, start, visited=None):
    visited = visited or set()
    if start in visited:
        return
    visited.add(start)
    print(start)
    for n in g.adj.get(start, []):
        dfs_ok(g, n, visited)
```
