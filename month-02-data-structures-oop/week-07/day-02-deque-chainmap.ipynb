## deque (Double-Ended Queue)

`deque` supports O(1) append and pop from both ends. Lists are O(n) for pop(0) at the front.

```python
from collections import deque

d = deque([1, 2, 3])
d.append(4)
d.appendleft(0)
print(d)
print(d.pop())
print(d.popleft())
print(d)
```

## deque: extend and rotate

`extend`/`extendleft` add multiple items. `rotate(n)` shifts elements; positive n rotates right.

```python
from collections import deque

d = deque([1, 2, 3, 4, 5])
d.rotate(2)
print("Rotate right 2:", list(d))

d.rotate(-2)
print("Rotate left 2:", list(d))
```

## deque: maxlen for bounded queue

Set `maxlen` to create a sliding window; when full, adding from one end drops from the other.

```python
from collections import deque

recent = deque(maxlen=3)
for i in range(5):
    recent.append(i)
    print(list(recent))
```

## ChainMap

`ChainMap` groups multiple dicts into a single view. Lookups search each dict in order until found.

```python
from collections import ChainMap

defaults = {'theme': 'dark', 'language': 'en'}
user_prefs = {'theme': 'light'}
config = ChainMap(user_prefs, defaults)

print(config['theme'])   # From user_prefs
print(config['language'])  # From defaults
print(dict(config))
```

## ChainMap: new_child and parents

`new_child(m)` adds a new dict at the front. `parents` skips the first mapping.

```python
from collections import ChainMap

base = {'a': 1, 'b': 2}
cm = ChainMap({'a': 10}, base)
print(cm['a'])

cm2 = cm.new_child({'a': 100})
print(cm2['a'])
print(cm2.parents['a'])
```

## namedtuple

`namedtuple` creates tuple subclasses with named fields. Access by name or index.

```python
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print(p.x, p.y)
print(p[0], p[1])
print(p)
```

## namedtuple: _replace and _asdict

`_replace()` returns a new tuple with updated fields. `_asdict()` converts to dict.

```python
from collections import namedtuple

Point = namedtuple('Point', ['x', 'y'])
p = Point(1, 2)
p2 = p._replace(x=10)
print(p, p2)
print(p2._asdict())
```

## Practical: Sliding window with deque

Use deque with maxlen for a moving average.

```python
from collections import deque

def moving_average(iterable, n=3):
    it = iter(iterable)
    window = deque(maxlen=n)
    for _ in range(n - 1):
        window.append(next(it))
    for x in it:
        window.append(x)
        yield sum(window) / n

data = [10, 20, 30, 40, 50, 60]
print(list(moving_average(data)))
```

## Practical: Configuration with ChainMap

Override defaults with user and environment settings.

```python
from collections import ChainMap

default_config = {'debug': False, 'port': 8080}
env_config = {'port': 3000}
config = ChainMap(env_config, default_config)
print(f"Port: {config['port']}, Debug: {config['debug']}")
```

## Common Mistakes

- **deque from string**: `deque('abc')` gives `['a','b','c']`; use `deque(['abc'])` for one element
- **ChainMap updates**: Changing `config['key']` updates the first mapping that has the key; new keys go to the first mapping
- **namedtuple immutability**: Can't do `p.x = 5`; use `_replace()`
- **deque index**: `d[0]` works but `d[1] = 9` is O(n) in the middle

```python
# WRONG: p.x = 5  # AttributeError: can't set attribute
# CORRECT:
from collections import namedtuple
Point = namedtuple('Point', ['x', 'y'])
p = Point(1, 2)
p = p._replace(x=5)
print(p)
```

## Key Takeaways

1. `deque` gives O(1) append/pop at both ends; use `maxlen` for sliding windows
2. `ChainMap` layers dicts; first match wins; good for config overrides
3. `namedtuple` gives named, immutable tuples; use `_replace()` for "updates"
4. `deque.rotate(n)` shifts elements; positive = right, negative = left
5. ChainMap lookups search mappings in order; writes go to the first mapping
