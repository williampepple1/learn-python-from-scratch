## Introduction to collections

The `collections` module provides specialized container datatypes beyond the built-in `list`, `dict`, and `tuple`. These are optimized for specific use cases.

```python
import collections
print([x for x in dir(collections) if not x.startswith('_')])
```

## Counter

`Counter` counts hashable objects. It's a dict subclass where elements are stored as keys and their counts as values.

```python
from collections import Counter

words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
counts = Counter(words)
print(counts)
print(counts['apple'])
print(counts['mango'])  # Missing key returns 0, no KeyError
```

## Counter: most_common and update

`most_common(n)` returns the n most frequent elements. `update()` adds counts from another iterable.

```python
from collections import Counter

text = "the quick brown fox jumps over the lazy dog the fox"
word_counts = Counter(text.split())
print("Top 3 words:", word_counts.most_common(3))

word_counts.update(['the', 'the', 'the'])  # Add more occurrences
print("After update:", word_counts['the'])
```

## defaultdict

`defaultdict` is a dict subclass that provides a default value for missing keys. You pass a callable (e.g., `list`, `int`) that returns the default.

```python
from collections import defaultdict

# Group words by first letter
words = ['apple', 'ant', 'banana', 'berry', 'cherry']
by_letter = defaultdict(list)
for w in words:
    by_letter[w[0]].append(w)

print(dict(by_letter))
```

## defaultdict with int

Using `int` as default (returns 0) is perfect for counting.

```python
from collections import defaultdict

dd = defaultdict(int)
dd['a'] += 1
dd['a'] += 1
dd['b'] += 1
print(dict(dd))
```

## OrderedDict

`OrderedDict` remembers the order in which keys were inserted. In Python 3.7+, regular `dict` also preserves order, but `OrderedDict` has extra methods like `move_to_end()`.

```python
from collections import OrderedDict

od = OrderedDict()
od['z'] = 1
od['a'] = 2
od['m'] = 3
print(list(od.keys()))

od.move_to_end('a')  # Move 'a' to end
print(list(od.keys()))

od.move_to_end('z', last=False)  # Move 'z' to beginning
print(list(od.keys()))
```

## Practical: Word frequency analysis

Real-world use: analyze word frequency in text.

```python
from collections import Counter

def word_frequency(text, top_n=5):
    words = text.lower().replace('.', '').replace(',', '').split()
    return Counter(words).most_common(top_n)

sample = "Python is great. Python is easy. Python is powerful. Learn Python today."
print(word_frequency(sample))
```

## Practical: Grouping with defaultdict

Group items by a key without checking if the key exists.

```python
from collections import defaultdict

students = [('Alice', 'A'), ('Bob', 'B'), ('Charlie', 'A'), ('Diana', 'B')]
by_grade = defaultdict(list)
for name, grade in students:
    by_grade[grade].append(name)

print(dict(by_grade))
```

## Common Mistakes

- **Counter with non-hashable keys**: Counter only works with hashable objects (no lists as keys)
- **defaultdict(None)**: Pass a callable, not None; use `lambda: value` for custom defaults
- **Assuming dict order in older Python**: In Python 3.6 and earlier, use OrderedDict for guaranteed order
- **Counter from empty**: `Counter()` is valid and returns empty; `Counter([])` too

```python
# WRONG: defaultdict(None) - None is not callable
# dd = defaultdict(None)  # TypeError!

# CORRECT: Use lambda for custom default
from collections import defaultdict
dd = defaultdict(lambda: 'unknown')
print(dd['missing'])
```

## Key Takeaways

2. `defaultdict(factory)` provides default values for missing keys
3. `OrderedDict` preserves insertion order; use `move_to_end()` for reordering
