## Project Overview

We'll build a task scheduler that:
- Stores tasks with **priority** (1 = highest, 5 = lowest)
- Uses **heapq** for O(log n) add and O(log n) get-next
- Supports **add_task**, **run_next**, **list_pending**
- Handles ties with a simple counter for FIFO within same priority

## Step 1: Task Representation

Each task is (priority, counter, description). Lower priority number = higher urgency. Counter breaks ties.

```python
import heapq

# (priority, tiebreaker, description)
# heapq is min-heap, so lower priority = popped first
task1 = (1, 0, "Fix critical bug")
task2 = (2, 1, "Review PR")
task3 = (1, 2, "Deploy hotfix")

heap = []
heapq.heappush(heap, task1)
heapq.heappush(heap, task2)
heapq.heappush(heap, task3)

print("Order of execution:")
while heap:
    prio, _, desc = heapq.heappop(heap)
    print(f"  Priority {prio}: {desc}")
```

## Step 2: TaskScheduler Class - Structure

Encapsulate the heap and a counter for tie-breaking.

```python
import heapq

class TaskScheduler:
    def __init__(self):
        self._heap = []
        self._counter = 0

    def add_task(self, description, priority=3):
        """Add a task. Priority 1=highest, 5=lowest."""
        self._counter += 1
        # (priority, counter, description) - counter ensures stable order
        heapq.heappush(self._heap, (priority, self._counter, description))

ts = TaskScheduler()
ts.add_task("Low priority task", priority=5)
ts.add_task("High priority task", priority=1)
print("Tasks added. Heap:", ts._heap)
```

## Step 3: run_next and list_pending

`run_next` pops and returns the highest-priority task. `list_pending` shows all without removing.

```python
import heapq

class TaskScheduler:
    def __init__(self):
        self._heap = []
        self._counter = 0

    def add_task(self, description, priority=3):
        self._counter += 1
        heapq.heappush(self._heap, (priority, self._counter, description))

    def run_next(self):
        """Run (remove and return) the next task. Returns None if empty."""
        if not self._heap:
            return None
        priority, _, description = heapq.heappop(self._heap)
        return {"priority": priority, "description": description}

    def list_pending(self):
        """Return sorted list of (priority, description) without removing."""
        return [(p, d) for p, _, d in sorted(self._heap)]

ts = TaskScheduler()
ts.add_task("Fix bug", priority=1)
ts.add_task("Write docs", priority=3)
ts.add_task("Refactor", priority=2)
print("Pending:", ts.list_pending())
print("Run next:", ts.run_next())
print("Pending after:", ts.list_pending())
```

## Step 5: Complete System - Full Code

All methods together with clear docstrings.

```python
import heapq

class TaskScheduler:
    """Priority-based task scheduler using heapq. Priority 1=highest, 5=lowest."""

    def __init__(self):
        self._heap = []
        self._counter = 0

    def add_task(self, description, priority=3):
        """Add a task with given priority."""
        self._counter += 1
        heapq.heappush(self._heap, (priority, self._counter, description))

    def run_next(self):
        """Run and return the highest-priority task. None if empty."""
        if not self._heap:
            return None
        priority, _, description = heapq.heappop(self._heap)
        return {"priority": priority, "description": description}

    def list_pending(self):
        """Return list of (priority, description) without removing."""
        return [(p, d) for p, _, d in sorted(self._heap)]

    def is_empty(self):
        return len(self._heap) == 0

    def __len__(self):
        return len(self._heap)
```

## Step 6: Run the Demo

Simulate a workflow: add tasks, run them in priority order.

```python
scheduler = TaskScheduler()

scheduler.add_task("Fix production bug", priority=1)
scheduler.add_task("Update documentation", priority=4)
scheduler.add_task("Code review", priority=2)
scheduler.add_task("Deploy to staging", priority=2)
scheduler.add_task("Plan next sprint", priority=5)

print("Pending tasks (by priority):")
for prio, desc in scheduler.list_pending():
    print(f"  [{prio}] {desc}")
```

```python
print("\nRunning tasks in order:")
while not scheduler.is_empty():
    task = scheduler.run_next()
    print(f"  Executing: {task['description']} (priority {task['priority']})")

print("\nAll tasks completed!")
```

## Step 7: Edge Cases

Handle empty scheduler and duplicate priorities.

```python
scheduler = TaskScheduler()

# Empty run
result = scheduler.run_next()
print("Run on empty:", result)

# Same priority - FIFO order
scheduler.add_task("First", priority=1)
scheduler.add_task("Second", priority=1)
scheduler.add_task("Third", priority=1)
print("Same priority order:", [scheduler.run_next()['description'] for _ in range(3)])
```
