## heapq Basics

`heapq` implements a min-heap on top of a list. The smallest element is always at index 0. Use `heappush`, `heappop`, `heapify`.

```python
import heapq

h = []
heapq.heappush(h, 5)
heapq.heappush(h, 2)
heapq.heappush(h, 8)
heapq.heappush(h, 1)
print(h)
print(heapq.heappop(h))
print(heapq.heappop(h))
print(h)
```

## heapify and nlargest/nsmallest

`heapify` converts a list in-place to a heap. `nlargest` and `nsmallest` return the n largest/smallest elements.

```python
import heapq

data = [3, 1, 4, 1, 5, 9, 2, 6]
heapq.heapify(data)
print("Heap:", data)
print("Smallest:", data[0])
print("3 largest:", heapq.nlargest(3, data))
print("3 smallest:", heapq.nsmallest(3, data))
```

## Priority Queue with heapq

For a priority queue, push tuples `(priority, item)`. Lower priority = higher urgency for min-heap.

```python
import heapq

pq = []
heapq.heappush(pq, (2, 'medium'))
heapq.heappush(pq, (1, 'high'))
heapq.heappush(pq, (3, 'low'))

while pq:
    priority, task = heapq.heappop(pq)
    print(f"Priority {priority}: {task}")
```

## heapreplace and heappushpop

`heapreplace(h, x)` pops smallest, then pushes x (more efficient than pop+push). `heappushpop(h, x)` pushes then pops smallest.

```python
import heapq

h = [1, 3, 5, 7]
heapq.heapify(h)
heapq.heapreplace(h, 2)
print("After heapreplace(h, 2):", h)
print("heappushpop(h, 0):", heapq.heappushpop(h, 0))
print(h)
```

## bisect: Binary Search

`bisect` provides binary search on sorted sequences. `bisect_left` and `bisect_right` return insertion points.

```python
import bisect

arr = [1, 3, 3, 5, 7, 9]
print(bisect.bisect_left(arr, 3))   # Leftmost position for 3
print(bisect.bisect_right(arr, 3))  # Rightmost position for 3
print(bisect.bisect_left(arr, 4))   # Where 4 would go
print(bisect.bisect_right(arr, 4))
```

## bisect: insort for sorted insertion

`insort_left` and `insort_right` insert while keeping the list sorted.

```python
import bisect

arr = [1, 3, 5, 7]
bisect.insort(arr, 4)
print(arr)
bisect.insort_left(arr, 3)
print(arr)
```

## bisect with key (Python 3.10+)

`bisect` functions accept a `key` parameter for custom comparison.

```python
import bisect

data = [{'x': 1}, {'x': 3}, {'x': 5}]
idx = bisect.bisect_left(data, 4, key=lambda d: d['x'])
print(f"Insert 4 at index {idx}")
```

## Practical: Merge sorted streams

Use heapq to merge multiple sorted iterables.

```python
import heapq

a = [1, 4, 7]
b = [2, 5, 8]
c = [3, 6, 9]
merged = list(heapq.merge(a, b, c))
print(merged)
```

## Practical: Find range in sorted list

Use bisect to find the range of indices where values fall in [lo, hi].

```python
import bisect

def find_range(arr, lo, hi):
    left = bisect.bisect_left(arr, lo)
    right = bisect.bisect_right(arr, hi)
    return arr[left:right]

arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(find_range(arr, 3, 6))
```

## Common Mistakes

- **heapq is min-heap**: For max-heap, negate priorities: `(-priority, item)`
- **Mutating heap**: After changing elements, call `heapify` again
- **bisect on unsorted list**: bisect assumes sorted input; wrong results otherwise
- **bisect_left vs bisect_right**: Use left for "first >= x", right for "first > x"

```python
# WRONG: heapq gives min-heap, so "high priority" = smallest number
# For max-heap (largest first), negate:
import heapq
h = []
heapq.heappush(h, -10)  # -10 = "highest" when we want 10 first
heapq.heappush(h, -5)
print(-heapq.heappop(h))  # 10
```

## Key Takeaways
